```python


# --- Make a toy dataframe ---
data = {
    'year1': [50, 120, 95, 130, 10, 2],
    'year2': [60, 80, 110, 140, 20, 2],
    'year3': [70, 90, 105, 150, 30, 2],
    'year4': [80, 100, 115, 160, 40, 2],
    'year5': [90, 110, 125, 170, 50, 2]
}

df = pd.DataFrame(data)

print("Full df:")
print(df)
```

```python
df.head()
```

```python
# Option 1: bracket notation (most common)
year3_col = df['year3']

# Option 2: using .loc with all rows, 'year3' column
year3_col_loc = df.loc[:, 'year3']

print("\nYear 3 column (by label):")
print(year3_col)
```
