```python
# -----------------------------
# Problem setting (same as MATLAB)
# -----------------------------
noc       = 10      # number of components
B         = 3       # number of subjects
Timesteps = 40      # observations (time)
Voxels    = 300     # dimensions (voxels)

# -----------------------------
# Data generation
# -----------------------------
subj = []
rng = np.random.default_rng(42)  # independent of opts['rngSEED'] (only for synthetic data)
for b in range(B):
    X = rng.standard_normal((Timesteps, Voxels)) ** 3         # zero-mean-ish heavy-tailed
    X = X - X.mean(axis=0, keepdims=True)                     # zero mean over time, per voxel
    sX = X                                                    # you can choose a smoothed/other version here
    subj.append(Subject(X=X, sX=sX))

# -----------------------------
# Optimization parameters
# -----------------------------
opts = {
    "maxiter":        100,
    "conv_crit":      1e-8,
    "fix_var_iter":   5,
    "use_gpu":        False,      # set True if you installed CuPy (pip install cupy-cuda12x)
    "heteroscedastic": True,
    "rngSEED":        12345,      # fixes random init (and FurthestSum seed)
    # "init":         "FurthestSum",  # default; or "random"
    # "numCstep":     10,
    # "numSstep":     20,
    # "initSstep":    250,
    # "noise_threshold": <float>,    # optional override
}

# -----------------------------
# Run MS-AA
# -----------------------------
results_subj, C, cost_fun, varexpl, time_taken = multi_subject_aa(subj, noc, opts)

print(f"Runtime (s): {time_taken:.3f}")
print(f"Variance explained: {varexpl*100:.2f}%")
print("C shape:", C.shape)
print("Subject 0: S shape", results_subj[0]["S"].shape, "sXC shape", results_subj[0]["sXC"].shape)

# -----------------------------
# Plot the negative log-likelihood
# -----------------------------
plt.figure()
plt.plot(cost_fun, marker="o", linewidth=1)
plt.title("Evolution of Negative Log-likelihood")
plt.xlabel(
    f"Iterations (The sharp dip near iter {opts['fix_var_iter']} is when\n"
    "voxel variance stops being fixed)"
)
plt.ylabel("NLL")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# -----------------------------
# Accessing outputs (same meaning as MATLAB comments)
# -----------------------------
# Shared archetypal generators:
#   C : (sV, noc)
# Per subject:
#   results_subj[b]["sXC"]   -> archetypes for subject b  (T_b, noc)
#   results_subj[b]["S"]     -> mixing matrix for subject b (noc, V)
#   results_subj[b]["sigmaSq"] -> voxel-specific noise (V, 1)

# Example: reconstruct subject 0 and compute Frobenius error
b = 0
recon0 = results_subj[b]["sXC"] @ results_subj[b]["S"]
fro_err0 = np.linalg.norm(subj[b].X - recon0, ord="fro")**2
print(f"Subject 0 SSE (recomputed): {fro_err0:.4f}")
print(f"Subject 0 SSE (from results): {results_subj[b]['SSE']:.4f}")
```

```python
# test with real data
pieman_name = 'data/pieman/raw/pieman_data.mat'
pieman_data = loadmat(pieman_name)
pieman_conds = ['intact', 'paragraph', 'word', 'rest']

posterior = loadmat('data/pieman/raw/pieman_posterior_K700.mat')
centers = posterior['posterior']['centers'][0][0][0][0][0]
widths = np.array(list(posterior['posterior']['widths'][0][0][0][0][0][:, 0].T))

data = []
conds = []
for c in pieman_conds:
    if c == 'paragraph':
        next_data = list(map(lambda i: pieman_data[c][:, i][0], np.where(np.arange(pieman_data[c].shape[1]) != 3)[0]))
    else:
        next_data = list(map(lambda i: pieman_data[c][:, i][0], np.arange(pieman_data[c].shape[1])))
    data.extend(next_data)
    conds.extend([c]*len(next_data))
```

```python
def coerce_TxV(x):
    x = np.asarray(x)
    if x.ndim != 2:
        raise ValueError(f"Expected 2D array, got {x.shape}")
    # We want T x V. If looks like V x T (V >> T), transpose it.
    T, V = x.shape
    if V < T:
        x = x.T  # now T x V
        T, V = x.shape
    # zero-mean per voxel (column) across time
    x = x - x.mean(axis=0, keepdims=True)
    return x.astype(np.float32, copy=False)
```

```python
subj = [Subject(X=coerce_TxV(X), sX=coerce_TxV(X)) for X in data]

opts = dict(maxiter=100, conv_crit=1e-6, fix_var_iter=5,
            heteroscedastic=True, use_gpu=False, rngSEED=12345)
K = 30
results_subj, C, cost_fun, varexpl, secs = multi_subject_aa(subj, noc=K, opts=opts)

print(f"MSAA runtime {secs:.1f}s | VarExpl {varexpl*100:.2f}%")
print("C:", C.shape)                                # (sV, K)
print("sXC (subj 0):", results_subj[0]["sXC"].shape)  # (T, K)  <-- temporal archetypes
print("S (subj 0):", results_subj[0]["S"].shape)      # (K, V)  <-- spatial maps
```
