```python
import matplotlib.pyplot as plt
from nilearn import plotting, datasets
from nibabel import freesurfer

# 1. Fetch standard brain surface meshes (fsaverage template)
fsaverage = datasets.fetch_surf_fsaverage()
# Load both left and right hemisphere meshes
mesh_left = fsaverage.pial_left
mesh_right = fsaverage.pial_right
# Use sulcal depth for shading the surface
bg_map_left = fsaverage.sulc_left
bg_map_right = fsaverage.sulc_right

# 2. Create a figure and a 3D axis manually
# Setting the facecolor and edgecolor to 'k' (black) for the figure itself
fig = plt.figure(figsize=(10, 8), facecolor='k', edgecolor='k')
ax = fig.add_subplot(111, projection='3d')

# 3. Plot both hemispheres using nilearn.plotting.plot_surf
# Pass the axis object to plot onto the specific 3D axis
# Set black_bg=True to ensure nilearn's internal plotting components use a black background
plotting.plot_surf(
    surf_mesh=mesh_left,
    hemi='left',
    bg_map=bg_map_left,
    bg_on_data=False, # Use background map for shading
    title="Full Brain 45-degree Profile View",
    axes=ax,
    colorbar=False
)

plotting.plot_surf(
    surf_mesh=mesh_right,
    hemi='right',
    bg_map=bg_map_right,
    bg_on_data=False,
    axes=ax,
    colorbar=False
)

# 4. Adjust the view angle (elevation and azimuth) to a 45-degree profile
# Default view is lateral (azimuth 0 or 180), we adjust it to a profile view.
# 'elev' is elevation, 'azim' is azimuth (rotation around Z axis).
ax.view_init(elev=15, azim=45) # Example 45 degree angle

# 5. Save the figure as a PDF file
# Use matplotlib's savefig method on the figure object
# Ensure facecolor and edgecolor are set to black again for savefig for a fully black background
plt.savefig("black_surface_plot_45deg.pdf", format='pdf', facecolor='k', edgecolor='k', bbox_inches='tight')
```

```python
from nilearn import datasets, surface, plotting
import numpy as np
import matplotlib.pyplot as plt

# --- Fetch fsaverage5 surfaces and sulcal background maps ---
fsavg = datasets.fetch_surf_fsaverage(mesh="fsaverage5")

mesh_left_path  = fsavg["pial_left"]
mesh_right_path = fsavg["pial_right"]
sulc_left_path  = fsavg["sulc_left"]
sulc_right_path = fsavg["sulc_right"]

# Load geometry + sulcal maps as arrays
left_mesh  = surface.load_surf_mesh(mesh_left_path)
right_mesh = surface.load_surf_mesh(mesh_right_path)
left_sulc  = surface.load_surf_data(sulc_left_path)
right_sulc = surface.load_surf_data(sulc_right_path)

# Handle nilearn versions that return a namedtuple vs (coords, faces)
def coords_faces(mesh):
    try:
        coords = np.asarray(mesh.coordinates)
        faces  = np.asarray(mesh.faces)
    except AttributeError:
        coords, faces = mesh  # tuple-like
        coords = np.asarray(coords)
        faces  = np.asarray(faces)
    return coords, faces

L_coords, L_faces = coords_faces(left_mesh)
R_coords, R_faces = coords_faces(right_mesh)

# Normalize sulcal depth to [0, 1] for interactive shading
def norm01(x):
    x = np.asarray(x, dtype=float)
    rng = np.ptp(x)
    return (x - x.min()) / (rng if rng > 0 else 1.0)

L_intensity = norm01(left_sulc)
R_intensity = norm01(right_sulc)

# --- 1) Interactive full-brain HTML (single scene) using plotly ---
# We use Mesh3d for each hemisphere and render them together.
import plotly.graph_objects as go

left_trace = go.Mesh3d(
    x=L_coords[:, 0], y=L_coords[:, 1], z=L_coords[:, 2],
    i=L_faces[:, 0],  j=L_faces[:, 1],  k=L_faces[:, 2],
    intensity=L_intensity, colorscale="Greys", reversescale=True,
    showscale=False, name="Left",
    lighting=dict(ambient=0.95, diffuse=0.9, specular=0.9, roughness=0.5),
)

right_trace = go.Mesh3d(
    x=R_coords[:, 0], y=R_coords[:, 1], z=R_coords[:, 2],
    i=R_faces[:, 0],  j=R_faces[:, 1],  k=R_faces[:, 2],
    intensity=R_intensity, colorscale="Greys", reversescale=True,
    showscale=False, name="Right",
    lighting=dict(ambient=0.95, diffuse=0.9, specular=0.9, roughness=0.5),
)

fig = go.Figure(data=[left_trace, right_trace])
fig.update_layout(
    title="Full Brain (fsaverage5) — interactive",
    scene=dict(
        xaxis_visible=False, yaxis_visible=False, zaxis_visible=False,
        aspectmode="data",
    ),
    margin=dict(l=0, r=0, t=40, b=0)
)

# Save as a self-contained HTML you can open anywhere
interactive_html = "full_brain_interactive.html"
fig.write_html(interactive_html, include_plotlyjs="cdn", full_html=True)
print(f"Saved interactive HTML -> {interactive_html}")

# Optional: open automatically in a browser
import webbrowser, pathlib
webbrowser.open(pathlib.Path(interactive_html).resolve().as_uri())

# --- 2) Static PDF export with nilearn (matplotlib engine) ---
# Lateral views of left and right hemispheres side-by-side.
fig_pdf = plt.figure(figsize=(10, 5), dpi=150)

ax_l = fig_pdf.add_subplot(1, 2, 1, projection="3d")
plotting.plot_surf(
    mesh_left_path, bg_map=sulc_left_path,
    hemi="left", view="anterior", cmap="bone_r", axes=ax_l,
)
ax_l.set_title("Left (lateral)")

ax_r = fig_pdf.add_subplot(1, 2, 2, projection="3d")
plotting.plot_surf(
    mesh_right_path, bg_map=sulc_right_path,
    hemi="right", view="anterior", cmap="bone_r", axes=ax_r,
)
ax_r.set_title("Right (lateral)")

plt.tight_layout()
pdf_path = "full_brain.pdf"
fig_pdf.savefig(pdf_path, bbox_inches="tight")
plt.close(fig_pdf)
print(f"Saved static PDF       -> {pdf_path}")
```
