
# Clean notebook: across-condition Spatial AA with decoding reported by condition

This notebook focuses on the workflow that makes the most sense for the paper:

1. Fit **one Spatial AA model across all subjects / all conditions**
2. Use that shared across-condition representation for all downstream analyses
3. Report **decoding separately within each condition**
4. Compute simple coefficient-space similarity clustering
5. Save brain plots safely without hanging the loop

## Key logic

- **Fit once across conditions**
- **Evaluate within condition**

That means:
- the archetypes / coefficients are learned jointly from all subjects
- but decoding is reported separately for:
  - intact
  - word
  - rest

This way the comparison is interpretable.


```

# =========================
# User configuration
# =========================

K_VALUES = [10, 25]              # start with [10] first, then expand
CONDITIONS_TO_USE = ["intact", "word", "rest"]
TRIM_REST_BY = 100

NREPS_DECODE = 20
NFOLDS_DECODE = 2
TOP_M_LIST = [1, 2, 3, 4, 5]
TOP_N_BRAINPLOTS = 5

RNG_SEED = 42

PIEMAN_MAT_PATH = "data/pieman/raw/pieman_data.mat"
POSTERIOR_MAT_PATH = "data/pieman/raw/pieman_posterior_K700.mat"
HELPERS_DIR = "helpers"
HYPERTOOLS_NORMALIZE_PATH = "hypertools/normalize.py"

OUTPUT_DIR = "msaa_clean_outputs_by_condition"

MSAA_OPTS = dict(
    maxiter=200,
    conv_crit=1e-6,
    fix_var_iter=5,
    heteroscedastic=False,
    use_gpu=False,
    rngSEED=RNG_SEED,
)
```

```

%matplotlib inline

import os
import sys
import time
import importlib.util

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from scipy.io import loadmat
from scipy.spatial.distance import cdist, squareform
from scipy.cluster.hierarchy import linkage, leaves_list, fcluster
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score

HT_NORMALIZE_AVAILABLE = False
try:
    spec = importlib.util.spec_from_file_location("normalize", HYPERTOOLS_NORMALIZE_PATH)
    ht = importlib.util.module_from_spec(spec)
    spec.loader.exec_module(ht)
    HT_NORMALIZE_AVAILABLE = True
    print("Loaded hypertools normalize.")
except Exception:
    print("hypertools normalize not available; using fallback normalization.")
    ht = None

sys.path.append(HELPERS_DIR)
from MultiSubject_AA import Subject, multi_subject_aa

try:
    from nilearn import plotting as niplot
    NILEARN_AVAILABLE = True
except Exception:
    NILEARN_AVAILABLE = False
    print("nilearn not available; brain plotting disabled.")

os.makedirs(OUTPUT_DIR, exist_ok=True)
print("Ready.")
```

```

def array_cutter(data_list, cut_length):
    out = [np.asarray(x).copy() for x in data_list]
    if cut_length <= 0:
        return out
    return [x[:-cut_length] for x in out]

def coerce_TxV(x):
    x = np.asarray(x)
    if x.ndim != 2:
        raise ValueError(f"Expected 2D array, got {x.shape}")
    T, V = x.shape
    if V < T:
        x = x.T
    x = x - x.mean(axis=0, keepdims=True)
    return x.astype(np.float32, copy=False)

def normalize_across_fallback(subject_list, eps=1e-8):
    subj = [coerce_TxV(x) for x in subject_list]
    stacked = np.vstack(subj)
    mu = stacked.mean(axis=0, keepdims=True)
    sd = stacked.std(axis=0, keepdims=True) + eps
    return [((x - mu) / sd).astype(np.float32, copy=False) for x in subj]

def normalize_subject_list(subject_list):
    subj = [coerce_TxV(x) for x in subject_list]
    if HT_NORMALIZE_AVAILABLE:
        try:
            return ht.normalize(subj, normalize="across")
        except Exception:
            return normalize_across_fallback(subj)
    return normalize_across_fallback(subj)

def load_pieman_data(mat_path=PIEMAN_MAT_PATH, conditions_to_use=CONDITIONS_TO_USE, trim_rest_by=TRIM_REST_BY):
    pieman_data = loadmat(mat_path)
    data = []
    conds = []
    for c in conditions_to_use:
        if c == "paragraph":
            next_data = list(map(lambda i: pieman_data[c][:, i][0],
                                 np.where(np.arange(pieman_data[c].shape[1]) != 3)[0]))
        else:
            next_data = list(map(lambda i: pieman_data[c][:, i][0],
                                 np.arange(pieman_data[c].shape[1])))
        data.extend(next_data)
        conds.extend([c] * len(next_data))

    conds_array = np.array(conds)
    condition_data = {}
    for c in conditions_to_use:
        cur = [data[i] for i in np.where(conds_array == c)[0]]
        if c == "rest" and trim_rest_by > 0:
            cur = array_cutter(cur, trim_rest_by)
        condition_data[c] = normalize_subject_list(cur)

    all_subjects = []
    all_condition_labels = []
    subject_condition_idx = []
    for c in conditions_to_use:
        for x in condition_data[c]:
            all_subjects.append(x)
            all_condition_labels.append(c)
            subject_condition_idx.append(c)

    return condition_data, np.array(all_condition_labels), all_subjects

condition_data, condition_labels_str, all_subjects = load_pieman_data()

condition_names = sorted(np.unique(condition_labels_str))
condition_to_code = {c: i for i, c in enumerate(condition_names)}
condition_map = {i: c for c, i in condition_to_code.items()}
condition_codes = np.array([condition_to_code[c] for c in condition_labels_str])

print("Conditions loaded:")
for c, arrs in condition_data.items():
    print(f"  {c}: n={len(arrs)}, example shape={coerce_TxV(arrs[0]).shape}")
print(f"Total subjects: {len(all_subjects)}")
print("Condition map:", condition_map)
```

```

def load_node_centers(posterior_mat_path=POSTERIOR_MAT_PATH):
    posterior = loadmat(posterior_mat_path)
    centers = posterior["posterior"]["centers"][0][0][0][0][0]
    centers = np.asarray(centers, dtype=float)
    if centers.shape[1] != 3 and centers.shape[0] == 3:
        centers = centers.T
    if centers.shape[1] != 3:
        raise ValueError(f"Expected centers to be V x 3, got {centers.shape}")
    return centers

centers = load_node_centers()
print("centers shape:", centers.shape)
```

## Fit one across-condition Spatial AA model per K

```

def build_subject_objects_spatial(subject_list):
    subj_objs = []
    for X in subject_list:
        Xtv = coerce_TxV(X)
        subj_objs.append(Subject(X=Xtv, sX=Xtv))
    return subj_objs

def fit_spatial_msaa(subject_list, K, opts=None, verbose=True):
    opts = dict(MSAA_OPTS if opts is None else opts)
    subj_objs = build_subject_objects_spatial(subject_list)

    t0 = time.time()
    results_subj, C, cost_fun, varexpl, secs = multi_subject_aa(
        subj_objs,
        noc=K,
        opts=opts,
    )
    elapsed = time.time() - t0

    if verbose:
        print(f"[Across-condition Spatial AA] K={K} | runtime={secs:.1f}s (wall={elapsed:.1f}s) | VarExpl={varexpl*100:.2f}%")
        if len(results_subj) > 0:
            print("subj0 keys:", list(results_subj[0].keys()))
            print("subj0 sXC shape:", np.asarray(results_subj[0]['sXC']).shape)
            print("subj0 S shape:", np.asarray(results_subj[0]['S']).shape)

    return {
        "results_subj": results_subj,
        "C": C,
        "cost_fun": cost_fun,
        "varexpl": varexpl,
        "secs": secs,
        "K": K,
    }

def fit_spatial_msaa_sweep(subject_list, k_values, opts=None, verbose=True):
    out = {}
    for K in k_values:
        out[K] = fit_spatial_msaa(subject_list, K, opts=opts, verbose=verbose)
    return out

results_by_k = fit_spatial_msaa_sweep(
    subject_list=all_subjects,
    k_values=K_VALUES,
    opts=MSAA_OPTS,
    verbose=True,
)
```

```

def print_spatial_diagnostics(results_subj, original_subjects, n_show=5):
    s0 = results_subj[0]
    X0 = coerce_TxV(original_subjects[0])
    T, V = X0.shape
    sXC = np.asarray(s0["sXC"], dtype=float)
    S = np.asarray(s0["S"], dtype=float)

    if sXC.shape[0] != T and sXC.shape[1] == T:
        sXC = sXC.T
    if S.shape[1] != V and S.shape[0] == V:
        S = S.T

    print("sXC shape (should be T x K):", sXC.shape)
    print("S shape   (should be K x V):", S.shape)
    print("S min/max:", float(np.min(S)), float(np.max(S)))
    print("sum across archetypes for first nodes:", np.sum(S, axis=0)[:n_show])

print_spatial_diagnostics(results_by_k[K_VALUES[0]]["results_subj"], all_subjects)
```

## Decoding helpers

```

def decoder(corrs):
    out = pd.DataFrame({'rank':[0.0], 'accuracy':[0.0], 'error':[0.0]})
    T = corrs.shape[0]
    for t in range(T):
        decoded_ind = int(np.argmax(corrs[t, :]))
        out.loc[0, 'error'] += np.mean(np.abs(decoded_ind - t)) / T
        out.loc[0, 'accuracy'] += (decoded_ind == t)
        out.loc[0, 'rank'] += np.mean((corrs[t, :] <= corrs[t, t]).astype(int))
    out['error'] /= T
    out['accuracy'] /= T
    out['rank'] /= T
    return out

def get_xval_assignments(ndata, nfolds, rng=None):
    rng = np.random.default_rng() if rng is None else rng
    group_assignments = np.zeros(ndata, dtype=int)
    groupsize = int(np.ceil(ndata / nfolds))
    for i in range(1, nfolds):
        inds = np.arange(i * groupsize, min((i + 1) * groupsize, ndata))
        group_assignments[inds] = i
    rng.shuffle(group_assignments)
    return group_assignments

def build_coeff_recon_stack(subjects, top_indices=None):
    Xhats = []
    for sub in subjects:
        sXC = np.asarray(sub["sXC"], dtype=float)
        S = np.asarray(sub["S"], dtype=float)
        if top_indices is None:
            Xhat = sXC @ S
        else:
            Xhat = sXC[:, top_indices] @ S[top_indices, :]
        Xhat = (Xhat - Xhat.mean(axis=0, keepdims=True)) / (Xhat.std(axis=0, keepdims=True) + 1e-8)
        Xhats.append(Xhat)
    return np.stack(Xhats, axis=0)

def run_timepoint_decoding(recon_stack, nfolds=2, nreps=20, seed=42):
    N, T, V = recon_stack.shape
    rng = np.random.default_rng(seed)
    all_results = []
    for rep in range(nreps):
        fold_ids = get_xval_assignments(N, nfolds, rng=rng)
        for i in range(nfolds):
            in_mask = (fold_ids == i)
            out_mask = ~in_mask
            in_mean = recon_stack[in_mask].mean(axis=0)
            out_mean = recon_stack[out_mask].mean(axis=0)
            corrs = 1.0 - cdist(in_mean, out_mean, metric='correlation')
            res = decoder(corrs)
            res["rep"] = rep
            res["fold"] = i
            all_results.append(res)
    return pd.concat(all_results, ignore_index=True)

def per_archetype_coeff_decoding(subjects, nfolds=2, nreps=20, seed=42):
    K = np.asarray(subjects[0]["S"]).shape[0]
    rows = []
    for k in range(K):
        recon_stack = build_coeff_recon_stack(subjects, top_indices=[k])
        res = run_timepoint_decoding(recon_stack, nfolds=nfolds, nreps=nreps, seed=seed)
        res["archetype"] = k
        rows.append(res)
    return pd.concat(rows, ignore_index=True)

def run_timepoint_decoding_topm(subjects, ranked_indices, top_ms=[1,2,3,4,5], nfolds=2, nreps=20, seed=42):
    rows = []
    for m in top_ms:
        top_idx = ranked_indices[:m]
        recon_stack = build_coeff_recon_stack(subjects, top_indices=top_idx)
        res = run_timepoint_decoding(recon_stack, nfolds=nfolds, nreps=nreps, seed=seed)
        res["top_m"] = m
        rows.append(res)
    return pd.concat(rows, ignore_index=True)
```

## Condition-specific decoding from the across-condition fit

```

def subjects_for_condition(results_subj, condition_labels_str, cond_name):
    idx = np.where(np.asarray(condition_labels_str) == cond_name)[0]
    return [results_subj[i] for i in idx], idx

def decode_by_condition(results_subj, condition_labels_str, top_m_list=TOP_M_LIST,
                        nfolds=NFOLDS_DECODE, nreps=NREPS_DECODE, seed=RNG_SEED):
    out = {}
    K = np.asarray(results_subj[0]["S"]).shape[0]

    # global ranking across all subjects in the across-condition fit
    per_arch_all = per_archetype_coeff_decoding(
        results_subj,
        nfolds=nfolds,
        nreps=nreps,
        seed=seed,
    )
    top_arch_global = (
        per_arch_all.groupby("archetype")["accuracy"]
        .mean()
        .sort_values(ascending=False)
    )

    for cond_name in np.unique(condition_labels_str):
        sub_cond, idx = subjects_for_condition(results_subj, condition_labels_str, cond_name)

        per_arch_cond = per_archetype_coeff_decoding(
            sub_cond,
            nfolds=nfolds,
            nreps=nreps,
            seed=seed,
        )

        top_arch_cond = (
            per_arch_cond.groupby("archetype")["accuracy"]
            .mean()
            .sort_values(ascending=False)
        )

        subset_cond = run_timepoint_decoding_topm(
            sub_cond,
            ranked_indices=top_arch_global.index.to_numpy(),   # use globally ranked across-condition archetypes
            top_ms=top_m_list,
            nfolds=nfolds,
            nreps=nreps,
            seed=seed,
        )

        out[cond_name] = {
            "subject_indices": idx,
            "per_arch_cond": per_arch_cond,
            "top_arch_cond": top_arch_cond,
            "subset_cond": subset_cond,
        }

    return {
        "per_arch_all": per_arch_all,
        "top_arch_global": top_arch_global,
        "by_condition": out,
    }
```

## Simple clustering similarity on coefficient vectors

```

def purity_score(y_cluster, y_class):
    y_cluster = np.asarray(y_cluster)
    y_class = np.asarray(y_class)
    total = 0
    for c in np.unique(y_cluster):
        mask = (y_cluster == c)
        _, counts = np.unique(y_class[mask], return_counts=True)
        total += counts.max()
    return total / len(y_cluster)

def cluster_archetype_from_similarity(Xk, cond, n_clusters=3, method="average"):
    sim = np.corrcoef(Xk)
    dist = 1 - sim
    np.fill_diagonal(dist, 0)
    Z = linkage(squareform(dist, checks=False), method=method)
    order = leaves_list(Z)
    labels = fcluster(Z, n_clusters, criterion="maxclust")
    sim_sorted = sim[np.ix_(order, order)]
    labels_sorted = labels[order]
    cond_sorted = np.asarray(cond)[order]
    boundaries = np.where(np.diff(labels_sorted) != 0)[0] + 1
    return {
        "sim": sim,
        "sim_sorted": sim_sorted,
        "order": order,
        "labels": labels,
        "labels_sorted": labels_sorted,
        "cond_sorted": cond_sorted,
        "boundaries": boundaries,
        "purity": purity_score(labels, cond),
        "ari": adjusted_rand_score(cond, labels),
        "nmi": normalized_mutual_info_score(cond, labels),
    }

def scan_archetypes_simple_similarity(results_subj, cond, n_clusters=3):
    K = np.asarray(results_subj[0]["S"]).shape[0]
    rows = []
    results = {}
    for k in range(K):
        Xk = np.stack([np.asarray(sub["S"])[k, :] for sub in results_subj], axis=0)
        result = cluster_archetype_from_similarity(Xk, cond, n_clusters=n_clusters, method="average")
        rows.append({
            "k": k,
            "purity": result["purity"],
            "ari": result["ari"],
            "nmi": result["nmi"],
        })
        results[k] = result
    df = pd.DataFrame(rows).sort_values("ari", ascending=False).reset_index(drop=True)
    return df, results

def plot_archetype_similarity(result, cond_colors=None, title="", show_subject_ids=True):
    if cond_colors is None:
        cond_colors = {0: "purple", 1: "green", 2: "black"}
    order = result["order"]
    cond_sorted = result["cond_sorted"]
    xticklabels = (order + 1) if show_subject_ids else False
    yticklabels = (order + 1) if show_subject_ids else False
    plt.figure(figsize=(7, 6))
    ax = sns.heatmap(
        result["sim_sorted"],
        cmap="coolwarm",
        vmin=-1.0,
        vmax=1.0,
        square=True,
        xticklabels=xticklabels,
        yticklabels=yticklabels,
        cbar_kws={"label": "Subject correlation"},
    )
    if show_subject_ids:
        tick_colors = [cond_colors.get(int(c), "gray") for c in cond_sorted]
        for tick_label, color in zip(ax.get_xticklabels(), tick_colors):
            tick_label.set_color(color)
            tick_label.set_rotation(90)
            tick_label.set_fontsize(8)
        for tick_label, color in zip(ax.get_yticklabels(), tick_colors):
            tick_label.set_color(color)
            tick_label.set_fontsize(8)
    for b in result["boundaries"]:
        ax.axhline(b, color="black", linewidth=2)
        ax.axvline(b, color="black", linewidth=2)
    plt.title(f"{title}\nPur={result['purity']:.2f}, ARI={result['ari']:.2f}, NMI={result['nmi']:.2f}")
    plt.xlabel("Subjects (clustered)")
    plt.ylabel("Subjects (clustered)")
    plt.tight_layout()
    plt.show()
```

## Safe brain plotting (save and close only)

```

def save_top_archetype_brainplots(
    S_group_mean,
    archetype_indices,
    centers,
    out_dir,
    prefix,
    rel_thresh=0.3,
    display_mode="lyrz",
    node_size=10,
):
    if not NILEARN_AVAILABLE:
        print("nilearn not available; skipping brain plots.")
        return []
    os.makedirs(out_dir, exist_ok=True)
    saved = []
    for k in archetype_indices:
        weights = np.asarray(S_group_mean[k], dtype=float).ravel()
        wmax = weights.max()
        keep = weights >= rel_thresh * wmax
        centers_sel = centers[keep]
        fig = plt.figure(figsize=(8, 6))
        display = niplot.plot_connectome(
            np.eye(len(centers_sel)),
            centers_sel,
            node_color="red",
            node_size=node_size,
            display_mode=display_mode,
            title=f"{prefix} | archetype {k}",
            figure=fig,
        )
        out_path = os.path.join(out_dir, f"{prefix}_arch_{k}.png")
        fig.savefig(out_path, dpi=150, bbox_inches="tight")
        try:
            display.close()
        except Exception:
            pass
        plt.close(fig)
        saved.append(out_path)
    return saved
```

## Small-step workflow: test one K

```

K_TO_TEST = K_VALUES[0]
cur = results_by_k[K_TO_TEST]

decode_res = decode_by_condition(
    cur["results_subj"],
    condition_labels_str=condition_labels_str,
    top_m_list=TOP_M_LIST,
    nfolds=NFOLDS_DECODE,
    nreps=NREPS_DECODE,
    seed=RNG_SEED,
)

print("Global top archetypes from across-condition fit:")
display(decode_res["top_arch_global"].head(10).to_frame("mean_accuracy"))

for cond_name, d in decode_res["by_condition"].items():
    print(f"\n===== {cond_name} =====")
    print("Top archetypes by within-condition decoding:")
    display(d["top_arch_cond"].head(10).to_frame("mean_accuracy"))
    print("Top-m decoding summary (using globally ranked archetypes):")
    display(d["subset_cond"].groupby("top_m")[["accuracy", "rank", "error"]].mean())
```

## Across-condition similarity clustering on coefficient vectors

```

cluster_df, cluster_results = scan_archetypes_simple_similarity(
    cur["results_subj"],
    cond=condition_codes,
    n_clusters=3,
)

print("Top archetypes by ARI:")
display(cluster_df.head(10))
```

```

top_k = int(cluster_df.iloc[0]["k"])
plot_archetype_similarity(
    cluster_results[top_k],
    cond_colors={0: "purple", 1: "green", 2: "black"},
    title=f"K={K_TO_TEST} | archetype {top_k} | across-condition coefficient similarity clustering"
)
```

## Save brain plots for top globally decoding archetypes

```

freq_selected_k = decode_res["top_arch_global"].head(TOP_N_BRAINPLOTS).index.to_numpy()
S_group_mean = np.mean([np.asarray(sub["S"], dtype=float) for sub in cur["results_subj"]], axis=0)

brainplot_dir = os.path.join(OUTPUT_DIR, f"K_{K_TO_TEST}", "brainplots")
saved_files = save_top_archetype_brainplots(
    S_group_mean=S_group_mean,
    archetype_indices=freq_selected_k,
    centers=centers,
    out_dir=brainplot_dir,
    prefix=f"spatialAA_acrossCond_K{K_TO_TEST}",
)

print("Saved brain plots:")
for f in saved_files:
    print(" ", f)
```

## Run the same workflow across all K values

```

all_summary = {}

for K, cur in results_by_k.items():
    print(f"\n===== K = {K} =====")

    decode_res = decode_by_condition(
        cur["results_subj"],
        condition_labels_str=condition_labels_str,
        top_m_list=TOP_M_LIST,
        nfolds=NFOLDS_DECODE,
        nreps=NREPS_DECODE,
        seed=RNG_SEED,
    )

    cluster_df, cluster_results = scan_archetypes_simple_similarity(
        cur["results_subj"],
        cond=condition_codes,
        n_clusters=3,
    )

    brainplot_dir = os.path.join(OUTPUT_DIR, f"K_{K}", "brainplots")
    S_group_mean = np.mean([np.asarray(sub["S"], dtype=float) for sub in cur["results_subj"]], axis=0)
    saved_files = save_top_archetype_brainplots(
        S_group_mean=S_group_mean,
        archetype_indices=decode_res["top_arch_global"].head(TOP_N_BRAINPLOTS).index.to_numpy(),
        centers=centers,
        out_dir=brainplot_dir,
        prefix=f"spatialAA_acrossCond_K{K}",
    )

    all_summary[K] = {
        "decode_res": decode_res,
        "cluster_df": cluster_df,
        "cluster_results": cluster_results,
        "saved_brainplots": saved_files,
    }

    print("Best global decoding archetypes:")
    print(decode_res["top_arch_global"].head(5))

    for cond_name, d in decode_res["by_condition"].items():
        print(f"{cond_name} top-m decoding mean accuracy:")
        print(d["subset_cond"].groupby("top_m")["accuracy"].mean())

    print("Top 5 clustering archetypes by ARI:")
    print(cluster_df.head(5))
    print(f"Saved {len(saved_files)} brain plots.")
```

## Summaries across K

```

summary_rows = []
for K, res in all_summary.items():
    row = {
        "K": K,
        "best_cluster_ari": float(res["cluster_df"]["ari"].max()),
        "best_cluster_purity": float(res["cluster_df"]["purity"].max()),
        "n_saved_brainplots": len(res["saved_brainplots"]),
    }
    for cond_name, d in res["decode_res"]["by_condition"].items():
        row[f"{cond_name}_best_single_arch_acc"] = float(d["top_arch_cond"].iloc[0])
        row[f"{cond_name}_best_topm_acc"] = float(d["subset_cond"].groupby("top_m")["accuracy"].mean().max())
    summary_rows.append(row)

summary_df = pd.DataFrame(summary_rows).sort_values("K").reset_index(drop=True)
display(summary_df)
```
