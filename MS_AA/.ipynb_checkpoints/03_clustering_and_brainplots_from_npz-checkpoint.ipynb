# 03_clustering_and_brainplots_from_npz.ipynb

Run this third.

```

LOAD_DIR = "msaa_outputs_npz"
K_VALUES = [10, 25]
N_CLUSTERS = 3
TOP_N_CLUSTER_ARCHETYPES = 3
COND_COLORS = {0: "purple", 1: "green", 2: "black"}
```

```

%matplotlib inline
import os
import numpy as np, pandas as pd, seaborn as sns, matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, leaves_list, fcluster
from scipy.spatial.distance import squareform
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score
try:
    from nilearn import plotting as niplot
    NILEARN_AVAILABLE = True
except Exception:
    NILEARN_AVAILABLE = False
    print("nilearn not available; brain plotting disabled.")
```

```

def load_msaa_npz(path):
    data = np.load(path, allow_pickle=True)
    results_subj = data["results_subj"].tolist()
    if isinstance(results_subj, np.ndarray):
        results_subj = results_subj.tolist()
    lookup_table = dict(zip(data["lookup_table_keys"].tolist(), data["lookup_table_vals"].tolist()))
    network_colors = dict(zip(data["network_color_keys"].tolist(), data["network_color_vals"].tolist()))
    return {
        "K": int(data["K"]),
        "results_subj": results_subj,
        "condition_labels_str": data["condition_labels_str"].tolist(),
        "condition_codes": data["condition_codes"],
        "centers": data["centers"],
        "widths": int(data["widths"]),
        "networks_cmu": pd.DataFrame(data["networks_cmu"].tolist()),
        "node_codes": pd.DataFrame(data["node_codes"].tolist()),
        "lookup_table": lookup_table,
        "network_colors": network_colors,
    }

loaded = {}
for K in K_VALUES:
    path = os.path.join(LOAD_DIR, f"spatialAA_acrossCond_K{K}.npz")
    loaded[K] = load_msaa_npz(path)
    print("Loaded:", path)
```

```

def purity_score(y_cluster, y_class):
    y_cluster = np.asarray(y_cluster)
    y_class = np.asarray(y_class)
    total = 0
    for c in np.unique(y_cluster):
        mask = (y_cluster == c)
        _, counts = np.unique(y_class[mask], return_counts=True)
        total += counts.max()
    return total / len(y_cluster)

def get_timecourse_archetype_subject_matrix(results_subj, k, zscore_each=True, eps=1e-8):
    Xk = np.stack([np.asarray(sub["sXC"], dtype=float)[:, k] for sub in results_subj], axis=0)
    if zscore_each:
        mu = Xk.mean(axis=1, keepdims=True)
        sd = Xk.std(axis=1, keepdims=True)
        Xk = (Xk - mu) / (sd + eps)
    Xk = np.nan_to_num(Xk, nan=0.0, posinf=0.0, neginf=0.0)
    return Xk

def cluster_from_timecourse_similarity(Xk, cond, n_clusters=3, linkage_method="average"):
    Xk = np.asarray(Xk, dtype=float)
    Xk = np.nan_to_num(Xk, nan=0.0, posinf=0.0, neginf=0.0)
    sim = np.corrcoef(Xk)
    sim = np.nan_to_num(sim, nan=0.0, posinf=0.0, neginf=0.0)
    sim = np.clip(sim, -1.0, 1.0)
    np.fill_diagonal(sim, 1.0)
    dist = 1.0 - sim
    dist = np.nan_to_num(dist, nan=1.0, posinf=1.0, neginf=1.0)
    np.fill_diagonal(dist, 0.0)
    Z = linkage(squareform(dist, checks=False), method=linkage_method)
    order = leaves_list(Z)
    labels = fcluster(Z, n_clusters, criterion="maxclust")
    sim_sorted = sim[np.ix_(order, order)]
    cond_sorted = np.asarray(cond)[order]
    labels_sorted = labels[order]
    boundaries = np.where(np.diff(labels_sorted) != 0)[0] + 1
    return {
        "sim": sim,
        "sim_sorted": sim_sorted,
        "order": order,
        "labels": labels,
        "labels_sorted": labels_sorted,
        "cond_sorted": cond_sorted,
        "boundaries": boundaries,
        "purity": purity_score(labels, cond),
        "ari": adjusted_rand_score(cond, labels),
        "nmi": normalized_mutual_info_score(cond, labels),
    }

def scan_archetypes_timecourse_similarity(results_subj, cond, n_clusters=3, linkage_method="average"):
    K = np.asarray(results_subj[0]["sXC"]).shape[1]
    rows, results = [], {}
    for k in range(K):
        Xk = get_timecourse_archetype_subject_matrix(results_subj, k, zscore_each=True)
        result = cluster_from_timecourse_similarity(Xk, cond=cond, n_clusters=n_clusters, linkage_method=linkage_method)
        rows.append({"k": k, "purity": result["purity"], "ari": result["ari"], "nmi": result["nmi"]})
        results[k] = result
    df = pd.DataFrame(rows).sort_values("ari", ascending=False).reset_index(drop=True)
    return df, results

def plot_timecourse_similarity(result, cond_colors=None, title="", show_subject_ids=True):
    if cond_colors is None:
        cond_colors = {0: "purple", 1: "green", 2: "black"}
    order = result["order"]
    cond_sorted = result["cond_sorted"]
    xticklabels = (order + 1) if show_subject_ids else False
    yticklabels = (order + 1) if show_subject_ids else False
    plt.figure(figsize=(7, 6))
    ax = sns.heatmap(
        result["sim_sorted"], cmap="coolwarm", vmin=-1, vmax=1, square=True,
        xticklabels=xticklabels, yticklabels=yticklabels,
        cbar_kws={"label": "subject correlation of archetype timecourses"},
    )
    if show_subject_ids:
        tick_colors = [cond_colors.get(int(c), "gray") for c in cond_sorted]
        for tick_label, color in zip(ax.get_xticklabels(), tick_colors):
            tick_label.set_color(color); tick_label.set_rotation(90); tick_label.set_fontsize(8)
        for tick_label, color in zip(ax.get_yticklabels(), tick_colors):
            tick_label.set_color(color); tick_label.set_fontsize(8)
    for b in result["boundaries"]:
        ax.axhline(b, color="black", linewidth=2)
        ax.axvline(b, color="black", linewidth=2)
    plt.title(f"{title}\nPur={result['purity']:.2f}, ARI={result['ari']:.2f}, NMI={result['nmi']:.2f}")
    plt.tight_layout()
    plt.show()

def plot_network_colored_spatial_coeff_map(results_subj, k, centers, node_codes, network_colors, title=""):
    if not NILEARN_AVAILABLE:
        print("nilearn not available; skipping.")
        return None
    color_list = [None] + list(network_colors.values())
    coeffs = np.stack([np.asarray(sub["S"], dtype=float)[k, :] for sub in results_subj], axis=0)
    weights = coeffs.mean(axis=0)
    weights = np.nan_to_num(weights, nan=0.0, posinf=0.0, neginf=0.0)
    thr = 0.30 * np.max(weights)
    keep = weights >= thr
    centers_sel = centers[keep]
    codes_sel = node_codes.loc[keep, "code"].to_numpy()
    fig = plt.figure(figsize=(6, 5))
    display = niplot.plot_connectome(
        np.eye(centers_sel.shape[0]), centers_sel,
        node_size=20,
        node_color=[color_list[i] for i in codes_sel],
        display_mode="lyrz",
        title=title,
        figure=fig,
    )
    plt.show()
    try:
        display.close()
    except Exception:
        pass
    plt.close(fig)
```

```

top_cluster_ks_by_k = {}
for K in K_VALUES:
    cur = loaded[K]
    cluster_df, cluster_results = scan_archetypes_timecourse_similarity(
        cur["results_subj"],
        cond=cur["condition_codes"],
        n_clusters=N_CLUSTERS,
        linkage_method="average",
    )
    top_cluster_ks = cluster_df.head(TOP_N_CLUSTER_ARCHETYPES)["k"].astype(int).tolist()
    top_cluster_ks_by_k[K] = top_cluster_ks

    print(f"\n===== K={K} =====")
    display(cluster_df.head(10))
    print("Top 3 clustering archetypes:", top_cluster_ks)

    top_k = top_cluster_ks[0]
    plot_timecourse_similarity(
        cluster_results[top_k],
        cond_colors=COND_COLORS,
        title=f"K={K} | archetype {top_k} | timecourse clustering"
    )

    plot_network_colored_spatial_coeff_map(
        cur["results_subj"],
        top_k,
        centers=cur["centers"],
        node_codes=cur["node_codes"],
        network_colors=cur["network_colors"],
        title=f"K={K} | archetype {top_k} | spatial coefficients"
    )

top_cluster_ks_by_k
```
