
# 04_spatialAA_within_decoding_inspection_from_flexible_outputs.ipynb

This notebook loads the outputs from the **flexible decoding pipeline** and the saved
**within-condition spatial AA fits**, then generates:

- per-archetype decoding summaries
- archetype timecourse plots
- coefficient brain plots
- network pie charts

## Assumptions

This notebook assumes:

### Fit outputs
Saved from the flexible fit notebook/script as:
- `spatialAA_within_intact_K{K}.npz`
- `spatialAA_within_word_K{K}.npz`
- `spatialAA_within_rest_K{K}.npz`

### Decoding outputs
Saved from the flexible decoding notebook/script in:
- `msaa_flexible_decoding_outputs/`

and in particular:
- `rankings.npy`
- optionally `per_archetype_mean_accuracy.csv`

## Spatial AA convention

For within-condition **spatial AA**:
- `sXC`: **T x K** shared temporal archetypes / motifs
- `S`: **K x V** spatial coefficients

Reconstruction:
\[
\hat X = sXC \cdot S
\]

## What gets plotted

For each selected condition / K / archetype:
1. decoding bar
2. archetype timecourse plot (`sXC[:, k]`)
3. coefficient brain plot (`S[k, :]`)
4. network pie chart from the coefficient weights


```

# =========================
# USER SETTINGS
# =========================

FIT_LOAD_DIR = "msaa_flexible_outputs_npz"
DECODE_LOAD_DIR = "msaa_flexible_decoding_outputs"

CONDITIONS = ["intact", "word", "rest"]
K_VALUES = [10, 25, 50]

TOP_N_REPORT = 5
USE_RANKINGS_FILE = True     # if False, rank from per_archetype_mean_accuracy.csv

# If you want to manually override:
# MANUAL_TOP_ARCHETYPES = {
#     ("intact", 50): [0, 4, 1, 8, 3],
# }
MANUAL_TOP_ARCHETYPES = {}

SCHAEFER_TXT = "data/pieman/raw/Schaefer2018_1000Parcels_7Networks_order.txt"
SCHAEFER_NII = "data/pieman/raw/Schaefer2018_1000Parcels_7Networks_order_FSLMNI152_2mm.nii.gz"
POSTERIOR_MAT = "data/pieman/raw/pieman_posterior_K700.mat"

COND_COLORS = {"intact": "purple", "word": "green", "rest": "black"}

OUTPUT_DIR = "spatialAA_within_decoding_inspection_outputs"
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```

%matplotlib inline
import os, warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import loadmat
from matplotlib.colors import to_rgba

try:
    import nibabel as nib
    from nilearn.input_data import NiftiMasker
    from nilearn import plotting as niplot
    NILEARN_AVAILABLE = True
except Exception:
    NILEARN_AVAILABLE = False
    print("nilearn / nibabel not available; brain plotting disabled.")

print("Ready.")
```

```

def to_float_array(x):
    return np.array(x, dtype=float)

def load_msaa_npz(path):
    data = np.load(path, allow_pickle=True)
    results_subj = data["results_subj"].tolist()
    if isinstance(results_subj, np.ndarray):
        results_subj = results_subj.tolist()
    return {
        "K": int(data["K"]),
        "results_subj": results_subj,
        "condition_labels_str": data["condition_labels_str"].tolist(),
        "condition_codes": data["condition_codes"] if "condition_codes" in data else None,
    }

fits = {}
for cond_name in CONDITIONS:
    fits[cond_name] = {}
    for K in K_VALUES:
        path = os.path.join(FIT_LOAD_DIR, f"spatialAA_within_{cond_name}_K{K}.npz")
        fits[cond_name][K] = load_msaa_npz(path)
        print("Loaded fit:", path)

example_sub = fits[CONDITIONS[0]][K_VALUES[0]]["results_subj"][0]
print("sXC dtype:", to_float_array(example_sub["sXC"]).dtype, "shape:", to_float_array(example_sub["sXC"]).shape)
print("S dtype:", to_float_array(example_sub["S"]).dtype, "shape:", to_float_array(example_sub["S"]).shape)
```

```

rankings = None
per_arch_df = None

rankings_path = os.path.join(DECODE_LOAD_DIR, "rankings.npy")
per_arch_path = os.path.join(DECODE_LOAD_DIR, "per_archetype_mean_accuracy.csv")

if USE_RANKINGS_FILE and os.path.exists(rankings_path):
    rankings = np.load(rankings_path, allow_pickle=True).item()
    print("Loaded rankings:", rankings_path)
else:
    print("Rankings file not used or not found.")

if os.path.exists(per_arch_path):
    per_arch_df = pd.read_csv(per_arch_path)
    print("Loaded per-archetype summary:", per_arch_path)
else:
    print("No per_archetype_mean_accuracy.csv found.")

def get_top_archetypes(cond_name, K, top_n=5):
    if (cond_name, K) in MANUAL_TOP_ARCHETYPES:
        return MANUAL_TOP_ARCHETYPES[(cond_name, K)]

    if rankings is not None and (cond_name, K) in rankings:
        return list(rankings[(cond_name, K)][:top_n])

    if per_arch_df is not None:
        sub = per_arch_df[(per_arch_df["condition"] == cond_name) & (per_arch_df["K"] == K)]
        if len(sub):
            return sub.sort_values("mean_accuracy", ascending=False)["archetype"].astype(int).tolist()[:top_n]

    raise ValueError(f"Could not find top archetypes for condition={cond_name}, K={K}")
```

```

posterior = loadmat(POSTERIOR_MAT)
centers = to_float_array(posterior['posterior']['centers'][0][0][0][0][0])
widths = to_float_array(list(posterior['posterior']['widths'][0][0][0][0][0][:, 0].T)).ravel()

lookup_table = {
    'Vis':'Visual',
    'SomMot':'Somatomotor',
    'DorsAttn':'Dorsal attention',
    'SalVentAttn':'Ventral attention',
    'Limbic':'Limbic',
    'Cont':'Frontoparietal',
    'Default':'Default mode'
}
network_colors = {
    'Visual':'#D7DF23',
    'Somatomotor':'#39B54A',
    'Dorsal attention':'#00A79D',
    'Ventral attention':'#27AAE1',
    'Limbic':'#1C75BC',
    'Frontoparietal':'#92278F',
    'Default mode':'#EE2A7B'
}
colors = ['#888888'] + [v for _, v in network_colors.items()]
network_codes = {k: i + 1 for i, k in enumerate(lookup_table.values())}

def nii2cmu(nifti_file, mask_file=None):
    def fullfact(dims):
        vals = np.asmatrix(range(1, dims[0] + 1)).T
        if len(dims) == 1:
            return vals
        aftervals = np.asmatrix(fullfact(dims[1:]))
        inds = np.asmatrix(np.zeros((np.prod(dims), len(dims))))
        row = 0
        for i in range(aftervals.shape[0]):
            inds[row:(row + len(vals)), 0] = vals
            inds[row:(row + len(vals)), 1:] = np.tile(aftervals[i, :], (len(vals), 1))
            row += len(vals)
        return inds

    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        img = nib.load(nifti_file) if type(nifti_file) == str else nifti_file
        mask = NiftiMasker(mask_strategy='background')
        mask.fit(nifti_file if mask_file is None else mask_file)

    Saff = img.get_sform()
    Y = np.float32(mask.transform(nifti_file)).copy()
    vmask = np.nonzero(np.array(np.reshape(mask.mask_img_.dataobj, (1, np.prod(mask.mask_img_.shape)), order='C')))[1]
    vox_coords = fullfact(img.shape[0:3])[vmask, ::-1] - 1
    R = np.array(np.dot(vox_coords, Saff[0:3, 0:3])) + Saff[:3, 3]
    return {'Y': Y, 'R': R}

def rbf(R, center, width):
    return np.exp(-np.sum((R - center) ** 2, axis=1) / width)

def node_labels(centers, widths, networks_cmu):
    labels = []
    for c, w in zip(centers, widths):
        r = rbf(networks_cmu['R'], c, w)
        label_weights = [sum(r[networks_cmu['Y'].ravel() == i]) for i in range(1, len(network_codes)+1)]
        labels.append(np.argmax(label_weights)+1)
    return pd.DataFrame({
        'code': labels,
        'Network': [list(lookup_table.values())[i-1] for i in labels]
    })

if NILEARN_AVAILABLE:
    key = pd.read_csv(SCHAEFER_TXT, sep='\t', header=None, names=['id','name','x','y','z','t']).drop('t', axis=1)
    key['network'] = key['name'].apply(lambda x: lookup_table[x.split('_')[2]])
    key['code'] = key['network'].apply(lambda x: network_codes[x])
    key.set_index('id', inplace=True)
    key.loc[0, 'code'] = 0

    networks_cmu = nii2cmu(SCHAEFER_NII)
    networks_cmu['Y'] = np.atleast_2d(np.array([key.loc[i, 'code'] for i in networks_cmu['Y']]).astype(float))
    node_code_df = node_labels(centers, widths, networks_cmu)
    display(node_code_df.head())
else:
    networks_cmu = None
    node_code_df = None
```

```

def coefficient_to_alpha(weights, alpha_min=0.15, alpha_max=1.0):
    w = np.asarray(weights, dtype=float).ravel()
    w = np.nan_to_num(w, nan=0.0, posinf=0.0, neginf=0.0)
    wmin, wmax = w.min(), w.max()
    if np.isclose(wmax, wmin):
        return np.full(len(w), alpha_max)
    scaled = (w - wmin) / (wmax - wmin)
    return alpha_min + (alpha_max - alpha_min) * scaled

def get_mean_timecourse(results_subj, k):
    # Spatial AA: temporal archetype / motif is column k of sXC (T x K)
    Xk = np.stack([to_float_array(sub["sXC"])[:, k] for sub in results_subj], axis=0)
    return Xk.mean(axis=0), Xk.std(axis=0) / np.sqrt(max(Xk.shape[0], 1))

def get_mean_coefficients(results_subj, k):
    # Spatial AA: coefficients for archetype k are row k of S (K x V)
    Ck = np.stack([to_float_array(sub["S"])[k, :] for sub in results_subj], axis=0)
    return Ck.mean(axis=0)

def get_decoding_value(cond_name, K, k):
    if per_arch_df is not None:
        sub = per_arch_df[
            (per_arch_df["condition"] == cond_name) &
            (per_arch_df["K"] == K) &
            (per_arch_df["archetype"] == k)
        ]
        if len(sub):
            row = sub.iloc[0]
            return float(row["mean_accuracy"])
    return None

def plot_decoding_bar(cond_name, K, k):
    val = get_decoding_value(cond_name, K, k)
    if val is None:
        print(f"No decoding value found for {cond_name}, K={K}, archetype={k}")
        return
    plt.figure(figsize=(4, 4))
    plt.bar([f"{cond_name}\nK={K}\na={k}"], [val], color=[COND_COLORS[cond_name]])
    plt.ylabel("Mean decoding accuracy")
    plt.title("Per-archetype decoding")
    plt.tight_layout()
    plt.show()

def plot_archetype_timecourse(results_subj, cond_name, K, k):
    mean_tc, sem_tc = get_mean_timecourse(results_subj, k)
    plt.figure(figsize=(9, 4))
    x = np.arange(len(mean_tc))
    plt.plot(x, mean_tc, color=COND_COLORS[cond_name], label=cond_name)
    plt.fill_between(x, mean_tc - sem_tc, mean_tc + sem_tc, color=COND_COLORS[cond_name], alpha=0.2)
    plt.xlabel("Time")
    plt.ylabel("Archetype value")
    plt.title(f"{cond_name} | K={K} | archetype {k} | temporal motif")
    plt.legend()
    plt.tight_layout()
    plt.show()

def plot_coefficient_brainplot(results_subj, cond_name, K, k):
    if not NILEARN_AVAILABLE:
        print("nilearn not available; skipping.")
        return None

    coeffs = get_mean_coefficients(results_subj, k)
    if node_code_df is None:
        print("Node code df not available; skipping.")
        return None

    codes = node_code_df["code"].to_numpy()
    alphas = coefficient_to_alpha(np.abs(coeffs))
    node_colors = []
    for code, alpha in zip(codes, alphas):
        base = to_rgba(colors[code])
        node_colors.append((base[0], base[1], base[2], float(alpha)))

    display = niplot.plot_connectome(
        np.eye(centers.shape[0]),
        centers,
        node_size=10,
        node_color=node_colors,
        display_mode="lyrz",
        title=f"{cond_name} | K={K} | archetype {k} | coefficient brain plot"
    )
    plt.show()
    plt.clf()
    return display

def plot_network_piechart(results_subj, cond_name, K, k):
    coeffs = np.abs(get_mean_coefficients(results_subj, k))
    network_df = node_code_df.copy()
    network_df["weight"] = coeffs

    counts = network_df.groupby("Network")["weight"].sum().reindex(list(network_colors.keys()), fill_value=0.0)
    pie_colors = [network_colors[name] for name in counts.index]

    fig, ax = plt.subplots(figsize=(4.8, 4.8))
    ax.pie(
        counts.values,
        labels=counts.index,
        colors=pie_colors,
        autopct=lambda p: f"{p:.1f}%" if p > 0 else "",
        startangle=90,
        counterclock=False
    )
    ax.set_title(f"{cond_name} | K={K} | archetype {k} | network pie chart")
    plt.tight_layout()
    plt.show()
```

```

top_ids = {}

for cond_name in CONDITIONS:
    top_ids[cond_name] = {}
    for K in K_VALUES:
        print("\n" + "=" * 90)
        print(f"Condition={cond_name} | K={K}")
        print("=" * 90)

        cur = fits[cond_name][K]
        results_subj = cur["results_subj"]
        top_arches = get_top_archetypes(cond_name, K, top_n=TOP_N_REPORT)
        top_ids[cond_name][K] = top_arches

        print("Top archetypes:", top_arches)

        for k in top_arches:
            plot_decoding_bar(cond_name, K, k)
            plot_archetype_timecourse(results_subj, cond_name, K, k)
            disp = plot_coefficient_brainplot(results_subj, cond_name, K, k)
            if disp is not None:
                try:
                    disp.close()
                except Exception:
                    pass
            if node_code_df is not None:
                plot_network_piechart(results_subj, cond_name, K, k)

top_ids
```

```

summary_rows = []
for cond_name in CONDITIONS:
    for K in K_VALUES:
        top_arches = top_ids[cond_name][K]
        for rank, k in enumerate(top_arches, start=1):
            summary_rows.append({
                "condition": cond_name,
                "K": K,
                "rank": rank,
                "archetype": int(k),
                "mean_accuracy": get_decoding_value(cond_name, K, int(k)),
            })

summary_df = pd.DataFrame(summary_rows)
summary_df.to_csv(os.path.join(OUTPUT_DIR, "top_spatial_within_archetypes_summary.csv"), index=False)
np.save(os.path.join(OUTPUT_DIR, "top_spatial_within_archetypes.npy"), top_ids, allow_pickle=True)

display(summary_df.head(20))
print("Saved summary outputs to:", OUTPUT_DIR)
```
