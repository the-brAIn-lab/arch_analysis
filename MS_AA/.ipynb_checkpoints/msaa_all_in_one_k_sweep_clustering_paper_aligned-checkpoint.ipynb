
# All-in-one MSAA sweep notebook aligned to paper definitions

This notebook runs the **full across-condition MSAA workflow** in one place, while matching the definitions used in the paper.

## Paper-aligned definitions

### **Spatial AA**
- **archetypes** = **timecourse motifs** (`K × T`) → unconstrained
- **coefficients** = **node weights** (`K × V`) → convex / nonnegative, expected to lie in `[0, 1]` and sum to 1 across archetypes for each node

### **Temporal AA**
- **archetypes** = **spatial maps** (`K × V`) → unconstrained
- **coefficients** = **time-varying weights** (`T × K`) → convex / nonnegative, expected to lie in `[0, 1]` and sum to 1 across archetypes for each timepoint

## What this notebook does

1. loads and normalizes the Pieman data
2. lets you choose **`AA_MODE = "spatial"`** or **`"temporal"`**
3. sweeps over a list of **K values**
4. fits **multi-subject archetypal analysis (MSAA)** for each K
5. standardizes the raw outputs into **paper-aligned arrays**
6. builds **subject-level feature tables**
7. runs **condition clustering** with **purity / ARI / NMI**, plus a permutation null
8. optionally performs a **flexible archetype search** over a range of cluster numbers
9. plots:
   - clustered subject correlation matrices
   - **node coefficient maps** for **Spatial AA**
   - spatial archetype maps for **Temporal AA** (optional)

## Important implementation note

The underlying `multi_subject_aa` code learns convex weights over the **rows** of the matrix it is given. Therefore:

- To get **Temporal AA** as defined in the paper, fit on **`T × V`**
- To get **Spatial AA** as defined in the paper, fit on **`V × T`**

This notebook explicitly enforces that mapping.


```

# =========================
# User configuration
# =========================

AA_MODE = "spatial"          # "spatial" or "temporal" (paper terminology)
K_VALUES = [5, 10, 20, 30]   # sweep over these K values

# Conditions to include in the across-condition analysis
CONDITIONS_TO_USE = ["intact", "word", "rest"]

# Trim rest to match the 300-TR conditions if needed
TRIM_REST_BY = 100

# Feature families for clustering / subject-level analysis
FEATURE_MODE = "all"         # "coeff_only" | "archetype_only" | "energy_only" | "all"

# Clustering settings
N_CLUSTERS = 3
N_PERM = 1000
RNG_SEED = 12345

# Paths
PIEMAN_MAT_PATH = "data/pieman/raw/pieman_data.mat"
POSTERIOR_MAT_PATH = "data/pieman/raw/pieman_posterior_K700.mat"
HELPERS_DIR = "helpers"
HYPERTOOLS_NORMALIZE_PATH = "hypertools/normalize.py"

# MSAA optimizer options
MSAA_OPTS = dict(
    maxiter=200,
    conv_crit=1e-6,
    fix_var_iter=5,
    heteroscedastic=False,
    use_gpu=False,
    rngSEED=RNG_SEED,
)

# Feature extraction settings
FS = 0.5
BANDS = ((0.01, 0.08), (0.08, 0.2))
LOG1P_ENERGY = True
ZSCORE_FEATURE_COLUMNS = False
```

## Data loading and normalization

```

def array_cutter(data_list, cut_length):
    out = [np.asarray(x).copy() for x in data_list]
    if cut_length <= 0:
        return out
    return [x[:-cut_length] for x in out]

def coerce_TxV(x):
    x = np.asarray(x)
    if x.ndim != 2:
        raise ValueError(f"Expected 2D array, got {x.shape}")
    T, V = x.shape
    # If input looks like V x T, transpose it
    if V < T:
        x = x.T
        T, V = x.shape
    x = x - x.mean(axis=0, keepdims=True)
    return x.astype(np.float32, copy=False)

def normalize_across_fallback(subject_list, eps=1e-8):
    subj = [coerce_TxV(x) for x in subject_list]
    stacked = np.vstack(subj)
    mu = stacked.mean(axis=0, keepdims=True)
    sd = stacked.std(axis=0, keepdims=True) + eps
    return [((x - mu) / sd).astype(np.float32, copy=False) for x in subj]

def normalize_subject_list(subject_list):
    subj = [coerce_TxV(x) for x in subject_list]
    if HT_NORMALIZE_AVAILABLE:
        try:
            return ht.normalize(subj, normalize="across")
        except Exception:
            return normalize_across_fallback(subj)
    return normalize_across_fallback(subj)

def load_pieman_data(mat_path=PIEMAN_MAT_PATH, conditions_to_use=CONDITIONS_TO_USE, trim_rest_by=TRIM_REST_BY):
    pieman_data = loadmat(mat_path)

    all_possible = ["intact", "paragraph", "word", "rest"]
    missing = [c for c in conditions_to_use if c not in all_possible]
    if missing:
        raise ValueError(f"Unknown condition(s): {missing}")

    data = []
    conds = []
    for c in conditions_to_use:
        if c == "paragraph":
            next_data = list(map(lambda i: pieman_data[c][:, i][0],
                                 np.where(np.arange(pieman_data[c].shape[1]) != 3)[0]))
        else:
            next_data = list(map(lambda i: pieman_data[c][:, i][0],
                                 np.arange(pieman_data[c].shape[1])))
        data.extend(next_data)
        conds.extend([c] * len(next_data))

    conds_array = np.array(conds)
    condition_data = {}
    for c in conditions_to_use:
        cur = [data[i] for i in np.where(conds_array == c)[0]]
        if c == "rest" and trim_rest_by > 0:
            cur = array_cutter(cur, trim_rest_by)
        condition_data[c] = normalize_subject_list(cur)

    all_subjects = []
    condition_labels = []
    for c in conditions_to_use:
        all_subjects.extend(condition_data[c])
        condition_labels.extend([c] * len(condition_data[c]))

    return condition_data, np.array(condition_labels), all_subjects

condition_data, condition_labels_str, all_subjects = load_pieman_data()

print("Conditions loaded:")
for c, arrs in condition_data.items():
    print(f"  {c}: n={len(arrs)}, example shape={coerce_TxV(arrs[0]).shape}")
print(f"Total subjects: {len(all_subjects)}")
```

## Orientation helpers aligned to paper definitions

```

def prepare_subject_matrix_for_mode(X, aa_mode):
    """
    Paper-aligned orientation:
      temporal AA -> fit on T x V
      spatial AA  -> fit on V x T
    """
    Xtv = coerce_TxV(X)
    if aa_mode == "temporal":
        return Xtv       # T x V
    elif aa_mode == "spatial":
        return Xtv.T     # V x T
    else:
        raise ValueError("aa_mode must be 'spatial' or 'temporal'")

def build_subject_objects(subject_list, aa_mode):
    subj_objs = []
    for X in subject_list:
        Xprep = prepare_subject_matrix_for_mode(X, aa_mode)
        subj_objs.append(Subject(X=Xprep, sX=Xprep))
    return subj_objs
```

## Run MSAA

```

def run_msaa_once(subject_list, aa_mode, K, opts=None, verbose=True):
    opts = dict(MSAA_OPTS if opts is None else opts)
    subj_objs = build_subject_objects(subject_list, aa_mode=aa_mode)

    t0 = time.time()
    results_subj, C, cost_fun, varexpl, secs = multi_subject_aa(
        subj_objs,
        noc=K,
        opts=opts,
    )
    elapsed = time.time() - t0

    if verbose:
        print(f"[{aa_mode}] K={K} | runtime={secs:.1f}s (wall={elapsed:.1f}s) | VarExpl={varexpl*100:.2f}%")
        print("  C shape:", np.asarray(C).shape)
        if len(results_subj) > 0:
            print("  subj0 keys:", list(results_subj[0].keys()))
            if "sXC" in results_subj[0]:
                print("  subj0 sXC shape:", np.asarray(results_subj[0]["sXC"]).shape)
            if "S" in results_subj[0]:
                print("  subj0 S shape:", np.asarray(results_subj[0]["S"]).shape)

    return {
        "results_subj": results_subj,
        "C": C,
        "cost_fun": cost_fun,
        "varexpl": varexpl,
        "secs": secs,
        "K": K,
        "aa_mode": aa_mode,
        "opts": opts,
    }

def run_msaa_sweep(subject_list, aa_mode, k_values, opts=None, verbose=True):
    out = {}
    for K in k_values:
        out[K] = run_msaa_once(subject_list, aa_mode, K, opts=opts, verbose=verbose)
    return out
```

## Standardize raw MSAA outputs into paper-aligned arrays

```

def standardize_results_arrays(results_subj, aa_mode, original_subject_list):
    """
    Convert raw per-subject MSAA outputs into paper-aligned arrays.

    Returns
    -------
    rep : dict

    For spatial AA:
      rep['temporal_archetypes'] : N x K x T   (unconstrained)
      rep['spatial_coeffs']      : N x K x V   (convex; expected in [0,1])
      rep['coeff_array']         : alias of spatial_coeffs
      rep['archetype_array']     : alias of temporal_archetypes

    For temporal AA:
      rep['spatial_archetypes']  : N x K x V   (unconstrained)
      rep['temporal_coeffs']     : N x T x K   (convex; expected in [0,1])
      rep['coeff_array']         : alias of temporal_coeffs
      rep['archetype_array']     : alias of spatial_archetypes
    """
    if aa_mode not in ("spatial", "temporal"):
        raise ValueError("aa_mode must be 'spatial' or 'temporal'")

    coeff_list = []
    archetype_list = []

    for i, sub in enumerate(results_subj):
        Xtv = coerce_TxV(original_subject_list[i])
        T, V = Xtv.shape

        sXC = np.asarray(sub["sXC"], dtype=float)
        S = np.asarray(sub["S"], dtype=float)

        if aa_mode == "temporal":
            # fit on T x V
            # expected raw: sXC = T x K  (convex time coefficients)
            #               S   = K x V  (spatial archetypes)
            if sXC.shape[0] != T and sXC.shape[1] == T:
                sXC = sXC.T
            if S.shape[1] != V and S.shape[0] == V:
                S = S.T
            if sXC.shape[0] != T:
                raise ValueError(f"Temporal AA: expected sXC first dim {T}, got {sXC.shape}")
            if S.shape[1] != V:
                raise ValueError(f"Temporal AA: expected S second dim {V}, got {S.shape}")

            coeff_list.append(sXC)    # T x K
            archetype_list.append(S)  # K x V

        elif aa_mode == "spatial":
            # fit on V x T
            # expected raw: sXC = V x K  (convex node coefficients)
            #               S   = K x T  (temporal archetypes)
            if sXC.shape[0] != V and sXC.shape[1] == V:
                sXC = sXC.T
            if S.shape[1] != T and S.shape[0] == T:
                S = S.T
            if sXC.shape[0] != V:
                raise ValueError(f"Spatial AA: expected sXC first dim {V}, got {sXC.shape}")
            if S.shape[1] != T:
                raise ValueError(f"Spatial AA: expected S second dim {T}, got {S.shape}")

            coeff_list.append(sXC.T)  # K x V
            archetype_list.append(S)  # K x T

    if aa_mode == "temporal":
        temporal_coeffs = np.stack(coeff_list, axis=0)       # N x T x K
        spatial_archetypes = np.stack(archetype_list, axis=0) # N x K x V
        return {
            "temporal_coeffs": temporal_coeffs,
            "spatial_archetypes": spatial_archetypes,
            "coeff_array": temporal_coeffs,
            "archetype_array": spatial_archetypes,
        }

    else:
        spatial_coeffs = np.stack(coeff_list, axis=0)          # N x K x V
        temporal_archetypes = np.stack(archetype_list, axis=0) # N x K x T
        return {
            "spatial_coeffs": spatial_coeffs,
            "temporal_archetypes": temporal_archetypes,
            "coeff_array": spatial_coeffs,
            "archetype_array": temporal_archetypes,
        }

def print_representation_diagnostics(rep, aa_mode, n_show=3):
    print(f"=== Diagnostics for {aa_mode} AA ===")
    if aa_mode == "spatial":
        C = rep["spatial_coeffs"]     # N x K x V
        A = rep["temporal_archetypes"] # N x K x T
        print("spatial_coeffs shape:", C.shape)
        print("temporal_archetypes shape:", A.shape)
        s = C[0]  # K x V
        print("Coeff min/max (subject 0):", float(np.min(s)), float(np.max(s)))
        print("Node-wise sums across archetypes (first 5 nodes):", np.sum(s, axis=0)[:n_show])
    else:
        C = rep["temporal_coeffs"]    # N x T x K
        A = rep["spatial_archetypes"] # N x K x V
        print("temporal_coeffs shape:", C.shape)
        print("spatial_archetypes shape:", A.shape)
        s = C[0]  # T x K
        print("Coeff min/max (subject 0):", float(np.min(s)), float(np.max(s)))
        print("Timepoint-wise sums across archetypes (first 5 timepoints):", np.sum(s, axis=1)[:n_show])
```

## Feature construction aligned to paper definitions

```

def _lag1_autocorr(x):
    x = np.asarray(x, dtype=float)
    x = x - np.mean(x)
    denom = np.dot(x, x)
    if denom == 0 or np.isnan(denom):
        return 0.0
    num = np.dot(x[:-1], x[1:])
    if np.isnan(num):
        return 0.0
    return float(num / denom)

def _spectral_entropy(x, fs, nperseg=None):
    from scipy.signal import welch
    from scipy.stats import entropy
    x = np.asarray(x, dtype=float)
    nperseg = nperseg or min(256, len(x))
    f, Pxx = welch(x, fs=fs, nperseg=nperseg)
    if np.sum(Pxx) == 0 or np.isnan(np.sum(Pxx)):
        return 0.0
    P = Pxx / (np.sum(Pxx) + 1e-12)
    return float(entropy(P))

def _band_power(x, fs, band, nperseg=None):
    from scipy.signal import welch
    x = np.asarray(x, dtype=float)
    nperseg = nperseg or min(256, len(x))
    f, Pxx = welch(x, fs=fs, nperseg=nperseg)
    if np.all(Pxx == 0):
        return 0.0
    idx = (f >= band[0]) & (f < band[1])
    return float(np.trapz(Pxx[idx], f[idx])) if np.any(idx) else 0.0

def _safe_gini(x):
    x = np.abs(np.asarray(x, dtype=float)).ravel()
    if np.allclose(x, 0):
        return 0.0
    x = np.sort(x)
    n = len(x)
    index = np.arange(1, n + 1)
    return float((2 * np.sum(index * x) / (n * np.sum(x))) - (n + 1) / n)

def make_df_features_from_rep(rep, aa_mode, condition_codes, condition_map=None,
                              include_coeff=True, include_archetype=True, include_energy=True,
                              fs=FS, bands=BANDS, log1p_energy=LOG1P_ENERGY,
                              zscore_across_subjects=ZSCORE_FEATURE_COLUMNS):
    """
    Build a subject-level feature dataframe aligned to the paper definitions.

    Spatial AA:
      coeff features     <- spatial_coeffs (K x V), i.e. node coefficients
      archetype features <- temporal_archetypes (K x T), i.e. timecourse motifs

    Temporal AA:
      coeff features     <- temporal_coeffs (T x K), i.e. time-varying convex coefficients
      archetype features <- spatial_archetypes (K x V), i.e. spatial maps
    """
    condition_codes = np.asarray(condition_codes)
    N = len(condition_codes)
    rows = []

    if aa_mode == "spatial":
        coeffs = rep["spatial_coeffs"]         # N x K x V
        archetypes = rep["temporal_archetypes"] # N x K x T
        N1, K, V = coeffs.shape
        N2, K2, T = archetypes.shape
        if N1 != N2 or K != K2:
            raise ValueError("Spatial AA arrays are inconsistent")

        for s in range(N):
            feats = {"Subject": s, "Condition": int(condition_codes[s]), "AnalysisType": aa_mode}
            for k in range(K):
                cvec = coeffs[s, k, :]     # node coefficients
                avec = archetypes[s, k, :] # temporal motif

                if include_energy:
                    energy = float(np.sum(cvec**2) * np.sum(avec**2))
                    if log1p_energy:
                        energy = float(np.log1p(energy))
                    feats[f"A{k+1}_energy"] = energy

                if include_coeff:
                    feats[f"A{k+1}_coeff_mean"] = float(np.mean(cvec))
                    feats[f"A{k+1}_coeff_std"] = float(np.std(cvec))
                    feats[f"A{k+1}_coeff_l1"] = float(np.sum(np.abs(cvec)))
                    feats[f"A{k+1}_coeff_l2"] = float(np.sqrt(np.sum(cvec**2)))
                    feats[f"A{k+1}_coeff_max"] = float(np.max(cvec))
                    feats[f"A{k+1}_coeff_gini"] = _safe_gini(cvec)

                if include_archetype:
                    feats[f"A{k+1}_arch_mean"] = float(np.mean(avec))
                    feats[f"A{k+1}_arch_std"] = float(np.std(avec))
                    feats[f"A{k+1}_arch_ac1"] = _lag1_autocorr(avec)
                    feats[f"A{k+1}_arch_specent"] = _spectral_entropy(avec, fs=fs)
                    for b_idx, band in enumerate(bands):
                        feats[f"A{k+1}_arch_bp{b_idx}"] = _band_power(avec, fs=fs, band=band)

                rows.append(feats)

    elif aa_mode == "temporal":
        coeffs = rep["temporal_coeffs"]        # N x T x K
        archetypes = rep["spatial_archetypes"] # N x K x V
        N1, T, K = coeffs.shape
        N2, K2, V = archetypes.shape
        if N1 != N2 or K != K2:
            raise ValueError("Temporal AA arrays are inconsistent")

        for s in range(N):
            feats = {"Subject": s, "Condition": int(condition_codes[s]), "AnalysisType": aa_mode}
            for k in range(K):
                cvec = coeffs[s, :, k]     # time coefficients
                avec = archetypes[s, k, :] # spatial map

                if include_energy:
                    energy = float(np.sum(cvec**2) * np.sum(avec**2))
                    if log1p_energy:
                        energy = float(np.log1p(energy))
                    feats[f"A{k+1}_energy"] = energy

                if include_coeff:
                    feats[f"A{k+1}_coeff_mean"] = float(np.mean(cvec))
                    feats[f"A{k+1}_coeff_std"] = float(np.std(cvec))
                    feats[f"A{k+1}_coeff_ac1"] = _lag1_autocorr(cvec)
                    feats[f"A{k+1}_coeff_specent"] = _spectral_entropy(cvec, fs=fs)
                    for b_idx, band in enumerate(bands):
                        feats[f"A{k+1}_coeff_bp{b_idx}"] = _band_power(cvec, fs=fs, band=band)

                if include_archetype:
                    feats[f"A{k+1}_arch_mean"] = float(np.mean(avec))
                    feats[f"A{k+1}_arch_std"] = float(np.std(avec))
                    feats[f"A{k+1}_arch_l1"] = float(np.sum(np.abs(avec)))
                    feats[f"A{k+1}_arch_l2"] = float(np.sqrt(np.sum(avec**2)))
                    feats[f"A{k+1}_arch_maxabs"] = float(np.max(np.abs(avec)))
                    feats[f"A{k+1}_arch_gini"] = _safe_gini(avec)

            rows.append(feats)
    else:
        raise ValueError("aa_mode must be 'spatial' or 'temporal'")

    df = pd.DataFrame(rows)

    if condition_map is not None:
        df["ConditionName"] = pd.Series(df["Condition"]).map(condition_map).astype("category")

    if zscore_across_subjects:
        non_feature_cols = {"Subject", "Condition", "ConditionName", "AnalysisType"}
        feature_cols = [c for c in df.columns if c not in non_feature_cols]
        scaler = StandardScaler()
        df[feature_cols] = scaler.fit_transform(df[feature_cols])

    return df

def build_subject_feature_matrix(df_features, mode=FEATURE_MODE):
    coeff_cols = [c for c in df_features.columns if "_coeff_" in c]
    arch_cols = [c for c in df_features.columns if "_arch_" in c]
    energy_cols = [c for c in df_features.columns if c.endswith("_energy")]

    if mode == "coeff_only":
        cols = coeff_cols
    elif mode == "archetype_only":
        cols = arch_cols
    elif mode == "energy_only":
        cols = energy_cols
    elif mode == "all":
        cols = energy_cols + coeff_cols + arch_cols
    else:
        raise ValueError("mode must be 'coeff_only', 'archetype_only', 'energy_only', or 'all'")

    X_raw = df_features[cols].to_numpy()
    cond = pd.Categorical(df_features["Condition"]).codes
    subj_ids = df_features["Subject"].to_numpy()

    imputer = SimpleImputer(strategy="constant", fill_value=0.0)
    X_imp = imputer.fit_transform(X_raw)
    X = StandardScaler().fit_transform(X_imp)

    return X, cond, subj_ids, cols
```

## Clustering metrics and plotting

```

def purity_score(y_cluster, y_class):
    y_cluster = np.asarray(y_cluster)
    y_class = np.asarray(y_class)
    total = 0
    for c in np.unique(y_cluster):
        mask = (y_cluster == c)
        _, counts = np.unique(y_class[mask], return_counts=True)
        total += counts.max()
    return total / len(y_cluster)

def clustering_metrics(y_cluster, y_class):
    return {
        "purity": purity_score(y_cluster, y_class),
        "ari": adjusted_rand_score(y_class, y_cluster),
        "nmi": normalized_mutual_info_score(y_class, y_cluster),
    }

def permutation_null_purity(y_cluster, y_class, n_perm=N_PERM, random_state=RNG_SEED):
    rng = np.random.default_rng(random_state)
    y_cluster = np.asarray(y_cluster)
    y_class = np.asarray(y_class)
    null_vals = np.zeros(n_perm, dtype=float)
    for i in range(n_perm):
        shuffled = rng.permutation(y_class)
        null_vals[i] = purity_score(y_cluster, shuffled)
    return null_vals

def hierarchical_cluster_subjects(X, cond, n_clusters=N_CLUSTERS, linkage_method="average", feature_metric="euclidean"):
    dvec = pdist(X, metric=feature_metric)
    Z = linkage(dvec, method=linkage_method)
    order = leaves_list(Z)
    cluster_labels = fcluster(Z, n_clusters, criterion="maxclust")

    metrics = clustering_metrics(cluster_labels, cond)

    sim = np.corrcoef(X)
    sim_sorted = sim[np.ix_(order, order)]
    cond_sorted = np.asarray(cond)[order]
    cluster_sorted = cluster_labels[order]
    boundaries = np.where(np.diff(cluster_sorted) != 0)[0] + 1

    return {
        "Z": Z,
        "order": order,
        "cluster_labels": cluster_labels,
        "cluster_sorted": cluster_sorted,
        "cond_sorted": cond_sorted,
        "sim": sim,
        "sim_sorted": sim_sorted,
        "boundaries": boundaries,
        **metrics,
    }

def evaluate_cluster_solution(X, cond, n_clusters=N_CLUSTERS, linkage_method="average", feature_metric="euclidean",
                              n_perm=N_PERM, random_state=RNG_SEED):
    result = hierarchical_cluster_subjects(
        X=X,
        cond=cond,
        n_clusters=n_clusters,
        linkage_method=linkage_method,
        feature_metric=feature_metric,
    )
    null_purity = permutation_null_purity(
        y_cluster=result["cluster_labels"],
        y_class=cond,
        n_perm=n_perm,
        random_state=random_state,
    )
    result["null_purity"] = null_purity
    result["null_mean"] = float(np.mean(null_purity))
    result["null_std"] = float(np.std(null_purity))
    result["null_p95"] = float(np.percentile(null_purity, 95))
    result["p_perm"] = float((np.sum(null_purity >= result["purity"]) + 1) / (len(null_purity) + 1))
    return result

def plot_clustered_similarity(result, cond_colors=None, title="Subject-by-subject similarity",
                              vmin=-0.5, vmax=1.0, figsize=(8, 7), show_subject_ids=True):
    if cond_colors is None:
        cond_colors = {0: "purple", 1: "green", 2: "black"}

    sim_sorted = result["sim_sorted"]
    order = result["order"]
    cond_sorted = result["cond_sorted"]
    boundaries = result["boundaries"]

    plt.figure(figsize=figsize)
    ax = sns.heatmap(
        sim_sorted,
        cmap="coolwarm",
        vmin=vmin,
        vmax=vmax,
        square=True,
        xticklabels=(order + 1) if show_subject_ids else False,
        yticklabels=(order + 1) if show_subject_ids else False,
        cbar_kws={"label": "Pearson correlation"},
    )

    if show_subject_ids:
        tick_colors = [cond_colors.get(int(c), "gray") for c in cond_sorted]
        for tick_label, color in zip(ax.get_xticklabels(), tick_colors):
            tick_label.set_color(color)
        for tick_label, color in zip(ax.get_yticklabels(), tick_colors):
            tick_label.set_color(color)

    for b in result["boundaries"]:
        ax.axhline(b, color="black", linewidth=2)
        ax.axvline(b, color="black", linewidth=2)

    plt.title(
        f"{title}\nPurity={result['purity']:.3f}, "
        f"ARI={result['ari']:.3f}, NMI={result['nmi']:.3f}"
    )
    plt.xlabel("Subjects (clustered)")
    plt.ylabel("Subjects (clustered)")
    plt.tight_layout()
    plt.show()

def plot_purity_null(result, bins=30, figsize=(6, 4)):
    plt.figure(figsize=figsize)
    plt.hist(result["null_purity"], bins=bins, alpha=0.8)
    plt.axvline(result["purity"], linestyle="--", linewidth=2)
    plt.title(
        "Permutation null for purity\n"
        f"Observed={result['purity']:.3f}, "
        f"Null mean={result['null_mean']:.3f}, "
        f"p={result['p_perm']:.4f}"
    )
    plt.xlabel("Purity")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()
```

## Flexible archetype search over a range of cluster numbers

```

def _impute_and_zscore_rows(X):
    X = np.asarray(X, dtype=float)
    imp = SimpleImputer(strategy="constant", fill_value=0.0)
    X = imp.fit_transform(X)
    X = zscore(X, axis=1, ddof=0, nan_policy="omit")
    X = np.nan_to_num(X)
    return X

def subject_vectors_for_archetype(rep, aa_mode, k, vector_kind=None):
    """
    Return a subject-by-feature matrix for a single archetype.

    Spatial AA:
      vector_kind='coeff'     -> N x V node coefficients (convex)
      vector_kind='archetype' -> N x T temporal motifs

    Temporal AA:
      vector_kind='coeff'     -> N x T time coefficients (convex)
      vector_kind='archetype' -> N x V spatial maps
    """
    if vector_kind is None:
        vector_kind = "coeff"

    if aa_mode == "spatial":
        if vector_kind == "coeff":
            return rep["spatial_coeffs"][:, k, :]       # N x V
        elif vector_kind == "archetype":
            return rep["temporal_archetypes"][:, k, :]  # N x T
    elif aa_mode == "temporal":
        if vector_kind == "coeff":
            return rep["temporal_coeffs"][:, :, k]      # N x T
        elif vector_kind == "archetype":
            return rep["spatial_archetypes"][:, k, :]   # N x V

    raise ValueError("Invalid aa_mode or vector_kind")

def evaluate_single_archetype_over_cluster_range(rep, aa_mode, condition_codes, k,
                                                 vector_kind="coeff",
                                                 cluster_range=range(2, 7),
                                                 linkage_method="average",
                                                 feature_metric="euclidean",
                                                 n_perm=500,
                                                 random_state=12345):
    Xk = subject_vectors_for_archetype(rep, aa_mode, k, vector_kind=vector_kind)
    Xk = _impute_and_zscore_rows(Xk)
    cond = np.asarray(condition_codes)

    rows = []
    all_results = {}

    for n_clusters in cluster_range:
        result = evaluate_cluster_solution(
            X=Xk,
            cond=cond,
            n_clusters=n_clusters,
            linkage_method=linkage_method,
            feature_metric=feature_metric,
            n_perm=n_perm,
            random_state=random_state,
        )

        try:
            sil = silhouette_score(Xk, result["cluster_labels"], metric="euclidean")
        except Exception:
            sil = np.nan

        row = {
            "k": k + 1,
            "vector_kind": vector_kind,
            "n_clusters": n_clusters,
            "purity": result["purity"],
            "ari": result["ari"],
            "nmi": result["nmi"],
            "silhouette": sil,
            "null_mean": result["null_mean"],
            "null_p95": result["null_p95"],
            "p_perm": result["p_perm"],
        }
        rows.append(row)
        all_results[n_clusters] = result

    all_results_df = pd.DataFrame(rows)

    best_idx = (
        all_results_df
        .sort_values(by=["ari", "nmi", "purity", "n_clusters"],
                     ascending=[False, False, False, True])
        .index[0]
    )

    best_row = all_results_df.loc[best_idx].to_dict()
    best_n_clusters = int(best_row["n_clusters"])
    best_result = all_results[best_n_clusters].copy()
    best_result.update(best_row)

    return best_result, all_results_df

def scan_all_archetypes_flexible_clusters(rep, aa_mode, condition_codes,
                                          vector_kind="coeff",
                                          cluster_range=range(2, 7),
                                          linkage_method="average",
                                          feature_metric="euclidean",
                                          n_perm=500,
                                          random_state=12345,
                                          sort_by="ari"):
    # determine K
    if aa_mode == "spatial":
        K = rep["spatial_coeffs"].shape[1]
    else:
        K = rep["spatial_archetypes"].shape[1]

    best_rows = []
    full_rows = []

    for k in range(K):
        best_result, all_results_df = evaluate_single_archetype_over_cluster_range(
            rep=rep,
            aa_mode=aa_mode,
            condition_codes=condition_codes,
            k=k,
            vector_kind=vector_kind,
            cluster_range=cluster_range,
            linkage_method=linkage_method,
            feature_metric=feature_metric,
            n_perm=n_perm,
            random_state=random_state,
        )

        best_rows.append({
            "k": best_result["k"],
            "vector_kind": best_result["vector_kind"],
            "best_n_clusters": best_result["n_clusters"],
            "purity": best_result["purity"],
            "ari": best_result["ari"],
            "nmi": best_result["nmi"],
            "silhouette": best_result["silhouette"],
            "null_mean": best_result["null_mean"],
            "null_p95": best_result["null_p95"],
            "p_perm": best_result["p_perm"],
        })

        full_rows.append(all_results_df)

    best_df = pd.DataFrame(best_rows).sort_values(sort_by, ascending=False).reset_index(drop=True)
    full_df = pd.concat(full_rows, ignore_index=True)
    return best_df, full_df

def plot_archetype_cluster_search(full_df, top_k=10, metric="ari"):
    top_arches = (
        full_df.groupby("k")[metric]
        .max()
        .sort_values(ascending=False)
        .head(top_k)
        .index.tolist()
    )

    plot_df = full_df[full_df["k"].isin(top_arches)].copy()
    plt.figure(figsize=(10, 5))
    sns.lineplot(data=plot_df, x="n_clusters", y=metric, hue="k", marker="o")
    plt.title(f"{metric.upper()} across cluster numbers for top archetypes")
    plt.xlabel("Number of clusters")
    plt.ylabel(metric.upper())
    plt.tight_layout()
    plt.show()
```

```

K_FOR_FLEX_SCAN = K_VALUES[0]
VECTOR_KIND = "coeff"  # for spatial AA this means node coefficients; for temporal AA time coefficients
CLUSTER_RANGE = range(2, 7)

cur = analysis_by_k[K_FOR_FLEX_SCAN]

best_df, full_df = scan_all_archetypes_flexible_clusters(
    rep=cur["rep"],
    aa_mode=AA_MODE,
    condition_codes=condition_codes,
    vector_kind=VECTOR_KIND,
    cluster_range=CLUSTER_RANGE,
    linkage_method="average",
    feature_metric="euclidean",
    n_perm=500,
    random_state=RNG_SEED,
    sort_by="ari",
)

display(best_df.head(10))
plot_archetype_cluster_search(full_df, top_k=8, metric="ari")
```

```

def plot_top_n_cluster_matrices(rep, aa_mode, condition_codes, best_df, top_n=6,
                                linkage_method="average", feature_metric="euclidean",
                                n_perm=500, random_state=12345,
                                figsize_per_panel=(4.2, 4.0)):
    n_show = min(top_n, len(best_df))
    ncols = min(3, n_show)
    nrows = int(np.ceil(n_show / ncols))

    fig, axes = plt.subplots(
        nrows, ncols,
        figsize=(figsize_per_panel[0] * ncols, figsize_per_panel[1] * nrows),
        squeeze=False
    )

    for idx in range(nrows * ncols):
        ax = axes.flat[idx]
        if idx >= n_show:
            ax.axis("off")
            continue

        row = best_df.iloc[idx]
        k = int(row["k"]) - 1
        vector_kind = row["vector_kind"]
        n_clusters = int(row["best_n_clusters"])

        Xk = subject_vectors_for_archetype(rep, aa_mode, k, vector_kind=vector_kind)
        Xk = _impute_and_zscore_rows(Xk)

        result = evaluate_cluster_solution(
            X=Xk,
            cond=np.asarray(condition_codes),
            n_clusters=n_clusters,
            linkage_method=linkage_method,
            feature_metric=feature_metric,
            n_perm=n_perm,
            random_state=random_state,
        )

        sns.heatmap(
            result["sim_sorted"],
            cmap="coolwarm",
            vmin=-0.5, vmax=1.0,
            square=True,
            xticklabels=False,
            yticklabels=False,
            cbar=(idx == 0),
            ax=ax
        )

        for b in result["boundaries"]:
            ax.axhline(b, color="black", linewidth=2)
            ax.axvline(b, color="black", linewidth=2)

        ax.set_title(
            f"A{int(row['k'])} | {vector_kind} | c={n_clusters}\n"
            f"ARI={row['ari']:.2f}, Pur={row['purity']:.2f}, NMI={row['nmi']:.2f}"
        )

    plt.tight_layout()
    plt.show()

plot_top_n_cluster_matrices(
    rep=cur["rep"],
    aa_mode=AA_MODE,
    condition_codes=condition_codes,
    best_df=best_df,
    top_n=6,
    linkage_method="average",
    feature_metric="euclidean",
    n_perm=500,
    random_state=RNG_SEED,
)
```

## Brain plotting helpers

```

def load_node_centers(posterior_mat_path=POSTERIOR_MAT_PATH):
    posterior = loadmat(posterior_mat_path)
    centers = posterior["posterior"]["centers"][0][0][0][0][0]
    centers = np.asarray(centers, dtype=float)
    if centers.shape[1] != 3 and centers.shape[0] == 3:
        centers = centers.T
    if centers.shape[1] != 3:
        raise ValueError(f"Expected centers to be V x 3, got {centers.shape}")
    return centers

def plot_node_map(weights, centers, title="", node_size=35, cmap="coolwarm", global_vmax=None):
    weights = np.asarray(weights, dtype=float).ravel()
    centers = np.asarray(centers, dtype=float)

    if global_vmax is None:
        vmax = np.max(np.abs(weights)) if np.max(np.abs(weights)) > 0 else 1.0
    else:
        vmax = float(global_vmax)
    vmin = -vmax

    if NILEARN_AVAILABLE:
        try:
            display = niplot.plot_markers(
                node_values=weights,
                node_coords=centers,
                node_size=node_size,
                node_cmap=cmap,
                colorbar=True,
                annotate=False,
                title=title,
                black_bg=False,
                vmin=vmin,
                vmax=vmax,
            )
        except TypeError:
            display = niplot.plot_markers(
                node_values=weights,
                node_coords=centers,
                node_size=node_size,
                node_cmap=cmap,
                colorbar=True,
                annotate=False,
                title=title,
                black_bg=False,
            )
        return display
    else:
        fig, axes = plt.subplots(1, 3, figsize=(12, 4))
        views = [(0, 1, "x-y"), (0, 2, "x-z"), (1, 2, "y-z")]
        sc = None
        for ax, (i, j, lbl) in zip(axes, views):
            sc = ax.scatter(centers[:, i], centers[:, j], c=weights, s=node_size, cmap=cmap, vmin=vmin, vmax=vmax)
            ax.set_title(lbl)
            ax.set_xlabel(["x", "y", "z"][i])
            ax.set_ylabel(["x", "y", "z"][j])
        fig.suptitle(title)
        fig.colorbar(sc, ax=axes.ravel().tolist(), shrink=0.8)
        plt.tight_layout()
        plt.show()
        return None
```


## Plot node coefficient maps (Spatial AA) or spatial archetype maps (Temporal AA)

These functions are explicit about what is being plotted:

- **Spatial AA**:
  - `plot_top_n_node_coefficient_maps(...)` plots **spatial coefficients** (`K × V`), which should be the node weights you expect
  - these are the quantities that should look like convex coefficients

- **Temporal AA**:
  - `plot_top_n_spatial_archetype_maps(...)` plots **spatial archetypes** (`K × V`)
  - these are unconstrained and may be negative


```

def compute_best_archetype_cluster_labels(rep, aa_mode, condition_codes, best_df, rank=0,
                                          linkage_method="average", feature_metric="euclidean",
                                          n_perm=500, random_state=12345):
    row = best_df.iloc[rank]
    k = int(row["k"]) - 1
    vector_kind = row["vector_kind"]
    n_clusters = int(row["best_n_clusters"])

    Xk = subject_vectors_for_archetype(rep, aa_mode, k, vector_kind=vector_kind)
    Xk = _impute_and_zscore_rows(Xk)

    result = evaluate_cluster_solution(
        X=Xk,
        cond=np.asarray(condition_codes),
        n_clusters=n_clusters,
        linkage_method=linkage_method,
        feature_metric=feature_metric,
        n_perm=n_perm,
        random_state=random_state,
    )
    return row, result

def plot_top_n_node_coefficient_maps(rep, condition_codes, condition_map, best_df, centers,
                                     top_n=3, aggregate="by_condition",
                                     linkage_method="average", feature_metric="euclidean",
                                     n_perm=500, random_state=12345,
                                     global_vmax=None):
    """
    Spatial AA only: plot spatial coefficients (node weights).
    """
    if "spatial_coeffs" not in rep:
        raise ValueError("Node coefficient maps are only defined for Spatial AA representations.")

    n_show = min(top_n, len(best_df))
    cond_codes = np.asarray(condition_codes)

    for rank in range(n_show):
        row, result = compute_best_archetype_cluster_labels(
            rep=rep,
            aa_mode="spatial",
            condition_codes=cond_codes,
            best_df=best_df,
            rank=rank,
            linkage_method=linkage_method,
            feature_metric=feature_metric,
            n_perm=n_perm,
            random_state=random_state,
        )

        k = int(row["k"]) - 1
        coeff_k = rep["spatial_coeffs"][:, k, :]  # N x V

        if aggregate == "overall":
            mean_w = coeff_k.mean(axis=0)
            plot_node_map(mean_w, centers, title=f"A{k+1} overall node coefficients | ARI={row['ari']:.2f}",
                          global_vmax=global_vmax)

        elif aggregate == "by_condition":
            for c in np.unique(cond_codes):
                mask = (cond_codes == c)
                mean_w = coeff_k[mask].mean(axis=0)
                cname = condition_map[int(c)] if isinstance(condition_map, dict) else str(c)
                plot_node_map(mean_w, centers, title=f"A{k+1} node coefficients | condition={cname} | ARI={row['ari']:.2f}",
                              global_vmax=global_vmax)

        elif aggregate == "by_cluster":
            cluster_labels = np.asarray(result["cluster_labels"])
            for cl in np.unique(cluster_labels):
                mask = (cluster_labels == cl)
                mean_w = coeff_k[mask].mean(axis=0)
                plot_node_map(mean_w, centers, title=f"A{k+1} node coefficients | cluster={cl} | ARI={row['ari']:.2f}",
                              global_vmax=global_vmax)

        else:
            raise ValueError("aggregate must be 'overall', 'by_condition', or 'by_cluster'")

def plot_top_n_spatial_archetype_maps(rep, condition_codes, condition_map, best_df, centers,
                                      top_n=3, aggregate="by_condition",
                                      linkage_method="average", feature_metric="euclidean",
                                      n_perm=500, random_state=12345,
                                      global_vmax=None):
    """
    Temporal AA only: plot spatial archetypes (unconstrained maps).
    """
    if "spatial_archetypes" not in rep:
        raise ValueError("Spatial archetype maps are only defined for Temporal AA representations.")

    n_show = min(top_n, len(best_df))
    cond_codes = np.asarray(condition_codes)

    for rank in range(n_show):
        row, result = compute_best_archetype_cluster_labels(
            rep=rep,
            aa_mode="temporal",
            condition_codes=cond_codes,
            best_df=best_df,
            rank=rank,
            linkage_method=linkage_method,
            feature_metric=feature_metric,
            n_perm=n_perm,
            random_state=random_state,
        )

        k = int(row["k"]) - 1
        arch_k = rep["spatial_archetypes"][:, k, :]  # N x V

        if aggregate == "overall":
            mean_w = arch_k.mean(axis=0)
            plot_node_map(mean_w, centers, title=f"A{k+1} spatial archetype | ARI={row['ari']:.2f}",
                          global_vmax=global_vmax)

        elif aggregate == "by_condition":
            for c in np.unique(cond_codes):
                mask = (cond_codes == c)
                mean_w = arch_k[mask].mean(axis=0)
                cname = condition_map[int(c)] if isinstance(condition_map, dict) else str(c)
                plot_node_map(mean_w, centers, title=f"A{k+1} spatial archetype | condition={cname} | ARI={row['ari']:.2f}",
                              global_vmax=global_vmax)

        elif aggregate == "by_cluster":
            cluster_labels = np.asarray(result["cluster_labels"])
            for cl in np.unique(cluster_labels):
                mask = (cluster_labels == cl)
                mean_w = arch_k[mask].mean(axis=0)
                plot_node_map(mean_w, centers, title=f"A{k+1} spatial archetype | cluster={cl} | ARI={row['ari']:.2f}",
                              global_vmax=global_vmax)

        else:
            raise ValueError("aggregate must be 'overall', 'by_condition', or 'by_cluster'")
```


## Notes

- If you run **Spatial AA**, the node maps plotted at the end come from **spatial coefficients** and should match the interpretation in the paper.
- If you run **Temporal AA**, the node maps are **spatial archetypes**, not coefficients, so they are **not** expected to lie in `[0, 1]`.
- The diagnostics printed after standardization are intended to catch exactly this issue:
  - Spatial AA: check sums across archetypes for each node
  - Temporal AA: check sums across archetypes for each timepoint
