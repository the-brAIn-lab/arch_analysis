
# All-in-one MSAA sweep + clustering notebook

This notebook is designed to run the **full across-condition MSAA workflow** in one place:

1. load and normalize the Pieman data
2. choose **AA orientation** (`"spatial"` or `"temporal"`, using the paper's terminology)
3. sweep over a list of **K values**
4. fit **multi-subject archetypal analysis (MSAA)** for each K
5. build **subject-level feature tables**
6. run **3-cluster condition clustering** with **Purity / ARI / NMI**
7. optionally scan individual archetypes to identify “winning” archetypes

## Important orientation note

The underlying `multi_subject_aa` code learns shared archetypes over the **feature dimension** of the input matrix `X`.

So in this notebook:

- **`AA_MODE = "temporal"`** means the paper's *Temporal AA*  
  - input each subject as **T × V**
  - shared archetypes are **spatial maps** (`V × K`)
  - subject coefficients are **time-varying** (`T × K`)

- **`AA_MODE = "spatial"`** means the paper's *Spatial AA*  
  - input each subject as **V × T** (transpose before fitting)
  - shared archetypes are **temporal motifs** (`T × K`)
  - subject coefficients index the transposed rows, but we convert outputs back into a common feature format for clustering

## Assumptions

This notebook assumes the following are available in your working directory / environment:

- `helpers/MultiSubject_AA.py`
- `data/pieman/raw/pieman_data.mat`
- optionally: `hypertools/normalize.py`

If `hypertools` is not available, the notebook uses a built-in across-subject z-scoring fallback.

## Expected MSAA output structure

Each subject result returned by `multi_subject_aa` is assumed to contain:

- `sXC`: subject expression matrix
- `S`: subject-specific archetype-by-feature matrix

The notebook standardizes these into:

- `time_array`: **N × T × K**
- `spatial_array`: **N × K × V**

so the downstream feature/clustering code works for both orientations.


```

# =========================
# User configuration
# =========================

AA_MODE = "spatial"          # "spatial" or "temporal" (paper terminology)
K_VALUES = [5, 10, 20, 30]   # sweep over these K values
CONDITIONS_TO_USE = ["intact", "word", "rest"]  # default across-condition analysis
TRIM_REST_BY = 100           # trim last 100 TRs from rest to match 300 timepoints
FEATURE_MODE = "all"         # "energy_only" | "temporal_only" | "spatial_only" | "all"
N_CLUSTERS = 3               # because there are 3 conditions
N_PERM = 1000                # permutation null iterations for purity
RNG_SEED = 12345

# Paths
PIEMAN_MAT_PATH = "data/pieman/raw/pieman_data.mat"
HELPERS_DIR = "helpers"
HYPERTOOLS_NORMALIZE_PATH = "hypertools/normalize.py"

# MSAA optimizer options
MSAA_OPTS = dict(
    maxiter=200,
    conv_crit=1e-6,
    fix_var_iter=5,
    heteroscedastic=False,
    use_gpu=False,
    rngSEED=RNG_SEED,
)

# Feature extraction settings
FS = 0.5
BANDS = ((0.01, 0.08), (0.08, 0.2))
LOG1P_ENERGY = True
ZSCORE_FEATURE_COLUMNS = False
```

```

# =========================
# Imports
# =========================
%matplotlib inline

import os
import sys
import time
import pathlib
import importlib.util
import warnings

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from scipy.io import loadmat
from scipy.cluster.hierarchy import linkage, leaves_list, fcluster
from scipy.spatial.distance import pdist
from scipy.stats import zscore

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score

# Try to import hypertools normalize if available
HT_NORMALIZE_AVAILABLE = False
try:
    spec = importlib.util.spec_from_file_location("normalize", HYPERTOOLS_NORMALIZE_PATH)
    ht = importlib.util.module_from_spec(spec)
    spec.loader.exec_module(ht)
    HT_NORMALIZE_AVAILABLE = True
    print("Loaded hypertools normalize.")
except Exception as e:
    print("hypertools normalize not available; will use built-in fallback normalization.")
    ht = None

# Import MSAA helpers
sys.path.append(HELPERS_DIR)
from MultiSubject_AA import Subject, multi_subject_aa

print("Ready.")
```

## Data loading and normalization

```

def array_cutter(data_list, cut_length):
    out = [np.asarray(x).copy() for x in data_list]
    if cut_length <= 0:
        return out
    return [x[:-cut_length] for x in out]

def coerce_TxV(x):
    x = np.asarray(x)
    if x.ndim != 2:
        raise ValueError(f"Expected 2D array, got {x.shape}")
    T, V = x.shape
    # If it looks like V x T, transpose
    if V < T:
        x = x.T
        T, V = x.shape
    # zero-mean each feature across time
    x = x - x.mean(axis=0, keepdims=True)
    return x.astype(np.float32, copy=False)

def normalize_across_fallback(subject_list, eps=1e-8):
    """
    Across-subject z-scoring fallback:
    stack all subject timepoints and z-score each feature globally.
    """
    subj = [coerce_TxV(x) for x in subject_list]
    stacked = np.vstack(subj)
    mu = stacked.mean(axis=0, keepdims=True)
    sd = stacked.std(axis=0, keepdims=True) + eps
    return [((x - mu) / sd).astype(np.float32, copy=False) for x in subj]

def normalize_subject_list(subject_list):
    subj = [coerce_TxV(x) for x in subject_list]
    if HT_NORMALIZE_AVAILABLE:
        try:
            return ht.normalize(subj, normalize="across")
        except Exception:
            return normalize_across_fallback(subj)
    return normalize_across_fallback(subj)

def load_pieman_data(
    mat_path=PIEMAN_MAT_PATH,
    conditions_to_use=CONDITIONS_TO_USE,
    trim_rest_by=TRIM_REST_BY,
):
    """
    Returns:
      condition_data: dict[str, list[np.ndarray]] with each subject as T x V
      condition_labels: list[str] for across-condition concatenated subjects
      all_subjects: list[np.ndarray]
    """
    pieman_data = loadmat(mat_path)

    # Original dataset keys in prior notebook:
    # intact, paragraph, word, rest
    all_possible = ["intact", "paragraph", "word", "rest"]
    missing = [c for c in conditions_to_use if c not in all_possible]
    if missing:
        raise ValueError(f"Unknown condition(s): {missing}")

    data = []
    conds = []
    for c in conditions_to_use:
        if c == "paragraph":
            # matches previous notebook behavior: exclude one subject at index 3
            next_data = list(map(lambda i: pieman_data[c][:, i][0],
                                 np.where(np.arange(pieman_data[c].shape[1]) != 3)[0]))
        else:
            next_data = list(map(lambda i: pieman_data[c][:, i][0],
                                 np.arange(pieman_data[c].shape[1])))
        data.extend(next_data)
        conds.extend([c] * len(next_data))

    conds_array = np.array(conds)
    condition_data = {}
    for c in conditions_to_use:
        cur = [data[i] for i in np.where(conds_array == c)[0]]
        if c == "rest" and trim_rest_by > 0:
            cur = array_cutter(cur, trim_rest_by)
        condition_data[c] = normalize_subject_list(cur)

    # across-condition concatenation
    all_subjects = []
    condition_labels = []
    for c in conditions_to_use:
        all_subjects.extend(condition_data[c])
        condition_labels.extend([c] * len(condition_data[c]))

    return condition_data, np.array(condition_labels), all_subjects

condition_data, condition_labels_str, all_subjects = load_pieman_data()

print("Conditions loaded:")
for c, arrs in condition_data.items():
    print(f"  {c}: n={len(arrs)}, example shape={coerce_TxV(arrs[0]).shape}")
print(f"Total subjects: {len(all_subjects)}")
```

## Orientation helpers

```

def prepare_subject_matrix_for_mode(X, aa_mode):
    """
    Convert a single subject matrix into the orientation expected by MSAA.

    Paper terminology:
      - temporal AA: shared spatial maps -> fit on T x V
      - spatial AA: shared temporal motifs -> fit on V x T
    """
    Xtv = coerce_TxV(X)
    if aa_mode == "temporal":
        return Xtv           # T x V
    elif aa_mode == "spatial":
        return Xtv.T         # V x T
    else:
        raise ValueError("aa_mode must be 'spatial' or 'temporal'")

def build_subject_objects(subject_list, aa_mode):
    subj_objs = []
    for X in subject_list:
        Xprep = prepare_subject_matrix_for_mode(X, aa_mode)
        subj_objs.append(Subject(X=Xprep, sX=Xprep))
    return subj_objs
```

## MSAA fitting

```

def run_msaa_once(subject_list, aa_mode, K, opts=None, verbose=True):
    """
    Fit MSAA once for a given orientation and K.
    """
    opts = dict(MSAA_OPTS if opts is None else opts)
    subj_objs = build_subject_objects(subject_list, aa_mode=aa_mode)

    t0 = time.time()
    results_subj, C, cost_fun, varexpl, secs = multi_subject_aa(
        subj_objs,
        noc=K,
        opts=opts,
    )
    elapsed = time.time() - t0

    if verbose:
        print(f"[{aa_mode}] K={K} | runtime={secs:.1f}s (wall={elapsed:.1f}s) | VarExpl={varexpl*100:.2f}%")
        print("  C shape:", np.asarray(C).shape)
        if len(results_subj) > 0:
            print("  subj0 keys:", list(results_subj[0].keys()))
            if "sXC" in results_subj[0]:
                print("  subj0 sXC shape:", np.asarray(results_subj[0]["sXC"]).shape)
            if "S" in results_subj[0]:
                print("  subj0 S shape:", np.asarray(results_subj[0]["S"]).shape)

    return {
        "results_subj": results_subj,
        "C": C,
        "cost_fun": cost_fun,
        "varexpl": varexpl,
        "secs": secs,
        "K": K,
        "aa_mode": aa_mode,
        "opts": opts,
    }

def run_msaa_sweep(subject_list, aa_mode, k_values, opts=None, verbose=True):
    all_results = {}
    for K in k_values:
        all_results[K] = run_msaa_once(
            subject_list=subject_list,
            aa_mode=aa_mode,
            K=K,
            opts=opts,
            verbose=verbose,
        )
    return all_results
```

## Standardize MSAA outputs into common arrays

```

def standardize_results_arrays(results_subj, aa_mode, original_subject_list):
    """
    Convert raw per-subject MSAA outputs into a common representation:
      time_array   : N x T x K
      spatial_array: N x K x V

    Inputs:
      results_subj: list of subject result dicts from multi_subject_aa
      aa_mode: 'temporal' or 'spatial' in the paper's terminology
      original_subject_list: the original subject data as T x V arrays

    Notes:
    - For temporal AA:
        raw input to MSAA was T x V
        expected raw shapes: sXC = T x K, S = K x V
        so we can use them directly.
    - For spatial AA:
        raw input to MSAA was V x T
        expected raw shapes: sXC = V x K, S = K x T
        We convert to:
          time_array[s,:,k]    = S[k,:]        (temporal motif over T)
          spatial_array[s,k,:] = sXC[:,k]      (spatial weights over V)
    """
    time_list = []
    spatial_list = []

    for i, sub in enumerate(results_subj):
        Xtv = coerce_TxV(original_subject_list[i])
        T, V = Xtv.shape

        sXC = np.asarray(sub["sXC"], dtype=float)
        S = np.asarray(sub["S"], dtype=float)

        if aa_mode == "temporal":
            # expected: sXC = T x K, S = K x V
            if sXC.shape[0] != T and sXC.shape[1] == T:
                sXC = sXC.T
            if S.shape[1] != V and S.shape[0] == V:
                S = S.T
            if sXC.shape[0] != T:
                raise ValueError(f"Temporal AA: expected sXC first dim {T}, got {sXC.shape}")
            if S.shape[1] != V:
                raise ValueError(f"Temporal AA: expected S second dim {V}, got {S.shape}")

            time_list.append(sXC)   # T x K
            spatial_list.append(S)  # K x V

        elif aa_mode == "spatial":
            # raw fit was V x T
            # expected: sXC = V x K, S = K x T
            if sXC.shape[0] != V and sXC.shape[1] == V:
                sXC = sXC.T
            if S.shape[1] != T and S.shape[0] == T:
                S = S.T
            if sXC.shape[0] != V:
                raise ValueError(f"Spatial AA: expected sXC first dim {V}, got {sXC.shape}")
            if S.shape[1] != T:
                raise ValueError(f"Spatial AA: expected S second dim {T}, got {S.shape}")

            # convert to common convention
            time_list.append(S.T)        # T x K
            spatial_list.append(sXC.T)   # K x V

        else:
            raise ValueError("aa_mode must be 'spatial' or 'temporal'")

    time_array = np.stack(time_list, axis=0)       # N x T x K
    spatial_array = np.stack(spatial_list, axis=0) # N x K x V
    return time_array, spatial_array
```

## Feature construction

```

def _lag1_autocorr(x):
    x = np.asarray(x, dtype=float)
    x = x - np.mean(x)
    denom = np.dot(x, x)
    if denom == 0 or np.isnan(denom):
        return 0.0
    num = np.dot(x[:-1], x[1:])
    if np.isnan(num):
        return 0.0
    return float(num / denom)

def _spectral_entropy(x, fs, nperseg=None):
    from scipy.signal import welch
    from scipy.stats import entropy
    x = np.asarray(x, dtype=float)
    nperseg = nperseg or min(256, len(x))
    f, Pxx = welch(x, fs=fs, nperseg=nperseg)
    if np.sum(Pxx) == 0 or np.isnan(np.sum(Pxx)):
        return 0.0
    P = Pxx / (np.sum(Pxx) + 1e-12)
    return float(entropy(P))

def _band_power(x, fs, band, nperseg=None):
    from scipy.signal import welch
    x = np.asarray(x, dtype=float)
    nperseg = nperseg or min(256, len(x))
    f, Pxx = welch(x, fs=fs, nperseg=nperseg)
    if np.all(Pxx == 0):
        return 0.0
    idx = (f >= band[0]) & (f < band[1])
    return float(np.trapz(Pxx[idx], f[idx])) if np.any(idx) else 0.0

def _safe_gini(x):
    x = np.abs(np.asarray(x, dtype=float)).ravel()
    if np.allclose(x, 0):
        return 0.0
    x = np.sort(x)
    n = len(x)
    index = np.arange(1, n + 1)
    return float((2 * np.sum(index * x) / (n * np.sum(x))) - (n + 1) / n)

def make_df_features_from_arrays(
    time_array,
    spatial_array,
    condition_codes,
    condition_map=None,
    analysis_type="spatial",
    include_energy=True,
    include_temporal=True,
    include_spatial=True,
    fs=FS,
    bands=BANDS,
    log1p_energy=LOG1P_ENERGY,
    zscore_across_subjects=ZSCORE_FEATURE_COLUMNS,
):
    """
    Build subject-level feature dataframe from:
      time_array   : N x T x K
      spatial_array: N x K x V
    """
    N, T, K = time_array.shape
    N2, K2, V = spatial_array.shape
    if N != N2 or K != K2:
        raise ValueError("time_array and spatial_array are inconsistent")

    rows = []
    for s in range(N):
        feats = {
            "Subject": s,
            "Condition": int(condition_codes[s]),
            "AnalysisType": analysis_type,
        }

        for k in range(K):
            tvec = time_array[s, :, k]
            svec = spatial_array[s, k, :]

            if include_energy:
                energy = float(np.sum(tvec**2) * np.sum(svec**2))
                if log1p_energy:
                    energy = float(np.log1p(energy))
                feats[f"A{k+1}_energy"] = energy

            if include_temporal:
                feats[f"A{k+1}_t_mean"] = float(np.mean(tvec))
                feats[f"A{k+1}_t_std"] = float(np.std(tvec))
                feats[f"A{k+1}_t_ac1"] = _lag1_autocorr(tvec)
                feats[f"A{k+1}_t_specent"] = _spectral_entropy(tvec, fs=fs)
                for b_idx, band in enumerate(bands):
                    feats[f"A{k+1}_t_bp{b_idx}"] = _band_power(tvec, fs=fs, band=band)

            if include_spatial:
                feats[f"A{k+1}_s_mean"] = float(np.mean(svec))
                feats[f"A{k+1}_s_std"] = float(np.std(svec))
                feats[f"A{k+1}_s_l1"] = float(np.sum(np.abs(svec)))
                feats[f"A{k+1}_s_l2"] = float(np.sqrt(np.sum(svec**2)))
                feats[f"A{k+1}_s_maxabs"] = float(np.max(np.abs(svec)))
                feats[f"A{k+1}_s_gini"] = _safe_gini(svec)

        rows.append(feats)

    df_features = pd.DataFrame(rows)

    if condition_map is not None:
        df_features["ConditionName"] = (
            pd.Series(df_features["Condition"]).map(condition_map).astype("category")
        )

    if zscore_across_subjects:
        non_feature_cols = {"Subject", "Condition", "ConditionName", "AnalysisType"}
        feature_cols = [c for c in df_features.columns if c not in non_feature_cols]
        scaler = StandardScaler()
        df_features[feature_cols] = scaler.fit_transform(df_features[feature_cols])

    return df_features

def build_subject_feature_matrix(
    df_features,
    mode="all",  # "energy_only" | "temporal_only" | "spatial_only" | "all"
):
    energy_cols = [c for c in df_features.columns if c.startswith("A") and c.endswith("_energy")]
    temporal_cols = [c for c in df_features.columns if c.startswith("A") and "_t_" in c]
    spatial_cols = [c for c in df_features.columns if c.startswith("A") and "_s_" in c]

    if mode == "energy_only":
        cols = energy_cols
    elif mode == "temporal_only":
        cols = temporal_cols
    elif mode == "spatial_only":
        cols = spatial_cols
    elif mode == "all":
        cols = energy_cols + temporal_cols + spatial_cols
    else:
        raise ValueError("mode must be 'energy_only', 'temporal_only', 'spatial_only', or 'all'")

    X_raw = df_features[cols].to_numpy()
    cond = pd.Categorical(df_features["Condition"]).codes
    subj_ids = df_features["Subject"].to_numpy()

    imputer = SimpleImputer(strategy="constant", fill_value=0.0)
    X_imp = imputer.fit_transform(X_raw)
    X = StandardScaler().fit_transform(X_imp)

    return X, cond, subj_ids, cols
```

## Clustering metrics and plotting

```

def purity_score(y_cluster, y_class):
    y_cluster = np.asarray(y_cluster)
    y_class = np.asarray(y_class)
    total = 0
    for c in np.unique(y_cluster):
        mask = (y_cluster == c)
        _, counts = np.unique(y_class[mask], return_counts=True)
        total += counts.max()
    return total / len(y_cluster)

def clustering_metrics(y_cluster, y_class):
    return {
        "purity": purity_score(y_cluster, y_class),
        "ari": adjusted_rand_score(y_class, y_cluster),
        "nmi": normalized_mutual_info_score(y_class, y_cluster),
    }

def permutation_null_purity(y_cluster, y_class, n_perm=N_PERM, random_state=RNG_SEED):
    rng = np.random.default_rng(random_state)
    y_cluster = np.asarray(y_cluster)
    y_class = np.asarray(y_class)
    null_vals = np.zeros(n_perm, dtype=float)
    for i in range(n_perm):
        shuffled = rng.permutation(y_class)
        null_vals[i] = purity_score(y_cluster, shuffled)
    return null_vals

def hierarchical_cluster_subjects(X, cond, n_clusters=N_CLUSTERS, linkage_method="average", feature_metric="euclidean"):
    dvec = pdist(X, metric=feature_metric)
    Z = linkage(dvec, method=linkage_method)
    order = leaves_list(Z)
    cluster_labels = fcluster(Z, n_clusters, criterion="maxclust")

    metrics = clustering_metrics(cluster_labels, cond)

    sim = np.corrcoef(X)
    sim_sorted = sim[np.ix_(order, order)]
    cond_sorted = np.asarray(cond)[order]
    cluster_sorted = cluster_labels[order]
    boundaries = np.where(np.diff(cluster_sorted) != 0)[0] + 1

    return {
        "Z": Z,
        "order": order,
        "cluster_labels": cluster_labels,
        "cluster_sorted": cluster_sorted,
        "cond_sorted": cond_sorted,
        "sim": sim,
        "sim_sorted": sim_sorted,
        "boundaries": boundaries,
        **metrics,
    }

def evaluate_cluster_solution(X, cond, n_clusters=N_CLUSTERS, linkage_method="average", feature_metric="euclidean",
                              n_perm=N_PERM, random_state=RNG_SEED):
    result = hierarchical_cluster_subjects(
        X=X,
        cond=cond,
        n_clusters=n_clusters,
        linkage_method=linkage_method,
        feature_metric=feature_metric,
    )
    null_purity = permutation_null_purity(
        y_cluster=result["cluster_labels"],
        y_class=cond,
        n_perm=n_perm,
        random_state=random_state,
    )
    result["null_purity"] = null_purity
    result["null_mean"] = float(np.mean(null_purity))
    result["null_std"] = float(np.std(null_purity))
    result["null_p95"] = float(np.percentile(null_purity, 95))
    result["p_perm"] = float((np.sum(null_purity >= result["purity"]) + 1) / (len(null_purity) + 1))
    return result

def plot_clustered_similarity(result, cond_colors=None, title="Subject-by-subject similarity",
                              vmin=-0.5, vmax=1.0, figsize=(8, 7), show_subject_ids=True):
    if cond_colors is None:
        cond_colors = {0: "purple", 1: "green", 2: "black"}

    sim_sorted = result["sim_sorted"]
    order = result["order"]
    cond_sorted = result["cond_sorted"]
    boundaries = result["boundaries"]

    plt.figure(figsize=figsize)
    ax = sns.heatmap(
        sim_sorted,
        cmap="coolwarm",
        vmin=vmin,
        vmax=vmax,
        square=True,
        xticklabels=(order + 1) if show_subject_ids else False,
        yticklabels=(order + 1) if show_subject_ids else False,
        cbar_kws={"label": "Pearson correlation"},
    )

    if show_subject_ids:
        tick_colors = [cond_colors.get(int(c), "gray") for c in cond_sorted]
        for tick_label, color in zip(ax.get_xticklabels(), tick_colors):
            tick_label.set_color(color)
        for tick_label, color in zip(ax.get_yticklabels(), tick_colors):
            tick_label.set_color(color)

    for b in boundaries:
        ax.axhline(b, color="black", linewidth=2)
        ax.axvline(b, color="black", linewidth=2)

    plt.title(
        f"{title}\nPurity={result['purity']:.3f}, "
        f"ARI={result['ari']:.3f}, NMI={result['nmi']:.3f}"
    )
    plt.xlabel("Subjects (clustered)")
    plt.ylabel("Subjects (clustered)")
    plt.tight_layout()
    plt.show()

def plot_purity_null(result, bins=30, figsize=(6, 4)):
    plt.figure(figsize=figsize)
    plt.hist(result["null_purity"], bins=bins, alpha=0.8)
    plt.axvline(result["purity"], linestyle="--", linewidth=2)
    plt.title(
        "Permutation null for purity\n"
        f"Observed={result['purity']:.3f}, "
        f"Null mean={result['null_mean']:.3f}, "
        f"p={result['p_perm']:.4f}"
    )
    plt.xlabel("Purity")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()
```

## Per-archetype scans

```

def _impute_and_zscore_rows(X):
    X = np.asarray(X, dtype=float)
    imp = SimpleImputer(strategy="constant", fill_value=0.0)
    X = imp.fit_transform(X)
    X = zscore(X, axis=1, ddof=0, nan_policy="omit")
    X = np.nan_to_num(X)
    return X

def subject_vectors_for_archetype(time_array, spatial_array, k, mode="A"):
    """
    mode:
      'A' -> use temporal trajectory for archetype k (N x T)
      'S' -> use spatial coefficient vector for archetype k (N x V)
    """
    if mode == "A":
        return time_array[:, :, k]
    elif mode == "S":
        return spatial_array[:, k, :]
    else:
        raise ValueError("mode must be 'A' or 'S'")

def evaluate_single_archetype(time_array, spatial_array, condition_codes, k, mode="A",
                              n_clusters=N_CLUSTERS, linkage_method="average",
                              feature_metric="euclidean", n_perm=N_PERM, random_state=RNG_SEED):
    Xk = subject_vectors_for_archetype(time_array, spatial_array, k=k, mode=mode)
    Xk = _impute_and_zscore_rows(Xk)
    cond = np.asarray(condition_codes)
    result = evaluate_cluster_solution(
        X=Xk,
        cond=cond,
        n_clusters=n_clusters,
        linkage_method=linkage_method,
        feature_metric=feature_metric,
        n_perm=n_perm,
        random_state=random_state,
    )
    result["k"] = k + 1
    result["mode"] = mode
    return result

def scan_all_archetypes(time_array, spatial_array, condition_codes, mode="A",
                        n_clusters=N_CLUSTERS, linkage_method="average",
                        feature_metric="euclidean", n_perm=500, random_state=RNG_SEED,
                        sort_by="purity"):
    K = time_array.shape[2]
    rows = []
    for k in range(K):
        result = evaluate_single_archetype(
            time_array=time_array,
            spatial_array=spatial_array,
            condition_codes=condition_codes,
            k=k,
            mode=mode,
            n_clusters=n_clusters,
            linkage_method=linkage_method,
            feature_metric=feature_metric,
            n_perm=n_perm,
            random_state=random_state,
        )
        rows.append({
            "k": result["k"],
            "mode": result["mode"],
            "purity": result["purity"],
            "ari": result["ari"],
            "nmi": result["nmi"],
            "null_mean": result["null_mean"],
            "null_p95": result["null_p95"],
            "p_perm": result["p_perm"],
        })
    df = pd.DataFrame(rows)
    return df.sort_values(sort_by, ascending=False).reset_index(drop=True)
```

## Run the full K sweep

```

results_by_k = run_msaa_sweep(
    subject_list=all_subjects,
    aa_mode=AA_MODE,
    k_values=K_VALUES,
    opts=MSAA_OPTS,
    verbose=True,
)
```

## Build features and clustering results for every K

```

analysis_by_k = {}
summary_rows = []

for K, res in results_by_k.items():
    time_array, spatial_array = standardize_results_arrays(
        results_subj=res["results_subj"],
        aa_mode=AA_MODE,
        original_subject_list=all_subjects,
    )

    df_features = make_df_features_from_arrays(
        time_array=time_array,
        spatial_array=spatial_array,
        condition_codes=condition_codes,
        condition_map=condition_map,
        analysis_type=AA_MODE,
        include_energy=True,
        include_temporal=True,
        include_spatial=True,
        fs=FS,
        bands=BANDS,
        log1p_energy=LOG1P_ENERGY,
        zscore_across_subjects=ZSCORE_FEATURE_COLUMNS,
    )

    X, cond, subj_ids, used_cols = build_subject_feature_matrix(
        df_features,
        mode=FEATURE_MODE,
    )

    cluster_result = evaluate_cluster_solution(
        X,
        cond,
        n_clusters=N_CLUSTERS,
        linkage_method="average",
        feature_metric="euclidean",
        n_perm=N_PERM,
        random_state=RNG_SEED,
    )

    analysis_by_k[K] = {
        **res,
        "time_array": time_array,
        "spatial_array": spatial_array,
        "df_features": df_features,
        "X": X,
        "cond": cond,
        "subj_ids": subj_ids,
        "used_cols": used_cols,
        "cluster_result": cluster_result,
    }

    summary_rows.append({
        "K": K,
        "aa_mode": AA_MODE,
        "feature_mode": FEATURE_MODE,
        "varexpl": res["varexpl"],
        "secs": res["secs"],
        "n_features": X.shape[1],
        "purity": cluster_result["purity"],
        "ari": cluster_result["ari"],
        "nmi": cluster_result["nmi"],
        "null_mean": cluster_result["null_mean"],
        "null_p95": cluster_result["null_p95"],
        "p_perm": cluster_result["p_perm"],
    })

summary_df = pd.DataFrame(summary_rows).sort_values("K").reset_index(drop=True)
summary_df
```


## Notes

- The **ideal purity is 1.0** when `N_CLUSTERS = 3` and the 3 conditions separate perfectly.
- Purity can still be optimistic, so **ARI** and **NMI** are included and should usually be reported alongside purity.
- `FEATURE_MODE` lets you compare:
  - `"energy_only"`
  - `"temporal_only"`
  - `"spatial_only"`
  - `"all"`
- For **Spatial AA** in the paper's terminology, the most interpretable archetype scan is often `ARCH_SCAN_MODE = "S"` because it clusters subjects based on the spatial coefficient vectors associated with each temporal motif.
- For **Temporal AA** in the paper's terminology, `ARCH_SCAN_MODE = "A"` often aligns more directly with time-resolved archetype expression.

## Suggested reporting

A clean write-up would be:

> Subjects were clustered into 3 groups, matching the 3 experimental conditions. We quantified clustering quality using purity, adjusted Rand index (ARI), and normalized mutual information (NMI). Because purity can be inflated even under imperfect clustering, we additionally compared observed purity against a permutation null in which condition labels were randomly shuffled while cluster assignments were held fixed.
