
# Spatial MSAA across-condition clustering rewrite

This notebook rewrites the **subject similarity / clustering / purity** portion of the across-all-conditions Spatial AA analysis.

## What changed

- **Fix the number of clusters at 3** by default, since there are 3 conditions.
- Compute clustering **directly on the subject feature matrix** rather than on a distance matrix built from a similarity matrix. This avoids the `ClusterWarning` you were seeing and is easier to interpret.
- Report **Purity, ARI, and NMI** together.
- Add a **permutation null** for purity so you can compare the observed value against a chance baseline.
- Keep the **clustered subject-by-subject similarity heatmap** for visualization.
- Provide helpers to scan **all archetypes** and identify “winning” archetypes.

## Interpretation

With 3 clusters and 3 conditions, the **ideal purity is 1.0**:
- each cluster contains subjects from only one condition.

Purity alone can still be misleading, so this notebook also reports:
- **ARI** = Adjusted Rand Index
- **NMI** = Normalized Mutual Information

Both also have an ideal value of **1.0**.


```

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from scipy.cluster.hierarchy import linkage, leaves_list, fcluster
from scipy.spatial.distance import pdist, squareform
from scipy.stats import zscore

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score
```


## Core metrics


```

def purity_score(y_cluster, y_class):
    """
    Purity = (1/N) * sum over clusters of the majority class count in each cluster.

    Parameters
    ----------
    y_cluster : array-like of shape (N,)
        Cluster assignments. Can be any integers/labels.
    y_class : array-like of shape (N,)
        Ground-truth condition labels.

    Returns
    -------
    purity : float
        Value in [0, 1]. Ideal = 1.0.
    """
    y_cluster = np.asarray(y_cluster)
    y_class = np.asarray(y_class)

    total = 0
    for c in np.unique(y_cluster):
        mask = (y_cluster == c)
        classes, counts = np.unique(y_class[mask], return_counts=True)
        total += counts.max()
    return total / len(y_cluster)


def clustering_metrics(y_cluster, y_class):
    """
    Compute purity, ARI, and NMI.
    """
    return {
        "purity": purity_score(y_cluster, y_class),
        "ari": adjusted_rand_score(y_class, y_cluster),
        "nmi": normalized_mutual_info_score(y_class, y_cluster),
    }


def permutation_null_purity(y_cluster, y_class, n_perm=2000, random_state=0):
    """
    Permutation null for purity:
    hold cluster assignments fixed, shuffle condition labels.
    """
    rng = np.random.default_rng(random_state)
    y_cluster = np.asarray(y_cluster)
    y_class = np.asarray(y_class)

    null_vals = np.zeros(n_perm, dtype=float)
    for i in range(n_perm):
        shuffled = rng.permutation(y_class)
        null_vals[i] = purity_score(y_cluster, shuffled)
    return null_vals
```


## Feature preparation

These helpers are designed to plug into your existing subject-level feature dataframes, e.g.
- energy-only features,
- temporal summary features,
- or all features together.

They preserve the logic you already had, but make clustering cleaner.


```

def build_subject_feature_matrix(
    df_features: pd.DataFrame,
    mode: str = "energy_only",      # 'energy_only' | 'temporal_only' | 'all'
    top_m_archetypes: int | None = None,
    arch_importance_df: pd.DataFrame | None = None,  # columns: ['Archetype','Importance']
    custom_archetype_idxs: list[int] | None = None,  # 1-based indices, e.g. [1, 7, 12]
):
    """
    Build a clean, subject x feature matrix for clustering/classification.

    Returns
    -------
    X : ndarray, shape (N_subjects, N_features)
        Imputed + z-scored feature matrix
    cond : ndarray, shape (N_subjects,)
        Condition labels encoded as categorical integers
    subj_ids : ndarray, shape (N_subjects,)
        Subject indices if available
    used_cols : list[str]
        Feature column names retained
    """
    df = df_features.copy()

    energy_cols = [c for c in df.columns if c.startswith("A") and c.endswith("_energy")]
    temporal_cols = [c for c in df.columns if c.startswith("A") and not c.endswith("_energy")]

    if mode == "energy_only":
        cols = energy_cols
    elif mode == "temporal_only":
        cols = temporal_cols
    elif mode == "all":
        cols = energy_cols + temporal_cols
    else:
        raise ValueError("mode must be 'energy_only', 'temporal_only', or 'all'")

    if top_m_archetypes is not None and arch_importance_df is not None:
        keep_arch = set(
            arch_importance_df.sort_values("Importance", ascending=False)
            .head(top_m_archetypes)["Archetype"].tolist()
        )
        cols = [c for c in cols if c.split("_")[0] in keep_arch]

    if custom_archetype_idxs is not None:
        keep_arch = {f"A{a}" for a in custom_archetype_idxs}
        cols = [c for c in cols if c.split("_")[0] in keep_arch]

    X_raw = df[cols].to_numpy()
    cond = pd.Categorical(df["Condition"]).codes
    subj_ids = df["Subject"].to_numpy() if "Subject" in df.columns else np.arange(len(df))

    imputer = SimpleImputer(strategy="constant", fill_value=0.0)
    X_imp = imputer.fit_transform(X_raw)
    X = StandardScaler().fit_transform(X_imp)

    return X, cond, subj_ids, cols
```


## Clustering directly on the feature matrix

This is the main change.

Instead of:
1. building a similarity matrix,
2. then clustering on Euclidean distances between similarity rows,

we cluster directly on the subject feature matrix `X`.

That makes the analysis easier to interpret:
- clustering is based on the actual subject-level archetype features;
- the heatmap is a visualization of the resulting organization.

We still compute the subject-by-subject similarity matrix for plotting.


```

def hierarchical_cluster_subjects(
    X,
    cond,
    n_clusters=3,
    linkage_method="average",
    feature_metric="euclidean",
):
    """
    Cluster subjects directly from the subject feature matrix X.

    Parameters
    ----------
    X : ndarray, shape (N, F)
        Subject-level feature matrix
    cond : ndarray, shape (N,)
        Condition labels
    n_clusters : int
        Number of clusters. Default = 3 for 3 conditions.
    linkage_method : str
        Linkage method for scipy.cluster.hierarchy.linkage
    feature_metric : str
        Distance metric for pdist on the subject feature matrix

    Returns
    -------
    result : dict
        clustering results and plotting helpers
    """
    # condensed distance vector for linkage (avoids ClusterWarning)
    dvec = pdist(X, metric=feature_metric)
    Z = linkage(dvec, method=linkage_method)

    order = leaves_list(Z)
    cluster_labels = fcluster(Z, n_clusters, criterion="maxclust")

    metrics = clustering_metrics(cluster_labels, cond)

    # subject similarity matrix for plotting only
    sim = np.corrcoef(X)
    sim_sorted = sim[np.ix_(order, order)]
    cond_sorted = np.asarray(cond)[order]
    cluster_sorted = cluster_labels[order]

    boundaries = np.where(np.diff(cluster_sorted) != 0)[0] + 1

    result = {
        "Z": Z,
        "order": order,
        "cluster_labels": cluster_labels,
        "cluster_sorted": cluster_sorted,
        "cond_sorted": cond_sorted,
        "sim": sim,
        "sim_sorted": sim_sorted,
        "boundaries": boundaries,
        **metrics,
    }
    return result
```

```

def plot_clustered_similarity(
    result,
    cond_colors=None,
    title="Subject-by-subject similarity",
    vmin=-0.5,
    vmax=1.0,
    figsize=(8, 7),
    show_subject_ids=True,
):
    """
    Plot the clustered subject-by-subject correlation heatmap.
    """
    if cond_colors is None:
        cond_colors = {0: "purple", 1: "green", 2: "black"}

    sim_sorted = result["sim_sorted"]
    order = result["order"]
    cond_sorted = result["cond_sorted"]
    boundaries = result["boundaries"]

    plt.figure(figsize=figsize)
    ax = sns.heatmap(
        sim_sorted,
        cmap="coolwarm",
        vmin=vmin,
        vmax=vmax,
        square=True,
        xticklabels=(order + 1) if show_subject_ids else False,
        yticklabels=(order + 1) if show_subject_ids else False,
        cbar_kws={"label": "Pearson correlation"},
    )

    if show_subject_ids:
        tick_colors = [cond_colors.get(int(c), "gray") for c in cond_sorted]
        for tick_label, color in zip(ax.get_xticklabels(), tick_colors):
            tick_label.set_color(color)
        for tick_label, color in zip(ax.get_yticklabels(), tick_colors):
            tick_label.set_color(color)

    for b in boundaries:
        ax.axhline(b, color="black", linewidth=2)
        ax.axvline(b, color="black", linewidth=2)

    plt.title(
        f"{title}\nPurity={result['purity']:.3f}, "
        f"ARI={result['ari']:.3f}, NMI={result['nmi']:.3f}"
    )
    plt.xlabel("Subjects (clustered)")
    plt.ylabel("Subjects (clustered)")
    plt.tight_layout()
    plt.show()
```


## Permutation null for purity


```

def evaluate_cluster_solution(
    X,
    cond,
    n_clusters=3,
    linkage_method="average",
    feature_metric="euclidean",
    n_perm=2000,
    random_state=0,
):
    """
    Run clustering + compute purity/ARI/NMI + permutation null for purity.
    """
    result = hierarchical_cluster_subjects(
        X=X,
        cond=cond,
        n_clusters=n_clusters,
        linkage_method=linkage_method,
        feature_metric=feature_metric,
    )

    null_purity = permutation_null_purity(
        y_cluster=result["cluster_labels"],
        y_class=cond,
        n_perm=n_perm,
        random_state=random_state,
    )

    result["null_purity"] = null_purity
    result["null_mean"] = float(np.mean(null_purity))
    result["null_std"] = float(np.std(null_purity))
    result["null_p95"] = float(np.percentile(null_purity, 95))
    result["p_perm"] = float((np.sum(null_purity >= result["purity"]) + 1) / (len(null_purity) + 1))
    return result


def plot_purity_null(result, bins=30, figsize=(6, 4)):
    """
    Plot the permutation null distribution for purity.
    """
    if "null_purity" not in result:
        raise ValueError("Run evaluate_cluster_solution(...) first.")

    plt.figure(figsize=figsize)
    plt.hist(result["null_purity"], bins=bins, alpha=0.8)
    plt.axvline(result["purity"], linestyle="--", linewidth=2)
    plt.title(
        "Permutation null for purity\n"
        f"Observed={result['purity']:.3f}, "
        f"Null mean={result['null_mean']:.3f}, "
        f"p={result['p_perm']:.4f}"
    )
    plt.xlabel("Purity")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()
```


## Per-archetype clustering scan

For your “winning archetypes” analysis, it usually makes more sense to:
1. build a subject-level vector for a given archetype,
2. cluster subjects into **3 clusters**,
3. report **Purity + ARI + NMI + permutation p-value**,
4. rank archetypes by one or more of these metrics.

Below are helpers for exactly that.


```

def _impute_and_zscore_rows(X):
    """
    Impute missing values and z-score each row.
    """
    X = np.asarray(X, dtype=float)
    imp = SimpleImputer(strategy="constant", fill_value=0.0)
    X = imp.fit_transform(X)
    X = zscore(X, axis=1, ddof=0, nan_policy="omit")
    X = np.nan_to_num(X)
    return X


def subject_vectors_for_archetype(group_archs_array, group_coeffs_array, k, mode="A"):
    """
    Build one subject-level vector per subject for archetype k.

    Parameters
    ----------
    group_archs_array : ndarray, shape (N, T, K)
        Subject-specific temporal archetype trajectories
    group_coeffs_array : ndarray, shape (N, K, V)
        Subject-specific spatial coefficient weights
    k : int
        0-based archetype index
    mode : str
        'A' -> use temporal trajectory for archetype k (shape N x T)
        'S' -> use spatial coefficient vector for archetype k (shape N x V)

    Returns
    -------
    Xk : ndarray
        subject x feature matrix for archetype k
    """
    if mode == "A":
        return group_archs_array[:, :, k]
    elif mode == "S":
        return group_coeffs_array[:, k, :]
    else:
        raise ValueError("mode must be 'A' or 'S'")
```

```

def evaluate_single_archetype(
    group_archs_array,
    group_coeffs_array,
    condition_labels,
    k,
    mode="A",
    n_clusters=3,
    linkage_method="average",
    feature_metric="euclidean",
    n_perm=1000,
    random_state=0,
):
    """
    Cluster subjects for a single archetype and return all metrics.
    """
    Xk = subject_vectors_for_archetype(group_archs_array, group_coeffs_array, k=k, mode=mode)
    Xk = _impute_and_zscore_rows(Xk)
    cond = np.asarray(condition_labels)

    result = evaluate_cluster_solution(
        X=Xk,
        cond=cond,
        n_clusters=n_clusters,
        linkage_method=linkage_method,
        feature_metric=feature_metric,
        n_perm=n_perm,
        random_state=random_state,
    )
    result["k"] = k + 1  # 1-based for reporting
    result["mode"] = mode
    return result
```

```

def plot_single_archetype(
    group_archs_array,
    group_coeffs_array,
    condition_labels,
    k,
    mode="A",
    n_clusters=3,
    linkage_method="average",
    feature_metric="euclidean",
    n_perm=1000,
    random_state=0,
    cond_colors=None,
    title_prefix="Archetype",
):
    """
    Plot clustered heatmap + null histogram for one archetype.
    """
    result = evaluate_single_archetype(
        group_archs_array=group_archs_array,
        group_coeffs_array=group_coeffs_array,
        condition_labels=condition_labels,
        k=k,
        mode=mode,
        n_clusters=n_clusters,
        linkage_method=linkage_method,
        feature_metric=feature_metric,
        n_perm=n_perm,
        random_state=random_state,
    )

    plot_clustered_similarity(
        result,
        cond_colors=cond_colors,
        title=f"{title_prefix} {k+1} ({mode})",
    )
    plot_purity_null(result)

    return result
```

```

def scan_all_archetypes(
    group_archs_array,
    group_coeffs_array,
    condition_labels,
    mode="A",
    n_clusters=3,
    linkage_method="average",
    feature_metric="euclidean",
    n_perm=500,
    random_state=0,
    sort_by="purity",
):
    """
    Scan all archetypes and summarize clustering metrics.
    """
    K = group_archs_array.shape[2]
    rows = []

    for k in range(K):
        result = evaluate_single_archetype(
            group_archs_array=group_archs_array,
            group_coeffs_array=group_coeffs_array,
            condition_labels=condition_labels,
            k=k,
            mode=mode,
            n_clusters=n_clusters,
            linkage_method=linkage_method,
            feature_metric=feature_metric,
            n_perm=n_perm,
            random_state=random_state,
        )
        rows.append({
            "k": result["k"],
            "mode": result["mode"],
            "purity": result["purity"],
            "ari": result["ari"],
            "nmi": result["nmi"],
            "null_mean": result["null_mean"],
            "null_p95": result["null_p95"],
            "p_perm": result["p_perm"],
        })

    df = pd.DataFrame(rows)
    return df.sort_values(sort_by, ascending=False).reset_index(drop=True)
```


## Recommended reporting language

A cleaner interpretation is:

- We clustered subjects into **3 clusters**, matching the **3 experimental conditions**.
- The **ideal purity is 1.0**, corresponding to complete condition separation.
- Because purity can be inflated even under imperfect clustering, we also report **ARI** and **NMI**, and compare observed purity against a **permutation null**.
- Archetypes can then be ranked by purity, ARI, NMI, or a composite score.

## Optional composite score

If you want a single ranking score, one reasonable choice is:
`0.5 * purity + 0.25 * ari + 0.25 * nmi`

I did not hard-code that here, but it is easy to add once you see which metric feels most stable in your data.
