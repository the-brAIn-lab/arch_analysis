
# 03_clustering_and_brainplots_from_npz_yeo_revised.ipynb

Run this third.

This version uses the **Yeo 7-network atlas directly** for node labeling, instead of the Schaefer parcel atlas.

It keeps the same overall plotting style from your uploaded notebook:
- Yeo network labels
- original network colors
- RBF-based node assignment from atlas voxels to your centers
- `plot_connectome(..., node_color=[colors[i] for i in node_codes['code']], display_mode='lyrz')`

Workflow:
1. load `.npz` MSAA outputs from Notebook 1
2. compute timecourse-archetype similarity clustering
3. select top clustering archetypes
4. plot their corresponding spatial coefficient maps using Yeo-based network colors


```

LOAD_DIR = "msaa_outputs_npz"
K_VALUES = [10, 25]
N_CLUSTERS = 3
TOP_N_CLUSTER_ARCHETYPES = 3

COND_COLORS = {0: "purple", 1: "green", 2: "black"}

# Point this to your Yeo 7-network atlas file
YEO_NII = os.path.join("data/pieman/raw", "Yeo2011_7Networks_MNI152_FreeSurferConformed1mm.nii.gz")
```

```

%matplotlib inline
import os
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from matplotlib.colors import ListedColormap
from scipy.cluster.hierarchy import linkage, leaves_list, fcluster
from scipy.spatial.distance import squareform
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score

try:
    import nibabel as nib
    from nilearn.input_data import NiftiMasker
    from nilearn import plotting as niplot
    NILEARN_AVAILABLE = True
except Exception:
    NILEARN_AVAILABLE = False
    print("nilearn / nibabel not available; brain plotting disabled.")

print("Ready.")
```

```

def load_msaa_npz(path):
    data = np.load(path, allow_pickle=True)
    results_subj = data["results_subj"].tolist()
    if isinstance(results_subj, np.ndarray):
        results_subj = results_subj.tolist()
    lookup_table = dict(zip(data["lookup_table_keys"].tolist(), data["lookup_table_vals"].tolist()))
    network_colors = dict(zip(data["network_color_keys"].tolist(), data["network_color_vals"].tolist()))
    return {
        "K": int(data["K"]),
        "results_subj": results_subj,
        "condition_labels_str": data["condition_labels_str"].tolist(),
        "condition_codes": data["condition_codes"],
        "centers": data["centers"],
        "widths": int(data["widths"]),
        "lookup_table": lookup_table,
        "network_colors": network_colors,
    }

loaded = {}
for K in K_VALUES:
    path = os.path.join(LOAD_DIR, f"spatialAA_acrossCond_K{K}.npz")
    loaded[K] = load_msaa_npz(path)
    print("Loaded:", path)
```

## Yeo 7-network labeling setup

```

# Original labels/colors matching your plotting style
lookup_table = {
    1: 'Visual',
    2: 'Somatomotor',
    3: 'Dorsal attention',
    4: 'Ventral attention',
    5: 'Limbic',
    6: 'Frontoparietal',
    7: 'Default mode'
}

network_colors = {
    'Visual': '#D7DF23',
    'Somatomotor': '#39B54A',
    'Dorsal attention': '#00A79D',
    'Ventral attention': '#27AAE1',
    'Limbic': '#1C75BC',
    'Frontoparietal': '#92278F',
    'Default mode': '#EE2A7B'
}

colors = ['#888888']
colors.extend([network_colors[lookup_table[i]] for i in range(1, 8)])

network_cmap = ListedColormap(colors, N=len(colors) * 2, name='networks')
network_codes = {v: k for k, v in lookup_table.items()}

print(network_codes)
```

```

def fullfact(levels):
    mesh = np.meshgrid(*[np.arange(1, int(l) + 1) for l in levels], indexing="ij")
    return np.stack([m.ravel(order="C") for m in mesh], axis=1)

def nii2cmu(nifti_file, mask_file=None):
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        if isinstance(nifti_file, str):
            img = nib.load(nifti_file)
        else:
            img = nifti_file

        mask = NiftiMasker(mask_strategy='background')
        if mask_file is None:
            mask.fit(nifti_file)
        else:
            mask.fit(mask_file)

    S = img.get_sform()
    Y = np.float32(mask.transform(nifti_file)).copy()
    vmask = np.nonzero(np.array(np.reshape(mask.mask_img_.dataobj, (1, np.prod(mask.mask_img_.shape)), order='C')))[1]
    vox_coords = fullfact(img.shape[0:3])[vmask, ::-1] - 1
    R = np.array(np.dot(vox_coords, S[0:3, 0:3])) + S[:3, 3]
    return {'Y': Y, 'R': R}

def rbf(R, center, width):
    return np.exp(-np.sum((R - center) ** 2, axis=1) / width)

def node_labels(centers, widths, networks_cmu):
    labels = []
    if np.isscalar(widths):
        widths_iter = [widths] * len(centers)
    else:
        widths_iter = widths

    Y = networks_cmu['Y'].ravel().astype(int)

    for c, w in zip(centers, widths_iter):
        r = rbf(networks_cmu['R'], c, w)
        label_weights = [np.sum(r[Y == i]) for i in range(1, 8)]
        labels.append(np.argmax(label_weights) + 1)

    return pd.DataFrame({
        'code': labels,
        'Network': [lookup_table[i] for i in labels]
    })
```

```

if not os.path.exists(YEO_NII):
    raise FileNotFoundError(
        f"Could not find YEO_NII at {YEO_NII}. "
        f"Update the YEO_NII path to your actual Yeo atlas file."
    )

networks_cmu = nii2cmu(YEO_NII)

# keep only canonical 7-network labels
networks_cmu['Y'] = np.atleast_2d(np.asarray(networks_cmu['Y']).ravel())
valid = np.isin(networks_cmu['Y'].ravel(), np.arange(1, 8))
networks_cmu['R'] = networks_cmu['R'][valid]
networks_cmu['Y'] = np.atleast_2d(networks_cmu['Y'].ravel()[valid]).T

print("R shape:", networks_cmu['R'].shape)
print("Y shape:", networks_cmu['Y'].shape)
```

## Template and legend plots

```

def plot_network_template(centers, widths, networks_cmu, display_mode='lyrz', node_size=10, title=None):
    node_codes = node_labels(centers, widths, networks_cmu)
    display = niplot.plot_connectome(
        np.eye(centers.shape[0]),
        centers,
        node_size=node_size,
        node_color=[colors[i] for i in node_codes['code']],
        display_mode=display_mode,
        title=title,
    )
    plt.show()
    plt.clf()
    return node_codes, display

def plot_network_legend(node_codes, out_file=None):
    network_legend_list = list(network_colors.keys())
    network_legend_list.reverse()

    fig = plt.figure(figsize=(4, 3))
    ax = plt.gca()

    sns.histplot(
        data=node_codes,
        x='code',
        color='k',
        shrink=0.75,
        hue='code',
        hue_order=range(1, 8),
        palette=colors[1:],
        alpha=1.0,
        legend=False,
        ax=ax,
        discrete=True,
        edgecolor=None
    )
    ax.set_xlabel('Network', fontsize=12)
    ax.set_ylabel('Number of nodes', fontsize=12)
    ax.set_xticks([])
    ax.spines[['right', 'top']].set_visible(False)
    ax.legend(
        network_legend_list,
        loc='upper center',
        bbox_to_anchor=(0.6, -0.1),
        fancybox=True,
        ncol=4
    )
    plt.show()
    if out_file is not None:
        fig.savefig(out_file, bbox_inches='tight')
    plt.clf()
    plt.close(fig)
```

## Timecourse-archetype similarity clustering

```

def purity_score(y_cluster, y_class):
    y_cluster = np.asarray(y_cluster)
    y_class = np.asarray(y_class)
    total = 0
    for c in np.unique(y_cluster):
        mask = (y_cluster == c)
        _, counts = np.unique(y_class[mask], return_counts=True)
        total += counts.max()
    return total / len(y_cluster)

def get_timecourse_archetype_subject_matrix(results_subj, k, zscore_each=True, eps=1e-8):
    Xk = np.stack([np.asarray(sub["sXC"], dtype=float)[:, k] for sub in results_subj], axis=0)
    if zscore_each:
        mu = Xk.mean(axis=1, keepdims=True)
        sd = Xk.std(axis=1, keepdims=True)
        Xk = (Xk - mu) / (sd + eps)
    Xk = np.nan_to_num(Xk, nan=0.0, posinf=0.0, neginf=0.0)
    return Xk

def cluster_from_timecourse_similarity(Xk, cond, n_clusters=3, linkage_method="average"):
    Xk = np.asarray(Xk, dtype=float)
    Xk = np.nan_to_num(Xk, nan=0.0, posinf=0.0, neginf=0.0)

    sim = np.corrcoef(Xk)
    sim = np.nan_to_num(sim, nan=0.0, posinf=0.0, neginf=0.0)
    sim = np.clip(sim, -1.0, 1.0)
    np.fill_diagonal(sim, 1.0)

    dist = 1.0 - sim
    dist = np.nan_to_num(dist, nan=1.0, posinf=1.0, neginf=1.0)
    np.fill_diagonal(dist, 0.0)

    Z = linkage(squareform(dist, checks=False), method=linkage_method)
    order = leaves_list(Z)
    labels = fcluster(Z, n_clusters, criterion="maxclust")

    sim_sorted = sim[np.ix_(order, order)]
    cond_sorted = np.asarray(cond)[order]
    labels_sorted = labels[order]
    boundaries = np.where(np.diff(labels_sorted) != 0)[0] + 1

    return {
        "sim": sim,
        "sim_sorted": sim_sorted,
        "order": order,
        "labels": labels,
        "labels_sorted": labels_sorted,
        "cond_sorted": cond_sorted,
        "boundaries": boundaries,
        "purity": purity_score(labels, cond),
        "ari": adjusted_rand_score(cond, labels),
        "nmi": normalized_mutual_info_score(cond, labels),
    }

def scan_archetypes_timecourse_similarity(results_subj, cond, n_clusters=3, linkage_method="average"):
    K = np.asarray(results_subj[0]["sXC"]).shape[1]
    rows, results = [], {}
    for k in range(K):
        Xk = get_timecourse_archetype_subject_matrix(results_subj, k, zscore_each=True)
        result = cluster_from_timecourse_similarity(Xk, cond=cond, n_clusters=n_clusters, linkage_method=linkage_method)
        rows.append({"k": k, "purity": result["purity"], "ari": result["ari"], "nmi": result["nmi"]})
        results[k] = result
    df = pd.DataFrame(rows).sort_values("ari", ascending=False).reset_index(drop=True)
    return df, results

def plot_timecourse_similarity(result, cond_colors=None, title="", show_subject_ids=True):
    if cond_colors is None:
        cond_colors = {0: "purple", 1: "green", 2: "black"}

    order = result["order"]
    cond_sorted = result["cond_sorted"]
    xticklabels = (order + 1) if show_subject_ids else False
    yticklabels = (order + 1) if show_subject_ids else False

    plt.figure(figsize=(7, 6))
    ax = sns.heatmap(
        result["sim_sorted"],
        cmap="coolwarm",
        vmin=-1,
        vmax=1,
        square=True,
        xticklabels=xticklabels,
        yticklabels=yticklabels,
        cbar_kws={"label": "subject correlation of archetype timecourses"},
    )

    if show_subject_ids:
        tick_colors = [cond_colors.get(int(c), "gray") for c in cond_sorted]
        for tick_label, color in zip(ax.get_xticklabels(), tick_colors):
            tick_label.set_color(color)
            tick_label.set_rotation(90)
            tick_label.set_fontsize(8)
        for tick_label, color in zip(ax.get_yticklabels(), tick_colors):
            tick_label.set_color(color)
            tick_label.set_fontsize(8)

    for b in result["boundaries"]:
        ax.axhline(b, color="black", linewidth=2)
        ax.axvline(b, color="black", linewidth=2)

    plt.title(f"{title}\nPur={result['purity']:.2f}, ARI={result['ari']:.2f}, NMI={result['nmi']:.2f}")
    plt.tight_layout()
    plt.show()
```

## Brain plots using Yeo-based network coloring

```

def plot_network_colored_spatial_coeff_map(results_subj, k, centers, widths, networks_cmu, title="", display_mode="lyrz", node_size=10, thr_frac=0.30):
    if not NILEARN_AVAILABLE:
        print("nilearn not available; skipping.")
        return None

    coeffs = np.stack([np.asarray(sub["S"], dtype=float)[k, :] for sub in results_subj], axis=0)
    weights = coeffs.mean(axis=0)
    weights = np.nan_to_num(weights, nan=0.0, posinf=0.0, neginf=0.0)

    keep = weights >= (thr_frac * np.max(weights))
    centers_sel = centers[keep]

    node_codes_full = node_labels(centers, widths, networks_cmu)
    codes_sel = node_codes_full.loc[keep, 'code'].to_numpy()

    display = niplot.plot_connectome(
        np.eye(centers_sel.shape[0]),
        centers_sel,
        node_size=node_size,
        node_color=[colors[i] for i in codes_sel],
        display_mode=display_mode,
        title=title,
    )
    plt.show()
    plt.clf()
    return display

def save_network_colored_spatial_coeff_map(results_subj, k, centers, widths, networks_cmu, out_file, title="", display_mode="lyrz", node_size=10, thr_frac=0.30):
    if not NILEARN_AVAILABLE:
        print("nilearn not available; skipping.")
        return None

    coeffs = np.stack([np.asarray(sub["S"], dtype=float)[k, :] for sub in results_subj], axis=0)
    weights = coeffs.mean(axis=0)
    weights = np.nan_to_num(weights, nan=0.0, posinf=0.0, neginf=0.0)

    keep = weights >= (thr_frac * np.max(weights))
    centers_sel = centers[keep]

    node_codes_full = node_labels(centers, widths, networks_cmu)
    codes_sel = node_codes_full.loc[keep, 'code'].to_numpy()

    fig = plt.figure(figsize=(6, 5))
    display = niplot.plot_connectome(
        np.eye(centers_sel.shape[0]),
        centers_sel,
        node_size=node_size,
        node_color=[colors[i] for i in codes_sel],
        display_mode=display_mode,
        title=title,
        figure=fig,
    )
    fig.savefig(out_file, dpi=150, bbox_inches='tight')
    try:
        display.close()
    except Exception:
        pass
    plt.clf()
    plt.close(fig)
    return out_file
```

```

# Plot template and legend once
first_K = K_VALUES[0]
cur0 = loaded[first_K]
node_codes0, _disp = plot_network_template(
    centers=cur0["centers"],
    widths=cur0["widths"],
    networks_cmu=networks_cmu,
    display_mode="lyrz",
    node_size=10,
    title="Yeo network-colored node template"
)
plot_network_legend(node_codes0)
try:
    _disp.close()
except Exception:
    pass
```

```

top_cluster_ks_by_k = {}

for K in K_VALUES:
    cur = loaded[K]
    cluster_df, cluster_results = scan_archetypes_timecourse_similarity(
        cur["results_subj"],
        cond=cur["condition_codes"],
        n_clusters=N_CLUSTERS,
        linkage_method="average",
    )
    top_cluster_ks = cluster_df.head(TOP_N_CLUSTER_ARCHETYPES)["k"].astype(int).tolist()
    top_cluster_ks_by_k[K] = top_cluster_ks

    print(f"\n===== K={K} =====")
    display(cluster_df.head(10))
    print("Top 3 clustering archetypes:", top_cluster_ks)

    top_k = top_cluster_ks[0]
    plot_timecourse_similarity(
        cluster_results[top_k],
        cond_colors=COND_COLORS,
        title=f"K={K} | archetype {top_k} | timecourse clustering"
    )

    _disp = plot_network_colored_spatial_coeff_map(
        cur["results_subj"],
        top_k,
        centers=cur["centers"],
        widths=cur["widths"],
        networks_cmu=networks_cmu,
        title=f"K={K} | archetype {top_k} | spatial coefficients",
        display_mode="lyrz",
        node_size=10,
        thr_frac=0.30,
    )
    try:
        _disp.close()
    except Exception:
        pass

top_cluster_ks_by_k
```
