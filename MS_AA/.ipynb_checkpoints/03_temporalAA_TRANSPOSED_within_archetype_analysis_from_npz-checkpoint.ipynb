
# 03_temporalAA_TRANSPOSED_within_archetype_analysis_from_npz.ipynb

This notebook mirrors the **across-condition temporal archetype analysis notebook**, but for
**within-condition temporal AA** fits.

It assumes the fits were created with:
- `ANALYSIS_TYPE = "temporal"`
- `FIT_SCOPE = "within"`

## Conventions

Temporal AA here means:
- `sXC`: **V x K** spatial motifs
- `S`: **K x T** timecourse coefficients

## Analysis choices mirrored from the current temporal notebook

- clustering uses the **spatial motifs** `sXC[:, k]` across subjects
- decoding uses reconstructed data from archetype `k` alone
- coefficient timecourse plot is retained
- subject × time coefficient heatmap is retained with condition-colored subject labels
- spatial motif map uses a **signed diverging colormap**
- network summary is a **bar chart of average archetype value within each network**


```

LOAD_DIR = "msaa_flexible_outputs_npz"
K_VALUES = [10, 25, 50]
CONDITIONS = ["intact", "word", "rest"]

CLUSTER_RANGE = range(2, 9)
TOP_N_REPORT = 3
PENALTY_MODE = "sqrt"

NFOLDS_DECODE = 2
NREPS_DECODE = 20
RNG_SEED = 42

COND_COLORS = {"intact": "purple", "word": "green", "rest": "black"}

SCHAEFER_TXT = "data/pieman/raw/Schaefer2018_1000Parcels_7Networks_order.txt"
SCHAEFER_NII = "data/pieman/raw/Schaefer2018_1000Parcels_7Networks_order_FSLMNI152_2mm.nii.gz"
POSTERIOR_MAT = "data/pieman/raw/pieman_posterior_K700.mat"

OUTPUT_DIR = "temporalAA_TRANSPOSED_within_analysis_outputs"
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```

%matplotlib inline
import os, warnings
import numpy as np, pandas as pd, seaborn as sns, matplotlib.pyplot as plt
from scipy.io import loadmat
from scipy.spatial.distance import cdist
from sklearn.cluster import SpectralClustering
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score
from matplotlib import cm, colors as mcolors

try:
    import nibabel as nib
    from nilearn.input_data import NiftiMasker
    from nilearn import plotting as niplot
    NILEARN_AVAILABLE = True
except Exception:
    NILEARN_AVAILABLE = False
    print("nilearn / nibabel not available; brain plotting disabled.")

print("Ready.")
```

```

def to_float_array(x):
    return np.array(x, dtype=float)

def load_msaa_npz(path):
    data = np.load(path, allow_pickle=True)
    results_subj = data["results_subj"].tolist()
    if isinstance(results_subj, np.ndarray):
        results_subj = results_subj.tolist()
    return {
        "K": int(data["K"]),
        "results_subj": results_subj,
        "condition_labels_str": data["condition_labels_str"].tolist(),
        "condition_codes": data["condition_codes"],
    }

loaded = {}
for cond_name in CONDITIONS:
    loaded[cond_name] = {}
    for K in K_VALUES:
        path = os.path.join(LOAD_DIR, f"temporalAA_within_{cond_name}_K{K}.npz")
        loaded[cond_name][K] = load_msaa_npz(path)
        print("Loaded:", path)

# diagnostic
example_sub = loaded[CONDITIONS[0]][K_VALUES[0]]["results_subj"][0]
print("sXC dtype:", to_float_array(example_sub["sXC"]).dtype, "shape:", to_float_array(example_sub["sXC"]).shape)
print("S dtype:", to_float_array(example_sub["S"]).dtype, "shape:", to_float_array(example_sub["S"]).shape)
```

```

posterior = loadmat(POSTERIOR_MAT)
centers = to_float_array(posterior['posterior']['centers'][0][0][0][0][0])
widths = to_float_array(list(posterior['posterior']['widths'][0][0][0][0][0][:, 0].T)).ravel()

lookup_table = {
    'Vis':'Visual',
    'SomMot':'Somatomotor',
    'DorsAttn':'Dorsal attention',
    'SalVentAttn':'Ventral attention',
    'Limbic':'Limbic',
    'Cont':'Frontoparietal',
    'Default':'Default mode'
}
network_colors = {
    'Visual':'#D7DF23',
    'Somatomotor':'#39B54A',
    'Dorsal attention':'#00A79D',
    'Ventral attention':'#27AAE1',
    'Limbic':'#1C75BC',
    'Frontoparietal':'#92278F',
    'Default mode':'#EE2A7B'
}
network_codes = {k: i + 1 for i, k in enumerate(lookup_table.values())}

def nii2cmu(nifti_file, mask_file=None):
    def fullfact(dims):
        vals = np.asmatrix(range(1, dims[0] + 1)).T
        if len(dims) == 1:
            return vals
        aftervals = np.asmatrix(fullfact(dims[1:]))
        inds = np.asmatrix(np.zeros((np.prod(dims), len(dims))))
        row = 0
        for i in range(aftervals.shape[0]):
            inds[row:(row + len(vals)), 0] = vals
            inds[row:(row + len(vals)), 1:] = np.tile(aftervals[i, :], (len(vals), 1))
            row += len(vals)
        return inds

    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        img = nib.load(nifti_file) if type(nifti_file) == str else nifti_file
        mask = NiftiMasker(mask_strategy='background')
        mask.fit(nifti_file if mask_file is None else mask_file)

    Saff = img.get_sform()
    Y = np.float32(mask.transform(nifti_file)).copy()
    vmask = np.nonzero(np.array(np.reshape(mask.mask_img_.dataobj, (1, np.prod(mask.mask_img_.shape)), order='C')))[1]
    vox_coords = fullfact(img.shape[0:3])[vmask, ::-1] - 1
    R = np.array(np.dot(vox_coords, Saff[0:3, 0:3])) + Saff[:3, 3]
    return {'Y': Y, 'R': R}

def rbf(R, center, width):
    return np.exp(-np.sum((R - center) ** 2, axis=1) / width)

def node_labels(centers, widths, networks_cmu):
    labels = []
    for c, w in zip(centers, widths):
        r = rbf(networks_cmu['R'], c, w)
        label_weights = [sum(r[networks_cmu['Y'].ravel() == i]) for i in range(1, len(network_codes)+1)]
        labels.append(np.argmax(label_weights)+1)
    return pd.DataFrame({
        'code': labels,
        'Network': [list(lookup_table.values())[i-1] for i in labels]
    })

key = pd.read_csv(SCHAEFER_TXT, sep='\t', header=None, names=['id','name','x','y','z','t']).drop('t', axis=1)
key['network'] = key['name'].apply(lambda x: lookup_table[x.split('_')[2]])
key['code'] = key['network'].apply(lambda x: network_codes[x])
key.set_index('id', inplace=True)
key.loc[0, 'code'] = 0

networks_cmu = nii2cmu(SCHAEFER_NII)
networks_cmu['Y'] = np.atleast_2d(np.array([key.loc[i, 'code'] for i in networks_cmu['Y']]).astype(float))
node_codes = node_labels(centers, widths, networks_cmu)
display(node_codes.head())
```

```

def decoder(corrs):
    out = pd.DataFrame({'rank':[0.0], 'accuracy':[0.0], 'error':[0.0]})
    T = corrs.shape[0]
    for t in range(T):
        decoded_ind = int(np.argmax(corrs[t, :]))
        out.loc[0, 'error'] += np.mean(np.abs(decoded_ind - t)) / T
        out.loc[0, 'accuracy'] += (decoded_ind == t)
        out.loc[0, 'rank'] += np.mean((corrs[t, :] <= corrs[t, t]).astype(int))
    out['error'] /= T
    out['accuracy'] /= T
    out['rank'] /= T
    return out

def get_xval_assignments(ndata, nfolds, rng=None):
    rng = np.random.default_rng() if rng is None else rng
    group_assignments = np.zeros(ndata, dtype=int)
    groupsize = int(np.ceil(ndata / nfolds))
    for i in range(1, nfolds):
        inds = np.arange(i * groupsize, min((i + 1) * groupsize, ndata))
        group_assignments[inds] = i
    rng.shuffle(group_assignments)
    return group_assignments

def build_recon_stack(subjects, top_indices=None):
    Xhats = []
    for sub in subjects:
        sXC = to_float_array(sub["sXC"])
        S = to_float_array(sub["S"])
        Yhat = sXC @ S if top_indices is None else sXC[:, top_indices] @ S[top_indices, :]
        Xhat = Yhat.T
        Xhat = (Xhat - Xhat.mean(axis=0, keepdims=True)) / (Xhat.std(axis=0, keepdims=True) + 1e-8)
        Xhats.append(Xhat)
    return np.stack(Xhats, axis=0)

def run_timepoint_decoding(recon_stack, nfolds=2, nreps=20, seed=42):
    N, T, V = recon_stack.shape
    rng = np.random.default_rng(seed)
    all_results = []
    for rep in range(nreps):
        fold_ids = get_xval_assignments(N, nfolds, rng=rng)
        for i in range(nfolds):
            in_mask = (fold_ids == i)
            out_mask = ~in_mask
            in_mean = recon_stack[in_mask].mean(axis=0)
            out_mean = recon_stack[out_mask].mean(axis=0)
            corrs = 1.0 - cdist(in_mean, out_mean, metric='correlation')
            res = decoder(corrs)
            res["rep"] = rep
            res["fold"] = i
            all_results.append(res)
    return pd.concat(all_results, ignore_index=True)

def summarize_single_archetype_decoding(results_subj, k, nfolds=2, nreps=20, seed=42):
    recon_stack = build_recon_stack(results_subj, top_indices=[k])
    dec_df = run_timepoint_decoding(recon_stack, nfolds=nfolds, nreps=nreps, seed=seed)
    summary = dec_df["accuracy"].agg(["mean","std","count"])
    sem = summary["std"] / np.sqrt(max(summary["count"], 1))
    return dec_df, float(summary["mean"]), float(sem)
```

```

def purity_score(y_cluster, y_class):
    total = 0
    for c in np.unique(y_cluster):
        mask = (y_cluster == c)
        _, counts = np.unique(np.asarray(y_class)[mask], return_counts=True)
        total += counts.max()
    return total / len(y_cluster)

def equal_size_balance(labels):
    _, counts = np.unique(labels, return_counts=True)
    N = counts.sum()
    k = len(counts)
    ideal = N / k
    return float(1.0 - np.sum(np.abs(counts - ideal)) / (2 * N))

def cluster_count_penalty(n_clusters, mode="sqrt"):
    return 1.0 / np.sqrt(n_clusters) if mode == "sqrt" else (1.0 / n_clusters if mode == "linear" else 1.0)

def score_clustering_solution(labels, pseudo_labels, penalty_mode="sqrt"):
    n_clusters = len(np.unique(labels))
    purity = purity_score(labels, pseudo_labels)
    balance = equal_size_balance(labels)
    penalty = cluster_count_penalty(n_clusters, penalty_mode)
    _, counts = np.unique(labels, return_counts=True)
    return {
        "n_clusters": n_clusters,
        "purity": purity,
        "balance": balance,
        "penalty": penalty,
        "score": purity * balance * penalty,
        "cluster_sizes": counts.tolist()
    }

def get_spatial_motif_subject_matrix(results_subj, k, eps=1e-8):
    Xk = np.stack([to_float_array(sub["sXC"])[:, k] for sub in results_subj], axis=0)  # N x V
    mu = Xk.mean(axis=1, keepdims=True)
    sd = Xk.std(axis=1, keepdims=True)
    Xk = (Xk - mu) / (sd + eps)
    return np.nan_to_num(Xk, nan=0.0, posinf=0.0, neginf=0.0)

def reorder_by_labels_and_within_similarity(sim, labels):
    unique_labels = np.unique(labels)
    order = []
    for lab in unique_labels:
        idx = np.where(labels == lab)[0]
        sub_sim = sim[np.ix_(idx, idx)]
        sub_order = idx[np.argsort(-sub_sim.mean(axis=1))]
        order.extend(sub_order.tolist())
    order = np.array(order)
    labels_sorted = labels[order]
    boundaries = np.where(np.asarray(labels_sorted[1:]) != np.asarray(labels_sorted[:-1]))[0] + 1
    return order, labels_sorted, boundaries

def make_result_from_labels(sim, labels):
    order, labels_sorted, boundaries = reorder_by_labels_and_within_similarity(sim, labels)
    return {
        "sim": sim,
        "sim_sorted": sim[np.ix_(order, order)],
        "order": order,
        "labels": labels,
        "labels_sorted": labels_sorted,
        "boundaries": boundaries,
    }

def cluster_spectral(sim, n_clusters=3, random_state=0):
    sim = np.asarray(sim, dtype=float)
    sim = np.nan_to_num(sim, nan=0.0, posinf=0.0, neginf=0.0)
    sim = np.clip(sim, -1.0, 1.0)
    aff = (sim + 1.0) / 2.0
    np.fill_diagonal(aff, 1.0)
    model = SpectralClustering(
        n_clusters=n_clusters,
        affinity="precomputed",
        assign_labels="kmeans",
        random_state=random_state
    )
    labels = model.fit_predict(aff) + 1
    return make_result_from_labels(sim, labels)

def scan_spectral_cluster_numbers_for_archetype(Xk, sim, cluster_range=range(2, 9), penalty_mode="sqrt", random_state=0):
    # pseudo-labels encourage near-equal partitions for within-condition runs
    N = Xk.shape[0]
    rows = []
    results = {}
    for n_clusters in cluster_range:
        res = cluster_spectral(sim, n_clusters=n_clusters, random_state=random_state)
        labels = res["labels"]
        # for within-condition, report balance and penalty; purity against cluster ids is trivially 1 if reused,
        # so use a pseudo-label vector based on equal-size bins to keep the combined score meaningful.
        pseudo = np.repeat(np.arange(n_clusters), np.ceil(N / n_clusters))[:N]
        scored = score_clustering_solution(labels, pseudo, penalty_mode)
        rows.append({
            "n_clusters": n_clusters,
            "purity": scored["purity"],
            "balance": scored["balance"],
            "penalty": scored["penalty"],
            "score": scored["score"],
            "cluster_sizes": scored["cluster_sizes"]
        })
        results[n_clusters] = res
    return pd.DataFrame(rows).sort_values("score", ascending=False).reset_index(drop=True), results
```

```

def analyze_archetypes_for_condition(cond_name, cur, K_value):
    results_subj = cur["results_subj"]
    n_archetypes = to_float_array(results_subj[0]["S"]).shape[0]

    rows = []
    cluster_store = {}
    decoding_store = {}

    for k in range(n_archetypes):
        Xk = get_spatial_motif_subject_matrix(results_subj, k)
        sim = np.corrcoef(Xk)
        sim = np.nan_to_num(sim, nan=0.0, posinf=0.0, neginf=0.0)
        sim = np.clip(sim, -1.0, 1.0)
        np.fill_diagonal(sim, 1.0)

        scan_df, scan_results = scan_spectral_cluster_numbers_for_archetype(
            Xk, sim, CLUSTER_RANGE, PENALTY_MODE, 0
        )
        best_cluster_row = scan_df.iloc[0]

        dec_df, dec_mean, dec_sem = summarize_single_archetype_decoding(
            results_subj, k, NFOLDS_DECODE, NREPS_DECODE, RNG_SEED
        )

        row = {
            "condition": cond_name,
            "K": K_value,
            "archetype": k,
            "best_n_clusters": int(best_cluster_row["n_clusters"]),
            "purity": float(best_cluster_row["purity"]),
            "balance": float(best_cluster_row["balance"]),
            "cluster_score": float(best_cluster_row["score"]),
            "cluster_sizes": best_cluster_row["cluster_sizes"],
            "decode_accuracy_overall": float(dec_mean),
            "decode_sem": float(dec_sem),
        }

        rows.append(row)
        cluster_store[k] = {"scan_df": scan_df, "scan_results": scan_results, "Xk": Xk}
        decoding_store[k] = {"dec_df": dec_df, "dec_mean": dec_mean, "dec_sem": dec_sem}

    summary_df = pd.DataFrame(rows)
    summary_df["cluster_rank"] = summary_df["cluster_score"].rank(ascending=False, method="average")
    summary_df["decode_rank"] = summary_df["decode_accuracy_overall"].rank(ascending=False, method="average")
    summary_df["joint_rank_score"] = -(summary_df["cluster_rank"] + summary_df["decode_rank"])
    return {"summary_df": summary_df, "cluster_store": cluster_store, "decoding_store": decoding_store}

all_condition_analysis = {}
all_rows = []

for cond_name in CONDITIONS:
    all_condition_analysis[cond_name] = {}
    for K in K_VALUES:
        res = analyze_archetypes_for_condition(cond_name, loaded[cond_name][K], K)
        all_condition_analysis[cond_name][K] = res
        all_rows.append(res["summary_df"])
        print(f"Condition={cond_name} | K={K}")
        display(res["summary_df"].sort_values(
            ["joint_rank_score","cluster_score","decode_accuracy_overall"],
            ascending=[False, False, False]
        ).head(10))

archetype_summary_all_df = pd.concat(all_rows, ignore_index=True)
display(archetype_summary_all_df.head())
```

```

def plot_cluster_scan_metrics(scan_df, archetype_k=None):
    plot_df = scan_df.sort_values("n_clusters")
    plt.figure(figsize=(8, 5))
    plt.plot(plot_df["n_clusters"], plot_df["purity"], marker="o", label="Purity")
    plt.plot(plot_df["n_clusters"], plot_df["balance"], marker="o", label="Balance")
    plt.plot(plot_df["n_clusters"], plot_df["score"], marker="o", label="Combined score")
    plt.xlabel("Number of clusters")
    plt.ylabel("Score")
    title = "Flexible spectral cluster-number scan"
    if archetype_k is not None:
        title += f" | archetype {archetype_k}"
    plt.title(title)
    plt.legend()
    plt.tight_layout()
    plt.show()

def plot_clustered_similarity(result, title="", show_subject_ids=True):
    order = result["order"]
    xticklabels = (order + 1) if show_subject_ids else False
    yticklabels = (order + 1) if show_subject_ids else False

    plt.figure(figsize=(7, 6))
    ax = sns.heatmap(
        result["sim_sorted"],
        cmap="coolwarm",
        vmin=-1,
        vmax=1,
        square=True,
        xticklabels=xticklabels,
        yticklabels=yticklabels,
        cbar_kws={"label": "subject correlation of spatial archetype values"}
    )
    if show_subject_ids:
        for tick_label in ax.get_xticklabels():
            tick_label.set_rotation(90)
            tick_label.set_fontsize(8)
        for tick_label in ax.get_yticklabels():
            tick_label.set_fontsize(8)
    for b in result["boundaries"]:
        ax.axhline(b, color="black", linewidth=2)
        ax.axvline(b, color="black", linewidth=2)
    plt.title(title)
    plt.tight_layout()
    plt.show()

def plot_mean_timecourse_coeff(results_subj, k, cond_name, title=""):
    coeffs = np.stack([to_float_array(sub["S"])[k, :] for sub in results_subj], axis=0)
    mean = coeffs.mean(axis=0)
    sem = coeffs.std(axis=0) / np.sqrt(max(coeffs.shape[0], 1))

    plt.figure(figsize=(9, 4))
    plt.plot(np.arange(len(mean)), mean, color=COND_COLORS[cond_name], label=cond_name)
    plt.fill_between(np.arange(len(mean)), mean - sem, mean + sem, color=COND_COLORS[cond_name], alpha=0.2)
    plt.xlabel("Time")
    plt.ylabel("Coefficient value")
    plt.title(title if title else f"Archetype {k}")
    plt.legend()
    plt.tight_layout()
    plt.show()

def plot_subject_time_heatmap(results_subj, k, cond_name, title=""):
    Xk = np.stack([to_float_array(sub["S"])[k, :] for sub in results_subj], axis=0)
    labels = np.array([cond_name] * Xk.shape[0])

    plt.figure(figsize=(10, 5))
    ax = sns.heatmap(Xk, cmap="coolwarm", center=0)
    plt.xlabel("Time")
    plt.ylabel("Subjects")
    plt.title(title if title else f"Archetype {k}")

    for tick_label in ax.get_yticklabels():
        tick_label.set_color(COND_COLORS.get(cond_name, "gray"))
        tick_label.set_fontsize(8)

    boundaries = np.where(np.asarray(labels[1:]) != np.asarray(labels[:-1]))[0] + 1
    for b in boundaries:
        ax.axhline(b, color="black", linewidth=1.5)

    plt.tight_layout()
    plt.show()

def plot_spatial_motif_map_diverging(results_subj, k, title="", cmap_name="coolwarm"):
    if not NILEARN_AVAILABLE:
        print("nilearn not available; skipping.")
        return None

    sXC_all = np.stack([to_float_array(sub["sXC"])[:, k] for sub in results_subj], axis=0)
    weights = np.nan_to_num(sXC_all.mean(axis=0), nan=0.0, posinf=0.0, neginf=0.0)

    vmax = np.max(np.abs(weights))
    norm = mcolors.TwoSlopeNorm(vmin=-vmax, vcenter=0.0, vmax=vmax)
    cmap = cm.get_cmap(cmap_name)

    node_colors = [cmap(norm(w)) for w in weights]
    node_sizes = 8 + 18 * (np.abs(weights) / (vmax + 1e-8))

    display = niplot.plot_connectome(
        np.eye(centers.shape[0]),
        centers,
        node_color=node_colors,
        node_size=node_sizes,
        display_mode="lyrz",
        title=title
    )
    plt.show()
    plt.clf()
    return display

def plot_network_mean_bar(results_subj, k, title="Average archetype value by network"):
    sXC_all = np.stack([to_float_array(sub["sXC"])[:, k] for sub in results_subj], axis=0)
    weights = np.nan_to_num(sXC_all.mean(axis=0), nan=0.0, posinf=0.0, neginf=0.0)

    network_df = node_labels(centers, widths, networks_cmu).copy()
    network_df["value"] = weights

    means = network_df.groupby("Network")["value"].mean().reindex(list(network_colors.keys()))
    bar_colors = [network_colors[n] for n in means.index]

    plt.figure(figsize=(7.5, 4))
    plt.bar(means.index, means.values, color=bar_colors)
    plt.axhline(0, color="black", linewidth=1)
    plt.ylabel("Mean archetype value")
    plt.title(title)
    plt.xticks(rotation=35, ha="right")
    plt.tight_layout()
    plt.show()

def plot_decoding_bar(dec_mean, dec_sem, cond_name, K, k):
    plt.figure(figsize=(4, 4))
    plt.bar([cond_name], [dec_mean], yerr=[dec_sem], color=[COND_COLORS[cond_name]], capsize=4)
    plt.ylabel("Decoding accuracy")
    plt.title(f"{cond_name} | K={K} | archetype {k}")
    plt.tight_layout()
    plt.show()
```

```

top_within_temporal_ks = {}

for cond_name in CONDITIONS:
    top_within_temporal_ks[cond_name] = {}
    for K in K_VALUES:
        print(f"\n==================== Condition={cond_name} | K={K} ====================")
        cur = loaded[cond_name][K]
        analysis_obj = all_condition_analysis[cond_name][K]
        df = analysis_obj["summary_df"].copy()

        top_df = df.sort_values("joint_rank_score", ascending=False).head(TOP_N_REPORT)
        top_within_temporal_ks[cond_name][K] = top_df["archetype"].astype(int).tolist()

        for _, row in top_df.iterrows():
            k = int(row["archetype"])
            best_n_clusters = int(row["best_n_clusters"])
            best_cluster_result = analysis_obj["cluster_store"][k]["scan_results"][best_n_clusters]
            dec_mean = analysis_obj["decoding_store"][k]["dec_mean"]
            dec_sem = analysis_obj["decoding_store"][k]["dec_sem"]

            plot_cluster_scan_metrics(analysis_obj["cluster_store"][k]["scan_df"], archetype_k=k)
            plot_clustered_similarity(
                best_cluster_result,
                title=f"{cond_name} | K={K} | Temporal AA archetype {k} | spectral clustering on spatial motifs"
            )
            plot_decoding_bar(dec_mean, dec_sem, cond_name, K, k)
            plot_mean_timecourse_coeff(
                cur["results_subj"], k, cond_name,
                title=f"{cond_name} | K={K} | archetype {k} | mean timecourse coefficient"
            )
            plot_subject_time_heatmap(
                cur["results_subj"], k, cond_name,
                title=f"{cond_name} | K={K} | archetype {k} | subject × time coefficient heatmap"
            )

            disp = plot_spatial_motif_map_diverging(
                cur["results_subj"], k,
                title=f"{cond_name} | K={K} | archetype {k} | spatial motif map (signed)"
            )
            if disp is not None:
                try:
                    disp.close()
                except Exception:
                    pass

            plot_network_mean_bar(
                cur["results_subj"], k,
                title=f"{cond_name} | K={K} | archetype {k} | mean value by network"
            )

            display(row.to_frame().T)

top_within_temporal_ks
```
