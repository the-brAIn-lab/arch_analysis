
# 03_clustering_and_brainplots_from_npz_exact_notebook_style.ipynb

This notebook rebuilds the clustering + brain-plot workflow using the **same network-labeling code structure**
from the uploaded notebook, rather than a simplified recreation.

Core split:
- clustering is based on **timecourse archetype similarity**
- brain plots use the corresponding **spatial coefficients**
- node labels are generated using the same:
  - `Schaefer2018_1000Parcels_7Networks_order.txt`
  - `Schaefer2018_1000Parcels_7Networks_order_FSLMNI152_2mm.nii.gz`
  - `pieman_posterior_K700.mat`
  - `nii2cmu(...)`
  - `rbf(...)`
  - `node_labels(...)`

It also includes:
- notebook-style network template + legend
- network-colored plots
- coefficient-opacity plots
- pie charts for selected-node network composition


```

LOAD_DIR = "msaa_outputs_npz"
K_VALUES = [10, 25]
N_CLUSTERS = 3
TOP_N_CLUSTER_ARCHETYPES = 3

COND_COLORS = {0: "purple", 1: "green", 2: "black"}

SCHAEFER_TXT = os.path.join("data/pieman/raw", "Schaefer2018_1000Parcels_7Networks_order.txt")
SCHAEFER_NII = os.path.join("data/pieman/raw", "Schaefer2018_1000Parcels_7Networks_order_FSLMNI152_2mm.nii.gz")
POSTERIOR_MAT = os.path.join("data/pieman/raw", "pieman_posterior_K700.mat")

OUTPUT_DIR = "clustering_brainplots_exact_notebook_style"
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```

%matplotlib inline
import os
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from scipy.io import loadmat
from matplotlib.colors import ListedColormap, to_rgba
from scipy.cluster.hierarchy import linkage, leaves_list, fcluster
from scipy.spatial.distance import squareform
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score

try:
    import nibabel as nib
    from nilearn.input_data import NiftiMasker
    from nilearn import plotting as niplot
    NILEARN_AVAILABLE = True
except Exception:
    NILEARN_AVAILABLE = False
    print("nilearn / nibabel not available; brain plotting disabled.")

print("Ready.")
```

```

def load_msaa_npz(path):
    data = np.load(path, allow_pickle=True)
    results_subj = data["results_subj"].tolist()
    if isinstance(results_subj, np.ndarray):
        results_subj = results_subj.tolist()
    return {
        "K": int(data["K"]),
        "results_subj": results_subj,
        "condition_labels_str": data["condition_labels_str"].tolist(),
        "condition_codes": data["condition_codes"],
    }

loaded = {}
for K in K_VALUES:
    path = os.path.join(LOAD_DIR, f"spatialAA_acrossCond_K{K}.npz")
    loaded[K] = load_msaa_npz(path)
    print("Loaded:", path)
```

## Exact network-labeling setup from the uploaded notebook

```

if not os.path.exists(POSTERIOR_MAT):
    raise FileNotFoundError(f"Missing file: {POSTERIOR_MAT}")
if not os.path.exists(SCHAEFER_TXT):
    raise FileNotFoundError(f"Missing file: {SCHAEFER_TXT}")
if not os.path.exists(SCHAEFER_NII):
    raise FileNotFoundError(f"Missing file: {SCHAEFER_NII}")

posterior = loadmat(POSTERIOR_MAT)
centers = posterior['posterior']['centers'][0][0][0][0][0]
widths = np.array(list(posterior['posterior']['widths'][0][0][0][0][0][:, 0].T))

centers = np.asarray(centers, dtype=float)
widths = np.asarray(widths, dtype=float).ravel()

print("centers shape:", centers.shape)
print("widths shape:", widths.shape)
```

```

lookup_table = {
    'Vis': 'Visual',
    'SomMot': 'Somatomotor',
    'DorsAttn': 'Dorsal attention',
    'SalVentAttn': 'Ventral attention',
    'Limbic': 'Limbic',
    'Cont': 'Frontoparietal',
    'Default': 'Default mode'
}

network_colors = {
    'Visual': '#D7DF23',
    'Somatomotor': '#39B54A',
    'Dorsal attention': '#00A79D',
    'Ventral attention': '#27AAE1',
    'Limbic': '#1C75BC',
    'Frontoparietal': '#92278F',
    'Default mode': '#EE2A7B'
}

colors = ['#888888']
colors.extend([v for k, v in network_colors.items()])
network_cmap = ListedColormap(colors, N=len(colors)*2, name='networks')

network_codes = {k: i + 1 for i, k in enumerate(lookup_table.values())}
```

```

def fullfact(levels):
    mesh = np.meshgrid(*[np.arange(1, int(l)+1) for l in levels], indexing="ij")
    return np.stack([m.ravel(order="C") for m in mesh], axis=1)

def nii2cmu(nifti_file, mask_file=None):
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")

        if isinstance(nifti_file, str):
            img = nib.load(nifti_file)
        else:
            img = nifti_file

        mask = NiftiMasker(mask_strategy='background')
        if mask_file is None:
            mask.fit(nifti_file)
        else:
            mask.fit(mask_file)

    S = img.get_sform()
    Y = np.float32(mask.transform(nifti_file)).copy()

    vmask = np.nonzero(
        np.array(np.reshape(mask.mask_img_.dataobj, (1, np.prod(mask.mask_img_.shape)), order='C'))
    )[1]

    vox_coords = fullfact(img.shape[0:3])[vmask, ::-1] - 1
    R = np.array(np.dot(vox_coords, S[0:3, 0:3])) + S[:3, 3]
    return {'Y': Y, 'R': R}

def rbf(R, center, width):
    return np.exp(-np.sum((R - center) ** 2, axis=1) / width)

def node_labels(centers, widths, networks_cmu):
    labels = []
    for c, w in zip(centers, widths):
        r = rbf(networks_cmu['R'], c, w)
        label_weights = [sum(r[networks_cmu['Y'].ravel() == i]) for i in range(1, len(network_codes) + 1)]
        labels.append(np.argmax(label_weights) + 1)

    return pd.DataFrame({
        'code': labels,
        'Network': [list(lookup_table.values())[i - 1] for i in labels]
    })
```

```

key = pd.read_csv(
    SCHAEFER_TXT,
    sep='\t',
    header=None,
    names=['id', 'name', 'x', 'y', 'z', 't']
).drop('t', axis=1)

key['study'] = key['name'].apply(lambda x: x.split('_')[0])
key['hemisphere'] = key['name'].apply(lambda x: x.split('_')[1][0])
key['network'] = key['name'].apply(lambda x: x.split('_')[2])
key.drop('name', axis=1, inplace=True)

key['network'] = key['network'].apply(lambda x: lookup_table[x])
key['code'] = key['network'].apply(lambda x: network_codes[x])
key.set_index('id', inplace=True)
key.loc[0, 'code'] = 0  # append a background (non-network) code

networks_cmu = nii2cmu(SCHAEFER_NII)

# convert voxel values in the reference image to network codes
networks_cmu['Y'] = np.atleast_2d(np.array([key.loc[i, 'code'] for i in networks_cmu['Y']]).astype(float))

node_codes = node_labels(centers, widths, networks_cmu)

display(node_codes.head())
display(node_codes["Network"].value_counts())
```

## Timecourse-archetype similarity clustering

```

def purity(y_clust, y_class):
    size_clust = np.max(y_clust)
    len_clust = len(y_clust)
    clusters_labels_try = [None] * size_clust

    for i in range(len_clust):
        index = y_clust[i]
        if clusters_labels_try[index - 1] is None:
            clusters_labels_try[index - 1] = y_class[i]
        else:
            clusters_labels_try[index - 1] = np.hstack((clusters_labels_try[index - 1], y_class[i]))

    purity_val = 0
    for c in clusters_labels_try:
        c = c.astype(int)
        if isinstance(c, np.int64):
            c = np.array([c])
        y = np.bincount(c)
        maximum = np.max(y)
        purity_val += maximum

    purity_val = purity_val / len_clust
    return purity_val

def get_timecourse_archetype_subject_matrix(results_subj, k, eps=1e-8):
    Xk = np.stack([np.asarray(sub["sXC"], dtype=float)[:, k] for sub in results_subj], axis=0)
    mu = Xk.mean(axis=1, keepdims=True)
    sd = Xk.std(axis=1, keepdims=True)
    Xk = (Xk - mu) / (sd + eps)
    Xk = np.nan_to_num(Xk, nan=0.0, posinf=0.0, neginf=0.0)
    return Xk

def cluster_from_timecourse_similarity(Xk, cond, n_clusters=3, linkage_method="average"):
    sim = np.corrcoef(Xk)
    sim = np.nan_to_num(sim, nan=0.0, posinf=0.0, neginf=0.0)
    sim = np.clip(sim, -1.0, 1.0)
    np.fill_diagonal(sim, 1.0)

    dist = 1.0 - sim
    dist = np.nan_to_num(dist, nan=1.0, posinf=1.0, neginf=1.0)
    np.fill_diagonal(dist, 0.0)

    Z = linkage(squareform(dist, checks=False), method=linkage_method)
    order = leaves_list(Z)
    labels = fcluster(Z, n_clusters, criterion="maxclust")

    sim_sorted = sim[np.ix_(order, order)]
    cond_sorted = np.asarray(cond)[order]
    labels_sorted = labels[order]
    boundaries = np.where(np.diff(labels_sorted) != 0)[0] + 1

    return {
        "sim": sim,
        "sim_sorted": sim_sorted,
        "order": order,
        "labels": labels,
        "labels_sorted": labels_sorted,
        "cond_sorted": cond_sorted,
        "boundaries": boundaries,
        "purity": purity(labels, cond),
        "ari": adjusted_rand_score(cond, labels),
        "nmi": normalized_mutual_info_score(cond, labels),
    }

def scan_archetypes_timecourse_similarity(results_subj, cond, n_clusters=3, linkage_method="average"):
    K = np.asarray(results_subj[0]["sXC"]).shape[1]
    rows, results = [], {}
    for k in range(K):
        Xk = get_timecourse_archetype_subject_matrix(results_subj, k)
        result = cluster_from_timecourse_similarity(Xk, cond=cond, n_clusters=n_clusters, linkage_method=linkage_method)
        rows.append({"k": k, "purity": result["purity"], "ari": result["ari"], "nmi": result["nmi"]})
        results[k] = result
    df = pd.DataFrame(rows).sort_values("ari", ascending=False).reset_index(drop=True)
    return df, results

def plot_timecourse_similarity(result, cond_colors=None, title="", show_subject_ids=True):
    if cond_colors is None:
        cond_colors = {0: "purple", 1: "green", 2: "black"}

    order = result["order"]
    cond_sorted = result["cond_sorted"]
    xticklabels = (order + 1) if show_subject_ids else False
    yticklabels = (order + 1) if show_subject_ids else False

    plt.figure(figsize=(7, 6))
    ax = sns.heatmap(
        result["sim_sorted"],
        cmap="coolwarm",
        vmin=-1,
        vmax=1,
        square=True,
        xticklabels=xticklabels,
        yticklabels=yticklabels,
        cbar_kws={"label": "subject correlation of archetype timecourses"},
    )

    if show_subject_ids:
        tick_colors = [cond_colors.get(int(c), "gray") for c in cond_sorted]
        for tick_label, color in zip(ax.get_xticklabels(), tick_colors):
            tick_label.set_color(color)
            tick_label.set_rotation(90)
            tick_label.set_fontsize(8)
        for tick_label, color in zip(ax.get_yticklabels(), tick_colors):
            tick_label.set_color(color)
            tick_label.set_fontsize(8)

    for b in result["boundaries"]:
        ax.axhline(b, color="black", linewidth=2)
        ax.axvline(b, color="black", linewidth=2)

    plt.title(f"{title}\nPur={result['purity']:.2f}, ARI={result['ari']:.2f}, NMI={result['nmi']:.2f}")
    plt.tight_layout()
    plt.show()
```

## Exact notebook-style network plots plus opacity and pies

```

def coefficient_to_alpha(weights, keep_mask, alpha_min=0.15, alpha_max=1.0):
    w = np.asarray(weights, dtype=float).ravel()
    w = np.nan_to_num(w, nan=0.0, posinf=0.0, neginf=0.0)
    wk = w[keep_mask]

    if len(wk) == 0:
        return np.array([])

    wmin, wmax = wk.min(), wk.max()
    if np.isclose(wmax, wmin):
        return np.full(len(wk), alpha_max)

    scaled = (wk - wmin) / (wmax - wmin)
    return alpha_min + (alpha_max - alpha_min) * scaled

def plot_network_colored_spatial_coeff_map(
    results_subj,
    k,
    title="",
    display_mode="lyrz",
    node_size=10,
    thr_frac=0.30,
    use_opacity=False,
    alpha_min=0.15,
    alpha_max=1.0,
):
    coeffs = np.stack([np.asarray(sub["S"], dtype=float)[k, :] for sub in results_subj], axis=0)
    weights = coeffs.mean(axis=0)
    weights = np.nan_to_num(weights, nan=0.0, posinf=0.0, neginf=0.0)

    wmax = np.max(weights)
    if wmax <= 0:
        print(f"Skipping archetype {k}: non-positive weights")
        return None, None, None

    keep = weights >= (thr_frac * wmax)
    centers_sel = centers[keep]
    node_codes_local = node_labels(centers, widths, networks_cmu)
    codes_sel = node_codes_local.loc[keep, 'code'].to_numpy()

    if use_opacity:
        alphas = coefficient_to_alpha(weights, keep, alpha_min=alpha_min, alpha_max=alpha_max)
        node_colors = []
        for code, alpha in zip(codes_sel, alphas):
            base = to_rgba(colors[code])
            node_colors.append((base[0], base[1], base[2], float(alpha)))
    else:
        node_colors = [colors[i] for i in codes_sel]

    display = niplot.plot_connectome(
        np.eye(centers_sel.shape[0]),
        centers_sel,
        node_size=node_size,
        node_color=node_colors,
        display_mode=display_mode,
        title=title,
    )
    plt.show()
    plt.clf()
    return display, node_codes_local, keep

def save_network_colored_spatial_coeff_map(
    results_subj,
    k,
    out_file,
    title="",
    display_mode="lyrz",
    node_size=10,
    thr_frac=0.30,
    use_opacity=False,
    alpha_min=0.15,
    alpha_max=1.0,
):
    coeffs = np.stack([np.asarray(sub["S"], dtype=float)[k, :] for sub in results_subj], axis=0)
    weights = coeffs.mean(axis=0)
    weights = np.nan_to_num(weights, nan=0.0, posinf=0.0, neginf=0.0)

    wmax = np.max(weights)
    if wmax <= 0:
        print(f"Skipping archetype {k}: non-positive weights")
        return None

    keep = weights >= (thr_frac * wmax)
    centers_sel = centers[keep]
    node_codes_local = node_labels(centers, widths, networks_cmu)
    codes_sel = node_codes_local.loc[keep, 'code'].to_numpy()

    if use_opacity:
        alphas = coefficient_to_alpha(weights, keep, alpha_min=alpha_min, alpha_max=alpha_max)
        node_colors = []
        for code, alpha in zip(codes_sel, alphas):
            base = to_rgba(colors[code])
            node_colors.append((base[0], base[1], base[2], float(alpha)))
    else:
        node_colors = [colors[i] for i in codes_sel]

    fig = plt.figure(figsize=(6, 5))
    display = niplot.plot_connectome(
        np.eye(centers_sel.shape[0]),
        centers_sel,
        node_size=node_size,
        node_color=node_colors,
        display_mode=display_mode,
        title=title,
        figure=fig,
    )
    fig.savefig(out_file, dpi=150, bbox_inches='tight')
    try:
        display.close()
    except Exception:
        pass
    plt.clf()
    plt.close(fig)
    return out_file

def plot_network_pie_for_selected_nodes(node_codes_local, keep_mask, title="Network composition", out_file=None):
    selected = node_codes_local.loc[keep_mask].copy()
    counts = selected["Network"].value_counts().reindex(list(network_colors.keys()), fill_value=0)

    fig, ax = plt.subplots(figsize=(4.5, 4.5))
    pie_colors = [network_colors[name] for name in counts.index]
    ax.pie(
        counts.values,
        labels=counts.index,
        colors=pie_colors,
        autopct=lambda p: f"{p:.1f}%" if p > 0 else "",
        startangle=90,
        counterclock=False,
    )
    ax.set_title(title)
    plt.tight_layout()
    plt.show()

    if out_file is not None:
        fig.savefig(out_file, dpi=150, bbox_inches='tight')
    plt.close(fig)
    return counts
```

```

top_cluster_ks_by_k = {}

for K in K_VALUES:
    cur = loaded[K]

    cluster_df, cluster_results = scan_archetypes_timecourse_similarity(
        cur["results_subj"],
        cond=cur["condition_codes"],
        n_clusters=N_CLUSTERS,
        linkage_method="average",
    )
    top_cluster_ks = cluster_df.head(TOP_N_CLUSTER_ARCHETYPES)["k"].astype(int).tolist()
    top_cluster_ks_by_k[K] = top_cluster_ks

    print(f"\n===== K={K} =====")
    display(cluster_df.head(10))
    print("Top 3 clustering archetypes:", top_cluster_ks)

    for top_k in top_cluster_ks:
        plot_timecourse_similarity(
            cluster_results[top_k],
            cond_colors=COND_COLORS,
            title=f"K={K} | archetype {top_k} | timecourse clustering"
        )

        d1, node_codes_local, keep = plot_network_colored_spatial_coeff_map(
            cur["results_subj"],
            top_k,
            title=f"K={K} | archetype {top_k} | spatial coefficients (network color)",
            display_mode="lyrz",
            node_size=10,
            thr_frac=0.30,
            use_opacity=False,
        )
        try:
            d1.close()
        except Exception:
            pass

        d2, _, _ = plot_network_colored_spatial_coeff_map(
            cur["results_subj"],
            top_k,
            title=f"K={K} | archetype {top_k} | spatial coefficients (opacity)",
            display_mode="lyrz",
            node_size=10,
            thr_frac=0.30,
            use_opacity=True,
            alpha_min=0.15,
            alpha_max=1.0,
        )
        try:
            d2.close()
        except Exception:
            pass

        base = os.path.join(OUTPUT_DIR, f"K{K}_arch{top_k}")
        save_network_colored_spatial_coeff_map(
            cur["results_subj"], top_k,
            out_file=base + "_networkcolor.png",
            title=f"K={K} | archetype {top_k} | spatial coefficients (network color)",
            use_opacity=False,
        )
        save_network_colored_spatial_coeff_map(
            cur["results_subj"], top_k,
            out_file=base + "_opacity.png",
            title=f"K={K} | archetype {top_k} | spatial coefficients (opacity)",
            use_opacity=True,
            alpha_min=0.15,
            alpha_max=1.0,
        )

        counts = plot_network_pie_for_selected_nodes(
            node_codes_local,
            keep,
            title=f"K={K} | archetype {top_k} | network composition",
            out_file=base + "_pie.png",
        )
        print(counts)

top_cluster_ks_by_k
```
